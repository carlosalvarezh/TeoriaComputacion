# **Capítulo 4 - Teorema de Kleene**

Pensemos en la última vez que usamos Ctrl + F para buscar una palabra en un documento. Ese ‘patrón’ que escriben es una ***expresión regular***. Por otro lado, imaginen una máquina que lee el texto letra a letra y, según reglas internas, decide si la acepta o no. Eso es un ***autómata finito***.

El *Teorema de Kleene ([o Algoritmo de Kleene](https://en.wikipedia.org/wiki/Kleene%27s_algorithm))* nos afirma que todo patrón que podamos describir con una expresión regular lo puede reconocer un autómata finito, y viceversa. En este capítulo veremos dos partes:

- *TK - P $I$: "Cómo, dado un patrón, construir esa máquina (expresión regular → autómata)"*.

- *TK - P $II$: "Cómo, dada la máquina, recuperar un patrón equivalente (autómata → expresión regular)"*.

Al final, entenderemos por qué en muchos motores de búsqueda o compiladores basta con uno u otro formalismo.


## **4.1 Teorema de Kleene - Parte $I$**

### **4.1.1 Introducción**

En las secciones anteriores hemos visto que los tres modelos de autómatas finitos —determinista (AFD), no‐determinista (AFN) y no‐determinista con transiciones $\lambda$ (AFN-$\lambda$)— son **equivalentes desde el punto de vista del lenguaje que reconocen**. Esto se resume con la notación

$$
\mathrm{AFD}\;\equiv\;\mathrm{AFN}\;\equiv\;\mathrm{AFN}\!-\lambda.
$$

**¿Qué implica esta equivalencia?**

* Para **cada** AFD existe un AFN y un AFN-$\lambda$ que reconocen exactamente el mismo conjunto de cadenas.
* Análogamente, dado un AFN-$\lambda$ o un AFN podemos construir un AFD equivalente.
* En todos los casos, el mecanismo de reconocimiento (determinista, no-determinista, con o sin $\lambda$) no **amplía** ni **reduce** la clase de lenguajes aceptables: siempre obtenemos los mismos **lenguajes regulares**.

El ***Teorema de Kleene*** proclama que **esa** familia de lenguajes —los que aceptan los autómatas finitos— coincide **precisamente** con los lenguajes que pueden describirse mediante expresiones regulares.

### **4.1.2 Teorema de Kleene**

> **Enunciado.**
> Un lenguaje $L$ es **regular** **si y solo si** existe un autómata finito (ya sea AFD, AFN o AFN-$\lambda$) que lo reconozca, es decir, tal que
>
> $$
> L = L(M).
> $$

Este “***si y solo si***” se demuestra constructivamente en dos etapas:

1. **De expresiones regulares a AFN-$\lambda$.**
   Mostramos cómo, dada cualquier expresión regular $R$, construir un AFN-$\lambda$ $M$ tal que $L(M)=L(R)$. Así probamos que **todo lenguaje descrito por una expresión regular** es reconocido por un AFN-$\lambda$.  

2. **De AFD a expresiones regulares.**
   Partiendo de un AFD $M$, desarrollamos un algoritmo para obtener una **expresión regular** $R$ con $L(R)=L(M)$. Con esto queda demostrado que **todo lenguaje aceptado por un autómata finito** se puede describir con una expresión regular.

> Ambas construcciones son **algorítmicas** y, en la práctica, sirven de base a herramientas de generación de analizadores léxicos que primero traducen una expresión a un AFN-$\lambda$ y luego lo convierten a un AFD optimizado.

### **4.1.3 Parte $I$ - De expresión regular a AFN-$\lambda$**

**Enunciado.**

> Dada una expresión regular
>
> $$
> R
> $$
>
> sobre un alfabeto $\Sigma$, existe un **algoritmo** para construir un AFN-$\lambda$
>
> $$
> M=(\Sigma,\,Q,\,q_{0},\,F,\,\Delta)
> $$
>
> de forma que
>
> $$
> L(M)\;=\;L(R).
> $$

Formalmente, la **Parte I** dice:  

> *Para toda expresión regular R existe un autómata finito no determinista con transiciones λ (AFN-λ) que reconoce exactamente el mismo lenguaje que R.*  

En palabras llanas: si sé cómo describir un conjunto de cadenas con símbolos y operadores (unión, concatenación, estrella), puedo **convertir** esa descripción en una máquina que, al leer cualquier cadena, me diga si pertenece o no al conjunto.”

A grandes rasgos, el método **recorre** la estructura de $R$ y, para cada operador (unión, concatenación, estrella) o caso base (∅, ε, símbolo), ensambla un “bloque” pequeño de autómata con transiciones $\lambda$, que luego se **pegan** entre sí para reflejar exactamente la semántica de $R$.

#### **1. Casos base**

1. **$R = \varnothing$ (lenguaje vacío)**

   * **Construcción:** dos estados $p,q$, **sin** flechas entre ellos.
   * **Comportamiento:** no hay camino de $p$ a $q$, así no se acepta ninguna cadena.

2. **$R = \varepsilon$ (cadena vacía)**

   * **Construcción:** dos estados $p\to q$ unidos por una **transición $\lambda$**.
   * **Comportamiento:** solo permite el salto nulo, acepta exactamente la cadena vacía.

3. **$R = a\in\Sigma$ (un símbolo)**

   * **Construcción:** dos estados $p\xrightarrow{a}q$.
   * **Comportamiento:** acepta únicamente la palabra “$a$”.

#### **2. Operador unión $R = R_{1}\cup R_{2}$**

Supongamos ya disponemos de AFN-$\lambda$ para $R_{1}$ y $R_{2}$, con estados iniciales $s_{1},s_{2}$ y finales $f_{1},f_{2}$.

* Crear **nuevo** inicio $s$ y final $f$.
* Añadir $\lambda$-transiciones:

  * $s \xrightarrow{\lambda} s_{1}$
  * $s \xrightarrow{\lambda} s_{2}$
  * $f_{1}\xrightarrow{\lambda} f$
  * $f_{2}\xrightarrow{\lambda} f$

De este modo, $M$ **elige** en silencio si procesa $R_{1}$ o $R_{2}$, y ambos caminos confluyen en $f$.

#### **3. Operador concatenación $R = R_{1}R_{2}$**

Dados los AFN-$\lambda$ de $R_{1}$ y $R_{2}$ con inicios $s_{1},s_{2}$ y finales $f_{1},f_{2}$:

* Conservar $s_{1}$ como inicio global.
* Conectar $f_{1}\xrightarrow{\lambda} s_{2}$.
* Conservar $f_{2}$ como estado de aceptación.

Así, el automáta consume primero cualquier palabra de $R_{1}$ y, sin leer símbolo extra, salta a $R_{2}$.

#### **4. Operador estrella $R = R_{1}^{*}$**

Partiendo del AFN-$\lambda$ de $R_{1}$ con inicio $s_{1}$ y final $f_{1}$:

* Crear **nuevo** inicio/aceptación $s$.
* Añadir $\lambda$-transiciones:

  1. $s \xrightarrow{\lambda} s_{1}$ (arrancar primera copia)
  2. $s \xrightarrow{\lambda} f$   (aceptar la cadena vacía)
  3. $f_{1}\xrightarrow{\lambda} s_{1}$ (repetir otro bloque)
  4. $f_{1}\xrightarrow{\lambda} f$   (salir tras un bloque)

Este diseño admite **cero o más** repeticiones de $R_{1}$, incluida la opción de no consumir ningún símbolo.

#### **Esquema global**

La construcción recursiva asegura que cada paso **preserva** el lenguaje:

* Los bloques base cubren ∅, ε y símbolos.
* Unión, concatenación y estrella los combinan exactamente según la definición de expresiones regulares.

  <img src="https://github.com/carlosalvarezh/TeoriaComputacion/blob/main/images/TC05_AFN02.PNG?raw=true" width="1000" />
</p>

A partir de aquí, cualquier expresión regular —por compleja que sea— se traduce de forma mecánica a un AFN-λ que **reconoce exactamente** el mismo conjunto de cadenas.

## **4.2 Lema de Arden**

Para completar la **Parte $II$** del Teorema de Kleene (convertir un autómata finito en expresión regular) necesitaremos este resultado fundamental.

 ### **4.2.1 Enunciado del Lema de Arden**

 Sean $A$ y $B$ dos lenguajes sobre el alfabeto $\Sigma$, con la condición de que la cadena vacía $\lambda$ **no** pertenezca a $A$.
Consideremos la ecuación en la variable lenguaje $X$:

$$
X \;=\; A\,X \;\cup\; B.
$$

El **Lema de Arden** afirma que:

> **La única solución** a esa ecuación es
>
> $$
> X \;=\; A^{*}\,B.
> $$

### **4.2.2 ¿Por qué funciona?**

1. **Interpretación de la ecuación**

   * $A\,X$ significa “*todas las cadenas que empiezan con un prefijo en $A$, seguido de algo en $X$*”.
   * La unión con $B$ permite además “*empezar directamente con algo de $B$*”.

   La ecuación $X = A\,X \cup B$ dice:

   > “Un string está en $X$ **exactamente** si
   >
   > * o bien comienza con un bloque de $A$ y luego sigue en $X$,
   > * o bien forma parte de $B$.”

2. **Construir progresivamente**

   * Si eliges la segunda opción, tomas una palabra de $B$.
   * Si eliges la primera, tomas un prefijo en $A$ y después vuelves a decidir si sigues con otro $A$ o paras en $B$.

   Este “*volver a decidir*” en cada paso repite $A$ **cero o más veces** antes de llegar a un elemento de $B$.

3. **Conclusión**
   Esa descripción es **precisamente** la definición de

   $$
   A^{*}\,B
   = \underbrace{A\,A\,\cdots\,A}_{\text{cero o más veces}}\,B.
   $$

   No puede haber otra solución distinta porque, al imponer que $\lambda\notin A$, evitamos ciclos vacíos infinitos; cada repetición efectiva de $A$ consume al menos un símbolo.


### **4.2.3 Ejemplos**

**- Ejemplo 1:** La ecuación $X=a X \cup b^* a b$ tiene solución única $X=a^* b^* a b$.  

**- Ejemplo 2:** La ecuación $X=a^2 X \cup b^{+} X \cup a b$ se puede escribir en la forma $X=\left(a^2 \cup b^{+}\right) X \cup a b$. Por el lema de Arden la ecuación tiene solución única $X=\left(a^2 \cup b^{+}\right)^* a b$.  

**- Ejemplo 3:** La ecuación $X=a b^2 X \cup a X \cup a^* b \cup b^* a$ se puede escribir como $X=\left(a b^2 \cup a\right) X \cup\left(a^* b \cup b^* a\right)$. Por lema de Arden la ecuación tiene solución única $X=\left(a b^2 \cup a\right)^*\left(a^* b \cup b^* a\right)$.  

**- Ejemplo aplicación práctica**
Imagina un filtro de correo que agrupa mensajes así:

* **Regla $B$:** Mensajes urgentes (“URGENTE: …”).
* **Regla $A$:** Mensajes estándar que añaden la etiqueta “\[INFO] ” al comienzo.

  La “bandeja” $X$ de mensajes que debes procesar se define recursivamente:

> “*Un mensaje está en la bandeja si*
>
> 1. *empieza con `[INFO] ` seguido de algo en la bandeja,*
> 2. *o bien es un mensaje urgente.*”

  Desplegar esa definición significa: puedes tener **cero o varias** etiquetas `[INFO] ` al inicio, y **finalmente** un mensaje urgente.
  Eso equivale a la forma $[INFO]^{*}\, \text{URGENTE}…$, exactamente “$A^{*}B$”.

El Lema de Arden será la **herramienta clave** para despejar, en forma de expresiones regulares, las ecuaciones de subconjuntos de estados que aparecen al convertir AFDs a expresiones regulares.

## **4.3 Teorema de Kleene - Parte $II$**

En esta sección probaremos que **todo** AFN

$$
M=(\Sigma,\,Q,\,q_{0},\,F,\,\Delta)
$$

admite una expresión regular $R$ tal que

$$
L(M)=L(R).
$$


### **4.3.1 Descomposición en subautómatas**

Aunque $M$ tiene un único estado inicial $q_{0}$, resulta útil considerar **-para cada** $q_{i}\in Q$ el autómata

$$
M_{i} = (\Sigma,\,Q,\,q_{i},\,F,\,\Delta),
$$

idéntico a $M$ salvo por el estado inicial. Denotemos

$$
A_{i} = L(M_{i})
   = \{\,w\in\Sigma^{*}\mid \Delta(q_{i},w)\cap F\neq\varnothing\},
$$

y en particular $A_{0}=L(M)$.  Nuestro objetivo será hallar una expresión para $A_{0}$.



### **4.3.2 Ecuaciones recursivas para los $A_i$**

Para cada $i$, podemos **describir** $A_{i}$ en función de los $A_{j}$ de esta manera:

* Si $q_{i}\notin F$, **todas** las cadenas en $A_{i}$ comienzan con un símbolo $a\in\Sigma$ y siguen en algún $A_{j}$ tal que $q_{j}\in\Delta(q_{i},a)$.
* Si $q_{i}\in F$, además de lo anterior, la cadena vacía $\lambda$ también pertenece a $A_{i}$.

Formalmente,

$$
A_{i} =
\begin{cases}
\displaystyle
  \bigcup_{a\in\Sigma}\;\bigcup_{q_{j}\in\Delta(q_{i},a)}\;a\,A_{j},
& q_{i}\notin F,\\[8pt]
\displaystyle
  \Bigl(\bigcup_{a\in\Sigma}\;\bigcup_{q_{j}\in\Delta(q_{i},a)}\;a\,A_{j}\Bigr)\;\cup\;\{\lambda\},
& q_{i}\in F.
\end{cases}
$$

Aquí $a\,A_{j}$ significa “concatenar el símbolo $a$ al comienzo de cada palabra de $A_{j}$”.


### **4.3.3 Sistema de ecuaciones**

Si $Q=\{q_{0},q_{1},\dots,q_{n}\}$, las igualdades anteriores generan un **sistema** de $n+1$ ecuaciones lineales en las incógnitas $A_{0},A_{1},\dots,A_{n}$:

$$
\begin{cases}
A_{0}=C_{0,1}A_{1}\;\cup\;C_{0,2}A_{2}\;\cup\;\cdots\;\cup\;C_{0,n}A_{n},\\
A_{1}=C_{1,0}A_{0}\;\cup\;C_{1,1}A_{1}\;\cup\;\cdots\;\cup\;C_{1,n}A_{n},\\
\quad\vdots\\
A_{n}=C_{n,0}A_{0}\;\cup\;C_{n,1}A_{1}\;\cup\;\cdots\;\cup\;C_{n,n}A_{n}\;\cup\;\Lambda,
\end{cases}
$$

donde cada coeficiente $C_{i,j}$ es o bien $\varnothing$ o bien un símbolo de $\Sigma$, y el término $\Lambda=\{\lambda\}$ aparece únicamente en las ecuaciones correspondientes a estados finales.

### **4.3.4 Resolución mediante el Lema de Arden**

Para **despejar** estas ecuaciones de lenguajes utilizamos repetidamente el Lema de Arden:

> Si $X = A\,X\;\cup\;B$ y $\lambda\notin A$, entonces la única solución es $X= A^{*}B$.

El procedimiento es el siguiente:

1. **Empezar por la última ecuación** ($A_{n}$):

   $$
   A_{n} = C_{n,n}\,A_{n}\;\cup\;(\text{resto})\;\cup\;\Lambda,
   $$

   donde $C_{n,n}$ no contiene $\lambda$.
   Aplicamos el Lema de Arden para escribir

   $$
     A_{n} = C_{n,n}^{*}\,\bigl(\text{resto}\;\cup\;\Lambda\bigr).
   $$
2. **Sustituir** esta expresión de $A_{n}$ en las ecuaciones de $A_{0},\dots,A_{n-1}$.
3. Obtener un **nuevo** sistema de $n$ ecuaciones en $A_{0},\dots,A_{n-1}$.
4. Repetir el proceso para $A_{n-1}$, $A_{n-2}$, … hasta quedarnos con **una sola** ecuación en $A_{0}$.
5. Aplicar finalmente el Lema de Arden a esa ecuación única y así obtener

   $$
     A_{0} = R,
   $$

   donde $R$ es una expresión regular construida a partir de los símbolos $C_{i,j}$ y de las operaciones de unión, concatenación y estrella.

Dado que cada paso **preserva** la igualdad de lenguajes y los coeficientes $C_{i,j}$ son símbolos de $\Sigma$, el resultado final $R$ cumple

$$
L(R) \;=\; A_{0} \;=\; L(M).
$$


Con esto concluimos la **Parte II** del Teorema de Kleene: **todo** lenguaje aceptado por un autómata finito es **regular**, es decir, puede escribirse mediante una expresión regular.


### **4.3.5 Ejemplos de la sección**

A continuación aplicamos el método para **extraer** una expresión regular $L(M)$ de un AFN

$$
M=(\Sigma,Q,q_{0},F,\Delta)
$$

mediante la resolución sucesiva de ecuaciones.


#### **Ejemplo 4.3.5.1:**

Considérese el AFN $M$ de la **Figura 1** sobre $\Sigma=\{a,b\}$:

<p float="center">
  <img src="https://github.com/carlosalvarezh/TeoriaComputacion/blob/main/images/TC05_AFN13.PNG?raw=true" width="500" />
</p>

Inspeccionando vemos que $M$ acepta cualquier número de $a$ o $b$, luego **dos** $a$ consecutivas, y a continuación de nuevo cualquier mezcla de $a$ y $b$. Es decir, intuitivamente

$$
L(M) \;=\;(a\cup b)^{*}\;a^{2}\;(a\cup b)^{*}.
$$

Pero aplicaremos el **procedimiento sistemático**:

**1. Definir los lenguajes $A_i$**

Para cada estado $q_i\in\{q_0,q_1,q_2\}$ definimos

$$
A_{i} \;=\;L(M_{i})
\;=\;
\{\,w\in\{a,b\}^{*} \mid \Delta(q_{i},w)\cap F\neq\varnothing\}.
$$

En particular $A_{0}=L(M)$.

**2. Escribir el sistema de ecuaciones**

Cada $A_i$ se descompone en función de un primer paso:

1. Desde **$q_0$** (no final), leer

   * un $a$ y volver a $q_0$,
   * o un $b$ y volver a $q_0$,
   * o un $a$ y pasar a $q_1$.
     De ello surge

   $$
     A_{0}
     \;=\;
     a\,A_{0}\;\cup\;b\,A_{0}\;\cup\;a\,A_{1}.
     \tag{1}
   $$

2. Desde **$q_1$** (no final), la única flecha es un $a$ a $q_2$:

   $$
     A_{1}
     \;=\;
     a\,A_{2}.
     \tag{2}
   $$

3. Desde **$q_2$** (final), puede leer

   * un $a$ y quedarse en $q_2$,
   * o un $b$ y quedarse en $q_2$,
   * o detenerse en $\lambda$.
     Por tanto

   $$
     A_{2}
     \;=\;
     a\,A_{2}\;\cup\;b\,A_{2}\;\cup\{\lambda\}.
     \tag{3}
   $$

**3. Resolver $A_{2}$ con el Lema de Arden**

La ecuación (3) se reescribe agrupando $\{a,b\}$:

$$
A_{2}
= (a\cup b)\,A_{2} \;\cup\;\{\lambda\}.
\tag{4}
$$

Como $\lambda\notin\{a,b\}$, aplicamos el **Lema de Arden** a $X=(a\cup b)X\cup\{\lambda\}$, obteniendo

$$
\boxed{
A_{2} \;=\;(a\cup b)^{*}\,\{\lambda\}
\;=\;(a\cup b)^{*}.
}
\tag{5}
$$

**4. Sustituir en $A_{1}$**

Reemplazamos (5) en (2):

$$
A_{1}
= a\,A_{2}
= a\,(a\cup b)^{*}.
\tag{6}
$$

**5. Sustituir en $A_{0}$ y aplicar Arden**

La ecuación (1) queda

$$
A_{0}
= a\,A_{0}\;\cup\;b\,A_{0}\;\cup\;a\,A_{1}
=
(a\cup b)\,A_{0}\;\cup\;a\bigl(a\,(a\cup b)^{*}\bigr).
$$

El término constante es $a\,A_{1}=a^{2}(a\cup b)^{*}$. Así:

$$
A_{0}
= (a\cup b)\,A_{0}\;\cup\;a^{2}\,(a\cup b)^{*}.
\tag{7}
$$

De nuevo, $\lambda\notin(a\cup b)$, así que por Arden:

$$
\boxed{
A_{0}
= (a\cup b)^{*}\;\bigl(a^{2}(a\cup b)^{*}\bigr)
= (a\cup b)^{*}\,a^{2}\,(a\cup b)^{*}.
}
$$

Esto coincide con la descripción intuitiva inicial.

**Conclusión:** el procedimiento ha generado la expresión regular

$$
R \;=\;(a\cup b)^{*}\,a^{2}\,(a\cup b)^{*}
$$

tal que $L(M)=R$, ilustrando la **Parte II** del Teorema de Kleene.


#### **Ejemplo 4.3.5.2**

Obtener una **expresión regular** para el lenguaje $L(M)$ reconocido por el siguiente AFN $M$:

<p float="center">
  <img src="https://github.com/carlosalvarezh/TeoriaComputacion/blob/main/images/TC05_AFN14.PNG?raw=true" width="750" />
</p>

**1. Sistema de ecuaciones**

Para cada estado $q_i$ definimos $A_i=L(M_i)$. El diagrama de transiciones indica:

1. Desde $q_0$ solo hay flecha $a\to q_1$:

   $$
   (1)\quad A_0 = a\,A_1.
   $$

2. Desde $q_1$ solo hay flecha $a\to q_2$:

   $$
   (2)\quad A_1 = a\,A_2.
   $$

3. En $q_2$ hay

   * un ciclo $b\to q_2$,
   * una flecha $b\to q_3$,
   * $q_2$ es **final** ($\lambda\in A_2$).

   $$
   (3)\quad A_2 = b\,A_2 \;\cup\; b\,A_3 \;\cup\;\{\lambda\}.
   $$

4. En $q_3$ hay

   * un ciclo $a\to q_3$,
   * una flecha $b\to q_4$.

   $$
   (4)\quad A_3 = a\,A_3 \;\cup\; b\,A_4.
   $$

5. En $q_4$ hay

   * flechas $a\to q_2$ y $a\to q_3$,
   * $q_4$ es **final** ($\lambda\in A_4$).

   $$
   (5)\quad A_4 = a\,A_2 \;\cup\; a\,A_3 \;\cup\;\{\lambda\}.
   $$

**2. Despejar $A_3$**

Primero sustituimos la ecuación (5) en (4) para escribir $A_3$ solo en función de sí mismo y de $A_2$:

$$
\begin{aligned}
A_3
&= a\,A_3 \;\cup\; b\,(a\,A_2 \;\cup\; a\,A_3 \;\cup\;\{\lambda\})\\
&= a\,A_3 
  \;\cup\; b\,a\,A_2 
  \;\cup\; b\,a\,A_3 
  \;\cup\; b.\\
\end{aligned}
$$

Agrupamos los términos que contienen $A_3$:

$$
(6)\quad
A_3 = (a \;\cup\; b\,a)\,A_3 
      \;\cup\;\bigl(b\,a\,A_2\;\cup\;b\bigr).
$$

* **Término en $X$:** $ (a\cup b\,a)\,A_3$.
* **Término constante:** $b\,a\,A_2\cup b$.

Como $\lambda\notin (a\cup b\,a)$, aplicamos el **Lema de Arden** a $X=(a\cup b\,a)X\cup D$, donde $D=b\,a\,A_2\cup b$. Obtenemos:

$$
(7)\quad
A_3 = (a\cup b\,a)^{*}\,\bigl(b\,a\,A_2\;\cup\;b\bigr)
     = (a\cup b\,a)^{*}\,b\,a\,A_2
       \;\cup\;(a\cup b\,a)^{*}\,b.
$$

**3. Despejar $A_2$**

Ahora sustituimos (7) en la ecuación (3):

$$
\begin{aligned}
A_2
&= b\,A_2
  \;\cup\; b\,A_3
  \;\cup\;\{\lambda\}\\
&= b\,A_2 
  \;\cup\; b\Bigl((a\cup b\,a)^{*}b\,a\,A_2\;\cup\;(a\cup b\,a)^{*}b\Bigr)
  \;\cup\;\{\lambda\}\\
&= b\,A_2
  \;\cup\; b\,(a\cup b\,a)^{*}b\,a\,A_2
  \;\cup\; b\,(a\cup b\,a)^{*}b
  \;\cup\;\{\lambda\}.
\end{aligned}
$$

Reagrupamos para aplicar Arden:

$$
(8)\quad
A_2
= \bigl[b \;\cup\; b(a\cup b\,a)^{*}b\,a\bigr]\,A_2
  \;\cup\;
  \bigl[b(a\cup b\,a)^{*}b\;\cup\;\{\lambda\}\bigr].
$$

* **Coeficiente en $A_2$:** $C = b\cup b(a\cup b\,a)^{*}b\,a$.
* **Término constante:** $D = b(a\cup b\,a)^{*}b\;\cup\;\{\lambda\}$.

Por Arden, como $\lambda\notin C$,

$$
(9)\quad
A_2
= C^{*}\;D
= \bigl[b\cup b(a\cup b\,a)^{*}b\,a\bigr]^{*}
  \bigl[b(a\cup b\,a)^{*}b\;\cup\;\{\lambda\}\bigr].
$$

**4. Obtener $A_1$ y $A_0$**

1. **Ecuación (2):**

   $$
   A_1 = a\,A_2
       = a\;\Bigl[\;C^{*}\,D\Bigr].
   $$

2. **Ecuación (1):**

   $$
   A_0 = a\,A_1
       = a^{2}\;\Bigl[C^{*}\,D\Bigr].
   $$

Sustituyendo $C$ y $D$ recuperamos la **expresión final**:

$$
\boxed{
L(M)=A_0
= a^{2}\,
\Bigl[b\cup b(a\cup b\,a)^{*}b\,a\Bigr]^{*}\,
\Bigl[b(a\cup b\,a)^{*}b\;\cup\;\lambda\Bigr].
}
$$

Esta fórmula, aunque algo extensa, **describe exactamente** cómo $M$ puede:

1. Leer dos $a$s iniciales,
2. repetir un bloque de “$b$” o “$b\,a\,(alguna mezcla)\,b\,a$” arbitrarias veces,
3. y finalmente, o bien consumir la porción “$\,b(a\cup b\,a)^{*}b$”, o bien terminar en vacío ($\lambda$).

Con ello queda ilustrado el cálculo completo de la **Parte II** del Teorema de Kleene.
