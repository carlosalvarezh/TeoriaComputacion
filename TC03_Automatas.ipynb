# **Capítulo 2: Autómatas Finitos**

Los autómatas finitos son modelos abstractos de máquinas que procesan cadenas de entrada y, en función de su estructura y transiciones, deciden si aceptan o rechazan dichas cadenas. Estos modelos surgieron en el contexto del estudio de la computación y tienen una amplia aplicación en el diseño de lenguajes de programación, compiladores y sistemas de control.

## **2.1 Introducción a los Autómatas Finitos**

Históricamente, el estudio de los autómatas tiene sus orígenes en los trabajos pioneros de Alan Turing en 1936, quien propuso la “máquina de Turing” para formalizar el concepto de computación. Posteriormente, durante las décadas de 1940 y 1950, se investigaron variantes más restringidas de estos modelos, lo que llevó al surgimiento de los **autómatas finitos**.

En esencia, un autómata finito es una máquina abstracta que procesa una cadena de entrada (una secuencia de símbolos) y, mediante un conjunto finito de estados y reglas de transición, determina si la cadena es aceptada o rechazada. La importancia de estos modelos radica en su capacidad para reconocer lenguajes regulares, que son fundamentales en la teoría de la computación y en aplicaciones prácticas como la verificación de patrones en cadenas (por ejemplo, en expresiones regulares).

## **2.2 Autómatas Finitos Deterministas (AFD)**

### **2.2.1 Definición Formal**

Un **Autómata Finito Determinista (AFD)** se define formalmente como una quíntupla

$$
M = (\Sigma, Q, q_0, F, \delta)
$$

donde:

- **$\Sigma$:** Es el *alfabeto de cinta*, es decir, el conjunto finito de símbolos que pueden aparecer en la cadena de entrada. Todas las cadenas procesadas por el autómata pertenecen a $\Sigma^*$.

- **$Q$:** Es el conjunto finito de estados internos del autómata. Se suele denotar como $Q = \{q_0, q_1, \ldots, q_n\}$.

- **$q_0 \in Q$:** Es el *estado inicial* en el que se encuentra el autómata antes de comenzar a procesar la cadena de entrada.

- **$F \subseteq Q$:** Es el conjunto de *estados finales* o de *aceptación*. Una cadena se considera aceptada por el autómata si, tras procesarla completamente, el autómata termina en alguno de los estados de $F$. Se requiere que $F \neq \varnothing$.

- **$\delta: Q \times \Sigma \to Q$:** s la *función de transición*. Para cada combinación de un estado $q \in Q$ y un símbolo $s \in \Sigma$, la función $\delta$ determina de forma única el siguiente estado $q' \in Q$. Es decir, $\delta(q, s) = q'$.

### **2.2.2 Funcionamiento del AFD**

El proceso de computación en un AFD se describe de la siguiente manera:

- **Inicio:** a cadena de entrada $u$ se coloca en una cinta semiinfinita, divida en celdas, donde cada celda contiene un símbolo de $u$. El primer símbolo de $u$ se ubica en la primera celda de la cinta.
  
- **Unidad de Control:** El autómata posee una *unidad de control* (o *cabeza lectora*) que comienza en el estado inicial $q_0$ y se sitúa sobre la primera celda de la cinta.  

<p float="center">
  <img src="https://github.com/carlosalvarezh/TeoriaComputacion/blob/main/images/TC03_AutomataFinito02.PNG?raw=true" width="500" />
</p>

- **Transiciones:** La función de transición $\delta$ indica al autómata qué hacer al leer un símbolo $s$ en un estado $q$. Concretamente, si se tiene $\delta(q, s) = q'$, el autómata:
  1. Cambia de estado, pasando de $q$ a $q'$.
  2. Se desplaza **hacia la derecha** para leer el siguiente símbolo.
  
  Cada cambio de estado y movimiento constituye un **paso computacional**.

<p float="center">
  <img src="https://github.com/carlosalvarezh/TeoriaComputacion/blob/main/images/TC03_AutomataFinito03.PNG?raw=true" width="500" />
</p>

- **Finalización:** El autómata procesa la cadena hasta que alcanza la primera celda vacía. La cadena de entrada se acepta si el autómata termina en un estado de aceptación ($q \in F$); de lo contrario, se rechaza.

  <img src="https://github.com/carlosalvarezh/TeoriaComputacion/blob/main/images/TC03_AutomataFinito04.PNG?raw=true" width="500" />
</p>

<p float="center">

</p>

- **Caso Especial:** La cadena vacía $\lambda$ es aceptada si, y solo si, el estado inicial $q_0$ se encuentra en el conjunto $F$.  
<p float="center">

#### **Ejemplo**

Consideremos el siguiente autómata finito determinista:

- **Alfabeto:** $\Sigma = \{a, b\}$.
- **Conjunto de Estados:** $Q = \{q_0, q_1, q_2\}$.
- **Estado Inicial:** $q_0$.
- **Estados de Aceptación:** $F = \{q_0, q_2\}$.
- **Función de Transición:** Se define en la siguiente tabla:

  | $\delta$ | $a$ | $b$ |
  |------------|-------|-------|
  | **$q_0$**  | $q_0$ | $q_1$ |
  | **$q_1$**  | $q_1$ | $q_2$ |
  | **$q_2$**  | $q_1$ | $q_1$ |

  Esto se lee de la siguiente manera:
  - $\delta(q_0, a) = q_0$ y $\delta(q_0, b) = q_1$.
  - $\delta(q_1, a) = q_1$ y $\delta(q_1, b) = q_2$.
  - $\delta(q_2, a) = q_1$ y $\delta(q_2, b) = q_1$.

**Procesamiento de Cadenas:**

1. **Cadena $u = aabab$:**
   - El autómata inicia en $q_0$.
   - Lee el primer $a$: $\delta(q_0, a) = q_0$.
   - Lee el segundo $a$: $\delta(q_0, a) = q_0$.
   - Lee $b$: $\delta(q_0, b) = q_1$.
   - Lee $a$: $\delta(q_1, a) = q_1$.
   - Lee $b$: $\delta(q_1, b) = q_2$.
   - La cadena se procesa completamente y el autómata termina en $q_2$, que es un estado de aceptación. Por lo tanto, $u$ es aceptada.

2. **Cadena $v = aabab a$:**
   - El procesamiento es similar, pero al final el autómata termina en $q_1$, que no es un estado de aceptación. Por lo tanto, $v$ es rechazada.

3. **Cadena Vacía $\lambda$:**
   - La cinta está vacía y el autómata no realiza transiciones.
   - Como el estado inicial $q_0$ está en $F$, $\lambda$ es aceptada.


### **2.2.3 Observaciones y Propiedades**

- **Determinismo:**  
  En un AFD, para cada combinación de estado $q \in Q$ y símbolo $s \in \Sigma$ la función $\delta(q, s)$ está definida de forma única. Esto significa que no hay ambigüedad en la evolución del autómata.

- **Direccionalidad:**  
  La unidad de control siempre se desplaza hacia la derecha y no tiene capacidad para retroceder o modificar el contenido de la cinta.

- **Aceptación de la Cadena Vacía:**  
  La cadena $\lambda$ es aceptada si, y solo si, el estado inicial $q_0$ pertenece al conjunto $F$.

El lenguaje aceptado por un autómata $M$ se denota por $L(M)$ y se define como:

$$
L(M) = \{\, u \in \Sigma^* : M \text{ termina el procesamiento de la cadena de entrada} u \text{ en un estado } q \in F \,\}.
$$

## **2.3. Diagrama de transiciones de un autómata**

Una cadena $u$ es aceptada si existe una trayectoria etiquetada con los símbolos de $u$, que comienza en el estado $q_0$ y termina en un estado de aceptación.

## **Capítulo 2.4: Diseño de Autómatas**

Los **autómatas finitos** son máquinas abstractas que, mediante un conjunto finito de estados y transiciones, "leen" una cadena de entrada y deciden si ésta pertenece o no a un lenguaje. En el proceso de diseñar tales autómatas, especialmente los deterministas, se han adoptado ciertas convenciones que simplifican los diagramas de transición.

### **2.4.1 Convenciones en los Diagramas de Transición**

En los diagramas de autómatas finitos deterministas (AFD) se utiliza la siguiente **convención adicional**:
- **Estado Limbo (o estado trampa):**  
  Se asume que cualquier arco o transición que no se dibuje explícitamente conduce a un estado "limbo" o trampa, el cual no es un estado de aceptación. Esto significa que solo se muestran los arcos que forman parte de trayectorias que, eventualmente, pueden conducir a la aceptación de una cadena.  
  **Ventaja:** Esta convención simplifica los diagramas, ya que se evita la saturación visual que produciría dibujar todas las transiciones posibles (incluso aquellas que resultan en rechazo).

### **2.4.2 Tipos de Problemas Abordados**

En el estudio y diseño de autómatas, se abordan dos tipos fundamentales de problemas:

1. **Diseño (Construcción) del Autómata a partir de un Lenguaje Regular:**  
   Se busca, dada una descripción formal de un lenguaje regular $ L $, construir un autómata $ M $ tal que $ L(M) = L $.  
   Es decir, el autómata debe reconocer todas y solo las cadenas que pertenezcan a $ L $.

2. **Determinación del Lenguaje Aceptado por un Autómata:**  
   Dado un autómata $ M $ (con una descripción o diagrama de transición), se busca determinar el conjunto $ L(M) $ de todas las cadenas que $ M $ acepta.

En este capítulo nos centraremos inicialmente en problemas del primer tipo: construir autómatas a partir de lenguajes regulares.

### **2.4.3 Ejemplos de Diseño de Autómatas**

A continuación se presentan varios ejemplos de diseño de AFD, desarrollados paso a paso. En cada caso se presenta una "versión completa" (con diagrama completo) y, a modo ilustrativo, una "versión simplificada" en la que se omiten los arcos que conducen al estado limbo.

#### **Ejemplo 1: $ L = a^* $**  
**Lenguaje:**  
$$
L = a^* = \{\lambda,\, a,\, a^2,\, a^3,\, \dots\}
$$
Es decir, todas las cadenas (incluida la vacía) compuestas únicamente del símbolo $a$.

**Diseño del AFD:**

1. **Análisis del Lenguaje:**  
   Toda cadena en $L$ consiste en cero o más repeticiones de $a$.  
   Se debe aceptar la cadena vacía $\lambda$ y cualquier secuencia de $a$'s.

2. **Construcción del Autómata:**  
   - **Estados:** Se puede diseñar un autómata con **un único estado** $q_0$ que actúa tanto de estado inicial como de estado de aceptación.
   - **Transición para $a$:** Desde $q_0$, al leer $a$, se permanece en $q_0$.
   - **Transiciones no especificadas:** Cualquier transición no dibujada (por ejemplo, lecturas de otros símbolos) se asume que llevan al estado limbo de rechazo.

#### **Ejemplo 2: $ L = a^+ $**  
**Lenguaje:**  
$$
L = a^+ = \{a,\, a^2,\, a^3,\, \dots\}
$$
Es decir, todas las cadenas compuestas de al menos una $a$.

**Diseño del AFD:**

1. **Análisis del Lenguaje:**  
   A diferencia de $a^*$, el lenguaje $a^+$ no acepta la cadena vacía, por lo que el autómata debe rechazar $\lambda$.

2. **Construcción del Autómata:**  
   - **Estados:**  
     Se requiere al menos dos estados:  
     - $q_0$: estado inicial, **no es de aceptación**.  
     - $q_1$: estado al que se transita tras leer al menos un $a$; este es un **estado de aceptación**.
   - **Transiciones:**  
     - Desde $q_0$, al leer $a$, se transita a $q_1$.  
     - Desde $q_1$, al leer $a$, se permanece en $q_1$ (o se regresa a sí mismo).

3. **Observación:**  
   La cadena vacía $\lambda$ es rechazada porque $q_0$ no es de aceptación.


#### **Ejemplo 3: $ L = $ Lenguaje de las cadenas que contienen exactamente dos $a$’s**  
**Alfabeto:**  
$\Sigma = \{a, b\}$.

**Lenguaje:**  
El lenguaje consiste en todas las cadenas que tengan exactamente dos ocurrencias del símbolo $a$. Una forma de representar este lenguaje es:
$$
L = b^* a \, b^* a \, b^*
$$
Lo cual indica que puede haber cualquier cantidad de $b$’s (incluyendo cero) antes de la primera $a$, entre las dos $a$’s y después de la segunda $a$.

**Diseño del AFD:**

1. **Análisis del Lenguaje:**  
   Se debe contar la aparición de $a$’s:  
   - Inicialmente, se han leído 0 $a$’s.  
   - Luego, al leer la primera $a$, se transita a un estado intermedio.  
   - Tras leer la segunda $a$, se transita a un estado final (de aceptación), en el cual ya no se debe aceptar ninguna otra $a$.

2. **Construcción del Autómata (diagrama en Imagen 5):**  

<p float="center">
  <img src="https://github.com/carlosalvarezh/TeoriaComputacion/blob/main/images/TC03_Graph10.PNG?raw=true" width="350" />
</p>

   - **Estados:**  
     - $q_0$: Estado inicial, donde todavía no se ha leído ningún $a$.  
     - $q_1$: Estado intermedio, tras leer la primera $a$.  
     - $q_2$: Estado final, tras leer la segunda $a$; es un estado de aceptación.  
   - **Transiciones:**  
     - En $q_0$:  
       - Al leer $b$, se permanece en $q_0$.  
       - Al leer $a$, se transita a $q_1$.
     - En $q_1$:  
       - Al leer $b$, se permanece en $q_1$.  
       - Al leer $a$, se transita a $q_2$.
     - En $q_2$:  
       - Al leer $b$, se permanece en $q_2$.  
       - Al leer $a$, se transita a un estado de no-aceptación (estado limbo) pues se excede el límite de dos $a$’s.
     
   **Observación:**  
   Solo las trayectorias que permanezcan dentro de $q_2$ (o se detengan allí) al final del procesamiento serán aceptadas.


#### **Ejemplo 4: $ L = $ Lenguaje de las cadenas de $\{0,1\}$ que tienen un número par de símbolos**  
**Alfabeto:**  
$\Sigma = \{0,1\}$.

**Lenguaje:**  
El lenguaje consiste en todas las cadenas de longitud par. Se puede observar que una cadena de longitud par es la concatenación de bloques de 2 símbolos.

**Diseño del AFD:**

1. **Análisis del Lenguaje:**  
   Se debe “contar” la longitud:  
   - Estado inicial con longitud 0 (par).  
   - Tras leer 1 símbolo (impar), se debe transitar a un estado que indica longitud impar.  
   - Después de leer el segundo símbolo (par), se regresa al estado de longitud par.
   
2. **Construcción del Autómata:**  

<p float="center">
  <img src="https://github.com/carlosalvarezh/TeoriaComputacion/blob/main/images/TC03_Graph11.PNG?raw=true" width="250" />
</p>

   - **Estados:**  
     - $q_0$: Estado de longitud par (estado inicial y de aceptación).  
     - $q_1$: Estado de longitud impar (no es de aceptación).
   - **Transiciones:**  
     - Desde $q_0$: Al leer cualquier símbolo (0 o 1), se transita a $q_1$.  
     - Desde $q_1$: Al leer cualquier símbolo, se regresa a $q_0$.
     
   **Observación:**  
   La cadena vacía, al tener longitud 0, se acepta (estando en $q_0$).

#### **Ejemplo 5. Lenguaje de Cadenas con un Número Par de Ceros**

**Descripción del Lenguaje**

- **Alfabeto:** $\Sigma = \{0, 1\}$.  
- **Lenguaje:** Conjunto de todas las cadenas de $\Sigma^*$ que contienen un **número par de ceros** (0).  
  - Ejemplos de cadenas en este lenguaje:  
    - $\lambda$ (la cadena vacía, que tiene 0 ceros, y 0 es par),  
    - $11$ (no contiene ceros, es par de ceros),  
    - $01$ (contiene 1 cero, no está en el lenguaje),  
    - $001$ (contiene 2 ceros, sí está en el lenguaje),  
    - $100100$ (contiene 3 ceros, no está en el lenguaje), etc.

**Idea de Diseño**

1. **Contar la Paridad de Ceros:**  
   Definimos dos estados para llevar la cuenta de la paridad de la cantidad de ceros encontrados hasta el momento:
   - $q_0$: Estado en el que el número de ceros leídos es **par**.  
   - $q_1$: Estado en el que el número de ceros leídos es **impar**.

2. **Estado Inicial y Aceptación:**  
   - El autómata arranca en $q_0$ (par = 0 ceros).  
   - $q_0$ es también un estado de **aceptación**, pues 0 ceros es par.

3. **Transiciones:**  
   - Cuando se lee un “0”:
     - Desde $q_0$ (par), se pasa a $q_1$ (ahora el conteo es impar).  
     - Desde $q_1$ (impar), se pasa a $q_0$ (par).
   - Cuando se lee un “1”:
     - Se permanece en el mismo estado, ya que no se modifica el conteo de ceros.

### Autómata Resultante

- **Estados:**
  - $q_0$: estado inicial y de aceptación (representado con doble círculo).  
  - $q_1$: estado no aceptante.

- **Diagrama:**

<p float="center">
  <img src="https://github.com/carlosalvarezh/TeoriaComputacion/blob/main/images/TC03_Graph12.PNG?raw=true" width="250" />
</p>

  $$
  \begin{aligned}
  &q_0 \xrightarrow{0} q_1, \quad q_0 \xrightarrow{1} q_0 \\
  &q_1 \xrightarrow{0} q_0, \quad q_1 \xrightarrow{1} q_1
  \end{aligned}
  $$

En el diagrama, el lazo en $q_0$ para el símbolo `1` indica que, si no se incrementa el número de ceros, se permanece en el mismo estado par. De manera análoga, en $q_1$ se permanece si se lee `1`, y se conmuta a $q_0$ si se lee `0`.

#### **Ejemplo 6. Lenguaje de Cadenas que Terminan en “b”**

**Descripción del Lenguaje**

- **Alfabeto:** $\Sigma = \{a, b\}$.  
- **Lenguaje:** Todas las cadenas que **terminan con el símbolo** `b`.  
  - Ejemplos de cadenas que pertenecen al lenguaje: `b`, `ab`, `bbb`, `aab`, etc.  
  - Ejemplos de cadenas que **no** pertenecen: `a`, `abba`, `\lambda` (la vacía), etc.

**Idea de Diseño**

1. **Estados según el último símbolo leído:**  
   - $q_0$: estado donde la última letra leída **no** es `b` (o no se ha leído nada aún).  
   - $q_1$: estado donde la última letra leída **es** `b`.

2. **Estado Inicial y de Aceptación:**  
   - El autómata inicia en $q_0$ (pues no ha leído nada aún, y no sabemos si al final terminará en b).  
   - $q_1$ es un estado de aceptación, ya que significa “la cadena actual termina en b”.

3. **Transiciones:**  
   - Desde $q_0$:  
     - Si se lee `a`, se sigue sin terminar en `b`, por lo que se permanece en $q_0$.  
     - Si se lee `b`, ahora la cadena termina en `b`, así que se pasa a $q_1$.
   - Desde $q_1$:  
     - Si se lee `a`, la cadena deja de terminar en `b`, así que se va a $q_0$.  
     - Si se lee `b`, se permanece terminando en `b`, quedando en $q_1$.

**Autómata Resultante**

- **Estados:**
  - $q_0$: estado inicial, no aceptante.  
  - $q_1$: estado de aceptación (doble círculo).
  
- **Diagrama:**  
<p float="center">
  <img src="https://github.com/carlosalvarezh/TeoriaComputacion/blob/main/images/TC03_Graph13.PNG?raw=true" width="250" />
</p>

  $$
  \begin{aligned}
  &q_0 \xrightarrow{a} q_0, \quad q_0 \xrightarrow{b} q_1 \\
  &q_1 \xrightarrow{a} q_0, \quad q_1 \xrightarrow{b} q_1
  \end{aligned}
  $$

De este modo, una cadena se considera aceptada si y solo si, tras su lectura completa, el autómata se encuentra en $q_1$. Por ejemplo, si la cadena es `abb`, la trayectoria sería:

- Inicia en $q_0$.  
- Lee `a`: permanece en $q_0$.  
- Lee `b`: pasa a $q_1$.  
- Lee `b`: permanece en $q_1$.  
- Al terminar, está en $q_1$ (aceptación).


### **Observaciones Generales a estos ejemplos**

- **Estado Limbo:**  
  Aunque no aparezca en los diagramas, las transiciones que no están dibujadas van a un estado de rechazo (o trampa). Esto ayuda a que el diagrama muestre solamente aquellas trayectorias que potencialmente pueden conducir a la aceptación.

- **Estado Inicial vs. Estado de Aceptación:**  
  Tener cuidado en distinguir cuándo un estado inicial es de aceptación y cuándo no.  
  - En el ejemplo del número par de ceros, $q_0$ es a la vez inicial y de aceptación, ya que la cadena vacía ($\lambda$) contiene 0 ceros, lo cual es par.  
  - En el ejemplo de “terminar en b”, el estado inicial no es de aceptación, pues una cadena sin leer (vacía) no termina en `b`.

- **Extensión a Otros Modelos:**  
  Los mismos principios se aplican al diseñar autómatas para cualquier lenguaje regular que se describa por condiciones como “tener un número determinado de apariciones”, “terminar con cierto símbolo”, “ser de longitud par o impar”, etc.

  La clave está en definir estados que representen la información relevante para el lenguaje (por ejemplo, si el conteo de ceros es par o impar, o si el último símbolo leído es `b`), y luego dibujar las transiciones adecuadas con base en el símbolo que se lee. Con la convención de omitir arcos al estado de trampa, los diagramas resultan más manejables y se centran en las trayectorias que pueden llevar a la aceptación.
