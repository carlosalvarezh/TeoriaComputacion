# **Capítulo 2: Autómatas Finitos**

En el capítulo anterior se definieron los conceptos fundamentales de la teoría de la computación: alfabetos, cadenas y lenguajes. Ahora construimos el primer modelo de **máquina abstracta** capaz de decidir si una cadena pertenece o no a un lenguaje dado.

El **autómata finito** procesa una cadena símbolo a símbolo y emite al final un veredicto: *acepta* o *rechaza*. Su restricción central es que solo dispone de **memoria finita** — únicamente recuerda en qué estado interno se encuentra, sin poder almacenar ni modificar la cadena de entrada.

El capítulo está organizado de la siguiente manera:

1. Definición formal y funcionamiento del **autómata finito determinista (AFD)**.
2. Representación mediante **diagramas de transición**.
3. Metodología y ejemplos de **diseño de AFD** para lenguajes regulares concretos.

El capítulo siguiente (TC04) extenderá este modelo al **autómata finito no determinista (AFN)**, que añade flexibilidad de diseño sin ampliar la clase de lenguajes reconocibles: ambos modelos reconocen exactamente los **lenguajes regulares**.

Los autómatas finitos son modelos abstractos de máquinas que procesan cadenas de entrada y, en función de su estructura y transiciones, deciden si aceptan o rechazan dichas cadenas. Estos modelos surgieron en el contexto del estudio de la computación y tienen una amplia aplicación en el diseño de lenguajes de programación, compiladores y sistemas de control.

## **2.1 Introducción a los Autómatas Finitos**

Históricamente, el estudio de los autómatas tiene sus orígenes en los trabajos pioneros de Alan Turing en 1936, quien propuso la “máquina de Turing” para formalizar el concepto de computación. Posteriormente, durante las décadas de 1940 y 1950, se investigaron variantes más restringidas de estos modelos, lo que llevó al surgimiento de los **autómatas finitos**.

En esencia, un autómata finito es una máquina abstracta que procesa una cadena de entrada (una secuencia de símbolos) y, mediante un conjunto finito de estados y reglas de transición, determina si la cadena es aceptada o rechazada. La importancia de estos modelos radica en su capacidad para reconocer lenguajes regulares, que son fundamentales en la teoría de la computación y en aplicaciones prácticas como la verificación de patrones en cadenas (por ejemplo, en expresiones regulares).

## **2.2 Autómatas Finitos Deterministas (AFD)**

### **2.2.1 Definición Formal**

Un **Autómata Finito Determinista (AFD)** se define formalmente como una quíntupla

$$
M = (\Sigma, Q, q_0, F, \delta)
$$

donde cada componente cumple un papel preciso:

| Componente | Nombre formal | Descripción |
|------------|---------------|-------------|
| $\Sigma$ | Alfabeto de entrada | Conjunto finito de símbolos que pueden aparecer en la cadena de entrada. Todas las cadenas procesadas pertenecen a $\Sigma^*$. |
| $Q$ | Conjunto de estados | Conjunto finito de configuraciones internas posibles del autómata. |
| $q_0 \in Q$ | Estado inicial | Estado en el que se encuentra el autómata antes de leer cualquier símbolo. |
| $F \subseteq Q,\; F \neq \varnothing$ | Estados de aceptación | Si al terminar de leer la cadena el autómata está en algún $q \in F$, la cadena es **aceptada**. |
| $\delta: Q \times \Sigma \to Q$ | Función de transición | Para cada par (estado, símbolo) determina de forma **única** el estado siguiente: $\delta(q, s) = q'$. |

**De la quíntupla a la tabla de transición**

La función $\delta$ se puede representar de forma equivalente como una **tabla**, donde las filas son los estados y las columnas son los símbolos del alfabeto. Cada celda contiene el estado destino:

$$
\begin{array}{c|cc}
\delta & s_1 & s_2 \\ \hline
q_0 & \delta(q_0,s_1) & \delta(q_0,s_2) \\
q_1 & \delta(q_1,s_1) & \delta(q_1,s_2) \\
\end{array}
$$

Esta tabla tiene exactamente $|Q| \times |\Sigma|$ celdas — una por cada combinación posible de estado y símbolo. El hecho de que **todas** las celdas estén definidas es lo que hace al autómata *determinista*: nunca hay ambigüedad sobre adónde ir.

**Traducción coloquial de cada componente**

| Componente | "Traducción" intuitiva | Pregunta que responde |
|------------|------------------------|-----------------------|
| $\Sigma$ | Lista de "eventos" posibles | ¿Qué tipos de entrada puedo recibir? |
| $Q$ | Las posibles "situaciones" del sistema | ¿En qué situación estoy ahora? |
| $q_0$ | Punto de partida | ¿Dónde comienzo antes de leer nada? |
| $F$ | Situaciones "exitosas" | ¿Cuándo considero la operación un éxito? |
| $\delta$ | Reglas del tipo *si-estoy-aquí y veo-esto → voy-allá* | ¿Cómo cambio de situación al recibir un símbolo? |

#### **Ejemplo: torniquete de acceso al Metro de Medellín**

Imagina el proceso de entrar a una estación del Metro:

**1. Alfabeto ($\Sigma$)**
$$\Sigma=\{T,\;I\}$$
- **T** = tarjeta contactless con saldo válido.
- **I** = tarjeta inválida (sin saldo o tarjeta incorrecta).

**2. Estados ($Q$)**
$$Q=\{q_\text{fuera},\;q_\text{bloqueado},\;q_\text{dentro}\}$$

| Estado | Significado |
|--------|-------------|
| $q_\text{fuera}$ | Persona aún fuera de la zona paga |
| $q_\text{bloqueado}$ | Tarjeta rechazada; torniquete bloqueado |
| $q_\text{dentro}$ | Persona ya cruzó el torniquete |

**3. Estado inicial y de aceptación**
$$q_0 = q_\text{fuera}, \qquad F = \{q_\text{dentro}\}$$

**4. Función de transición ($\delta$)**

| Estado actual | $T$ (válida) | $I$ (inválida) |
|:-------------:|:------------:|:--------------:|
| $q_\text{fuera}$ | $q_\text{dentro}$ | $q_\text{bloqueado}$ |
| $q_\text{bloqueado}$ | $q_\text{dentro}$ | $q_\text{bloqueado}$ |
| $q_\text{dentro}$ | $q_\text{dentro}$ | $q_\text{dentro}$ |

La tabla cubre las $3 \times 2 = 6$ combinaciones posibles: $\delta$ está **totalmente definida**.

**5. Ejecución de tres cadenas**

- **`T`**: $q_\text{fuera} \xrightarrow{T} q_\text{dentro} \in F$ ⇒ **Aceptada** (entró directamente).
- **`I T`**: $q_\text{fuera} \xrightarrow{I} q_\text{bloqueado} \xrightarrow{T} q_\text{dentro} \in F$ ⇒ **Aceptada** (segundo intento exitoso).
- **`I I`**: $q_\text{fuera} \xrightarrow{I} q_\text{bloqueado} \xrightarrow{I} q_\text{bloqueado} \notin F$ ⇒ **Rechazada** (nunca cruzó).

Sin importar cuántos intentos haga, en cada paso existe **una única** transición definida: eso es lo que hace al autómata **determinista**.

#### **Ideas clave**

1. **Determinista** — para cada par *(estado, símbolo)* hay **exactamente una** transición de salida. No hay ambigüedad.
2. **Finito** — tanto $\Sigma$ como $Q$ tienen tamaño limitado. La máquina no puede crecer.
3. **Aceptación** — la historia completa de la cadena termina en un estado "bueno" ($q \in F$).
4. **La tabla = la función** — la tabla de transición es simplemente otra forma de escribir $\delta$; ambas representaciones son equivalentes y se usarán indistintamente.

### **2.2.2 Funcionamiento del AFD**

El proceso de computación en un AFD se describe de la siguiente manera:

- **Inicio:** La cadena de entrada $u$ se coloca en una cinta semiinfinita, divida en celdas, donde cada celda contiene un símbolo de $u$. El primer símbolo de $u$ se ubica en la primera celda de la cinta.
  
- **Unidad de Control:** El autómata posee una *unidad de control* (o *cabeza lectora*) que comienza en el estado inicial $q_0$ y se sitúa sobre la primera celda de la cinta.  

<p float="center">
  <img src="https://github.com/carlosalvarezh/TeoriaComputacion/blob/main/images/TC03_AutomataFinito02.PNG?raw=true" width="350" />
</p>

- **Transiciones:** La función de transición $\delta$ indica al autómata qué hacer al leer un símbolo $s$ en un estado $q$. Concretamente, si se tiene $\delta(q, s) = q'$, el autómata:
  1. Cambia de estado, pasando de $q$ a $q'$.
  2. Se desplaza **hacia la derecha** para leer el siguiente símbolo.
  
  Cada cambio de estado y movimiento constituye un **paso computacional**.

<p float="center">
  <img src="https://github.com/carlosalvarezh/TeoriaComputacion/blob/main/images/TC03_AutomataFinito03.PNG?raw=true" width="350" />
</p>

- **Finalización:** El autómata procesa la cadena hasta que alcanza la primera celda vacía. La cadena de entrada se acepta si el autómata termina en un estado de aceptación ($q \in F$); de lo contrario, se rechaza.

- **Caso Especial:** La cadena vacía $\lambda$ es aceptada si, y solo si, el estado inicial $q_0$ se encuentra en el conjunto $F$.

#### **Resumen: Cómo “corre” un AFD paso a paso**

1. La cinta es solo la historia completa de lo que va a pasar.
   Cada símbolo ocupa una celda; el *cursor* empieza en la primera.

2. La unidad de control es un “punto” que recuerda su estado.
   Arranca en el estado inicial $q_0$ y “mira” la primera celda.

3. Regla de oro (función $\delta$).
   – Ve un símbolo $s$ estando en $q$.
   – Consulta la tabla: “si estoy en $q$ y veo $s$, salto a $q’$”.
   – Se mueve una celda a la derecha.
   Eso cuenta como un paso computacional. Repite hasta llegar a la primera celda vacía (fin de la cadena).

4. Decisión final.
   Si al terminar la lectura el estado pertenece a $F$, la cadena se acepta; si no, se rechaza.

5. Cadena vacía $\lambda$.
   La unidad de control ni siquiera avanza; se acepta solo si $q_0\in F$.



#### **Ejemplo**

Consideremos el siguiente autómata finito determinista:

- **Alfabeto:** $\Sigma = \{a, b\}$.
- **Conjunto de Estados:** $Q = \{q_0, q_1, q_2\}$.
- **Estado Inicial:** $q_0$.
- **Estados de Aceptación:** $F = \{q_0, q_2\}$.
- **Función de Transición:** Se define en la siguiente tabla:

  | $\delta$ | $a$ | $b$ |
  |------------|-------|-------|
  | **$q_0$**  | $q_0$ | $q_1$ |
  | **$q_1$**  | $q_1$ | $q_2$ |
  | **$q_2$**  | $q_1$ | $q_1$ |

  Esto se lee de la siguiente manera:
  - $\delta(q_0, a) = q_0$ y $\delta(q_0, b) = q_1$.
  - $\delta(q_1, a) = q_1$ y $\delta(q_1, b) = q_2$.
  - $\delta(q_2, a) = q_1$ y $\delta(q_2, b) = q_1$.

**Procesamiento de Cadenas:**

1. **Cadena $u = aabab$:**
   - El autómata inicia en $q_0$.
   - Lee el primer $a$: $\delta(q_0, a) = q_0$.
   - Lee el segundo $a$: $\delta(q_0, a) = q_0$.
   - Lee $b$: $\delta(q_0, b) = q_1$.
   - Lee $a$: $\delta(q_1, a) = q_1$.
   - Lee $b$: $\delta(q_1, b) = q_2$.
   - La cadena se procesa completamente y el autómata termina en $q_2$, que es un estado de aceptación. Por lo tanto, $u$ es aceptada.

2. **Cadena $v = aabab a$:**
   - El procesamiento es similar, pero al final el autómata termina en $q_1$, que no es un estado de aceptación. Por lo tanto, $v$ es rechazada.

3. **Cadena Vacía $\lambda$:**
   - La cinta está vacía y el autómata no realiza transiciones.
   - Como el estado inicial $q_0$ está en $F$, $\lambda$ es aceptada.

#### **Ejemplo – Torniquete del Metro (misma idea, ahora “animada”)**

| Símbolo en la cinta | Significado                                             |
| :------------------:| ------------------------------------------------------- |
| T                   | Tarjeta con saldo válido (lector emite “beep” verde)    |
| I                   | Tarjeta inválida / sin saldo (lector emite “beep” rojo) |

**Conjunto de estados**

$$
Q=\{q_{\text{fuera}},\;q_{\text{dentro}},\;q_{\text{bloq}}\}
$$

| Estado              | Qué representa                     | Está en $F$? |
| ------------------- | ---------------------------------- | ------------ |
| $q_{\text{fuera}}$  | Persona antes del torniquete       | No           |
| $q_{\text{bloq}}$   | Torniquete bloqueado tras un error | No           |
| $q_{\text{dentro}}$ | Persona ya cruzó                   | Sí       |

$q_0 = q_{\text{fuera}}$ y $F=\{q_{\text{dentro}}\}$.

**Tabla de transición $\delta$**

| Estado actual       | T               | I               |
| ------------------- | ------------------- | ------------------- |
| $q_{\text{fuera}}$  | $q_{\text{dentro}}$ | $q_{\text{bloq}}$   |
| $q_{\text{bloq}}$   | $q_{\text{dentro}}$ | $q_{\text{bloq}}$   |
| $q_{\text{dentro}}$ | $q_{\text{dentro}}$ | $q_{\text{dentro}}$ |

**Cómo se ve la “película” sobre la cinta**

- **Cadena 1: `T`**

| Paso | Celda leída | Estado antes       | Estado después      |
| ---- | ----------- | ------------------ | ------------------- |
| 1    | T       | $q_{\text{fuera}}$ | $q_{\text{dentro}}$ |

Se llega al final de la cinta en $q_{\text{dentro}}\in F$ ⇒ Aceptada (persona entró al sistema).

- **Cadena 2: `I T`**

| Paso | Celda | Estado antes       | Estado después      |
| ---- | ----- | ------------------ | ------------------- |
| 1    | I | $q_{\text{fuera}}$ | $q_{\text{bloq}}$   |
| 2    | T | $q_{\text{bloq}}$  | $q_{\text{dentro}}$ |

Fin de la cinta, estado aceptado ⇒ finalmente logró pasar (segundo intento).

- **Cadena 3: `I I`**

| Paso | Celda | Estado antes       | Estado después    |
| ---- | ----- | ------------------ | ----------------- |
| 1    | I | $q_{\text{fuera}}$ | $q_{\text{bloq}}$ |
| 2    | I | $q_{\text{bloq}}$  | $q_{\text{bloq}}$ |

Termina en $q_{\text{bloq}}\notin F$ ⇒ Rechazada (nunca logró cruzar).

- **Cadena vacía $\lambda$**

* No se leen celdas; permanecemos en $q_{\text{fuera}}\notin F$ ⇒ Rechazada.
  (Solo se aceptaría si el torniquete considerara “ya estás dentro” desde el inicio, es decir, si poníamos $q_{\text{fuera}}\in F$).

**Resumiendo**

- La cinta = la secuencia de eventos que suceden (intentos de pago, luces del semáforo, caracteres de una contraseña…).
- La cabeza = “dónde voy” y “qué veo ahora”.
- Determinismo = nunca dudo: por cada pareja *(estado, símbolo)* hay exactamente una flecha de salida en la tabla.
- Aceptación = la historia termina en un estado “final feliz”.

### **2.2.3 Observaciones y Propiedades**

- **Determinismo:**
  En un AFD, para cada combinación de estado $q \in Q$ y símbolo $s \in \Sigma$, la función $\delta(q, s)$ está definida de forma única. Esto significa que no hay ambigüedad en la evolución del autómata: dado el estado actual y el símbolo leído, el siguiente estado es siempre exactamente uno.

- **Totalidad de $\delta$:**
  La función de transición está definida para **todas** las combinaciones posibles de estado y símbolo. En la práctica, cuando un símbolo en un estado conduce al rechazo, se representa con un estado *trampa* implícito (ver sección 2.4.1). La tabla de transición siempre tiene $|Q| \times |\Sigma|$ entradas.

- **Direccionalidad:**
  La unidad de control siempre se desplaza hacia la derecha y no tiene capacidad para retroceder ni modificar el contenido de la cinta. La cadena se lee exactamente una vez, de izquierda a derecha.

- **Aceptación de la cadena vacía:**
  La cadena $\lambda$ es aceptada si, y solo si, el estado inicial $q_0$ pertenece al conjunto $F$. En ese caso, el autómata ya está en un estado de aceptación antes de leer ningún símbolo.

- **La tabla como representación de $\delta$:**
  La tabla de transición es equivalente a la función $\delta$. Cada fila corresponde a un estado, cada columna a un símbolo del alfabeto, y el contenido de cada celda es el estado destino. Leer la tabla es exactamente lo mismo que evaluar $\delta$ en cada par (estado, símbolo).

El **lenguaje aceptado** por el autómata $M$ se denota $L(M)$ y se define como el conjunto de todas las cadenas que, al ser procesadas por $M$, lo llevan a un estado de aceptación:

$$
L(M) = \{\, u \in \Sigma^* \mid \delta(q_0, u) \in F \,\}.
$$

> **Lectura:** $L(M)$ es el conjunto de todas las cadenas $u$ tal que, partiendo del estado inicial $q_0$ y procesando $u$ símbolo a símbolo, el autómata termina en algún estado de $F$.

## **2.3 Diagrama de transiciones de un autómata**

Un autómata finito se representa visualmente mediante un **grafo dirigido y etiquetado** (digrafo etiquetado), donde:

- Los **vértices** (nodos) representan los estados del autómata.
- Las **aristas** (flechas) representan las transiciones: una flecha de $q$ a $p$ etiquetada con $s$ indica que $\delta(q, s) = p$.

Las convenciones gráficas son:

| Elemento | Representación gráfica |
|----------|------------------------|
| Estado $q$ | Círculo con el nombre $q$ |
| Estado inicial $q_0$ | Flecha entrante sin origen (o flecha doble) |
| Estado de aceptación $q \in F$ | Círculo con **doble borde** |
| Transición $\delta(q, s) = p$ | Flecha de $q$ a $p$ etiquetada con $s$ |

Una cadena $u$ es **aceptada** si existe una trayectoria en el diagrama, comenzando en $q_0$, que al recorrer las aristas con los símbolos de $u$ en orden termina en un nodo con doble borde.

**Ejemplo: diagrama del autómata de la sección 2.2.2**

La función de transición del autómata con $Q=\{q_0,q_1,q_2\}$ y $\Sigma=\{a,b\}$ es:

| $\delta$ | $a$ | $b$ |
|----------|-----|-----|
| $q_0$ | $q_0$ | $q_1$ |
| $q_1$ | $q_1$ | $q_2$ |
| $q_2$ | $q_1$ | $q_1$ |

Con $F = \{q_0, q_2\}$.

$$
\delta(q_0,a)=q_0,\quad\delta(q_0,b)=q_1,\quad\delta(q_1,a)=q_1,\quad\delta(q_1,b)=q_2,\quad\delta(q_2,a)=q_1,\quad\delta(q_2,b)=q_1
$$

</p>

Examinando el diagrama se puede observar directamente que la cadena $aaababbb$ es aceptada (sigue un recorrido que termina en $q_2 \in F$), mientras que $aabaaaba$ es rechazada (termina en $q_1 \notin F$).

> **Por qué el diagrama es útil:** permite identificar de un vistazo las trayectorias de aceptación sin necesidad de calcular $\delta$ paso a paso. Es la herramienta principal para verificar o comunicar el diseño de un autómata.

---

**Ejemplo: diagrama del torniquete del Metro de Medellín**

El mismo autómata presentado en la sección 2.2.1 se representa gráficamente así:

| Estado | Descripción | ¿Aceptación? |
|--------|-------------|:------------:|
| $q_\text{fuera}$ | Antes del torniquete | No |
| $q_\text{bloqueado}$ | Tarjeta rechazada | No |
| $q_\text{dentro}$ | Ya cruzó | **Sí** |

**Tabla de transición $\delta$:**

| Origen | $T$ | $I$ |
|--------|-----|-----|
| $q_\text{fuera}$ | $q_\text{dentro}$ | $q_\text{bloqueado}$ |
| $q_\text{bloqueado}$ | $q_\text{dentro}$ | $q_\text{bloqueado}$ |
| $q_\text{dentro}$ | $q_\text{dentro}$ | $q_\text{dentro}$ |

<p float="center">

</p>

**Cómo leer el diagrama:**

* El nodo $q_\text{dentro}$ tiene doble borde → es el único estado de aceptación.
* La flecha de entrada apunta a $q_\text{fuera}$ → es el estado inicial.
* Cada flecha muestra qué ocurre al presentar cada tipo de tarjeta desde cada estado.

> La ejecución paso a paso sobre cadenas concretas (`T`, `I T`, `I I`) fue desarrollada en detalle en la sección 2.2.2.

**Lenguaje reconocido:**

$$
L = \{\,w \in \{T,I\}^* \mid w \text{ contiene al menos una } T\,\}.
$$

Se verifica: cualquier cadena con al menos un $T$ tiene un camino hasta $q_\text{dentro}$, donde todas las transiciones posteriores son lazos (permanece en aceptación). Una cadena sin $T$ nunca abandona $q_\text{fuera}$ o $q_\text{bloqueado}$, que no son de aceptación.

## **2.4 Diseño de Autómatas**

Los **autómatas finitos** son máquinas abstractas que, mediante un conjunto finito de estados y transiciones, "leen" una cadena de entrada y deciden si ésta pertenece o no a un lenguaje. En el proceso de diseñar tales autómatas, especialmente los deterministas, se han adoptado ciertas convenciones que simplifican los diagramas de transición.

---

### **2.4.1 Convenciones en los Diagramas de Transición**

En los diagramas de autómatas finitos deterministas (AFD) se utiliza la siguiente **convención adicional**:  

- **Estado Limbo (o estado trampa):** Se asume que cualquier arco o transición que no se dibuje explícitamente conduce a un estado "limbo" o trampa, el cual no es un estado de aceptación. Esto significa que solo se muestran los arcos que forman parte de trayectorias que, eventualmente, pueden conducir a la aceptación de una cadena.  


  **Ventaja:** Esta convención simplifica los diagramas, ya que se evita la saturación visual que produciría dibujar todas las transiciones posibles (incluso aquellas que resultan en rechazo).

### **2.4.2 Tipos de Problemas Abordados**

En el estudio y diseño de autómatas, se abordan dos tipos fundamentales de problemas:

1. **Diseño (Construcción) del Autómata a partir de un Lenguaje Regular:** Se busca, dada una descripción formal de un lenguaje regular $ L $, construir un autómata $ M $ tal que $ L(M) = L $. Es decir, el autómata $(M)$ debe reconocer (aceptar) exactamente todas y solo las cadenas que pertenezcan a $ L $.

2. **Determinación del Lenguaje Aceptado por un Autómata:**  
   Dado un autómata $ M $ (con una descripción o diagrama de transición), se busca determinar el conjunto $ L(M) $ de todas las cadenas que $ M $ acepta.

En este capítulo nos centraremos inicialmente en problemas del primer tipo: construir autómatas a partir de lenguajes regulares.

### **2.4.3 Ejemplos de Diseño de Autómatas**

Diseñar un AFD para un lenguaje dado es un proceso sistemático. La siguiente estrategia en cinco pasos sirve como guía general:

| Paso | Pregunta clave | Acción |
|------|---------------|--------|
| 1. Identificar la información relevante | ¿Qué debe "recordar" el autómata para decidir la aceptación? | Determinar las distintas situaciones posibles al leer la cadena. |
| 2. Crear los estados | ¿Cuántas situaciones distintas existen? | Asignar un estado a cada situación identificada. |
| 3. Establecer el estado inicial | ¿Cómo comienza el autómata, sin haber leído nada? | Marcar el estado que representa la situación inicial. |
| 4. Establecer los estados de aceptación | ¿En qué situaciones la cadena leída hasta ahora pertenece al lenguaje? | Marcar como aceptadores los estados correspondientes. |
| 5. Definir las transiciones | Al estar en una situación y leer un símbolo, ¿a qué situación se pasa? | Completar la función de transición $\delta$ para cada par $(q, s)$. |

> **Nota sobre el estado limbo:** Si para algún par $(q, s)$ ningún estado del diseño es apropiado (la cadena ya no puede pertenecer al lenguaje), la transición conduce a un **estado trampa** (limbo) implícito desde el que no hay salida. Por claridad, los diagramas suelen omitir este estado.

A continuación se presentan varios ejemplos desarrollados paso a paso. En cada caso se muestra una **versión completa** (con todas las transiciones, incluidas las que van al limbo) y una **versión simplificada** (omitiendo los arcos hacia el estado limbo).

#### **Ejemplo 1: $ L = a^* $**  
**Lenguaje:**  
$$
L = a^* = \{\lambda,\, a,\, a^2,\, a^3,\, \dots\}
$$
Es decir, todas las cadenas (incluida la vacía) compuestas únicamente del símbolo $a$.

**Diseño del AFD:**

1. **Análisis del Lenguaje:**  
   Toda cadena en $L$ consiste en cero o más repeticiones de $a$.  
   Se debe aceptar la cadena vacía $\lambda$ y cualquier secuencia de $a$'s.

2. **Construcción del Autómata:**  
   - **Estados:** Se puede diseñar un autómata con **un único estado** $q_0$ que actúa tanto de estado inicial como de estado de aceptación.
   - **Transición para $a$:** Desde $q_0$, al leer $a$, se permanece en $q_0$.
   - **Transiciones no especificadas:** Cualquier transición no dibujada (por ejemplo, lecturas de otros símbolos) se asume que llevan al estado limbo de rechazo.

3. **Observación:**  
   El autómata acepta $\lambda$ ya que el estado inicial $q_0$ es de aceptación.

**Analogía cotidiana:**  

Una máquina contadora de fichas que solo reconoce fichas rojas ilustra el mismo principio. La máquina comienza en un estado de conteo activo: cada vez que se introduce una ficha roja, la máquina sigue en ese estado y actualiza el total. Si se introduce una ficha de cualquier otro color, la máquina pasa inmediatamente a un modo de error del que no sale; las fichas siguientes ya no se cuentan.  

El esquema interno de control es idéntico al autómata para $a^{*}$: un estado que acepta las "entradas correctas" y un estado trampa implícito que absorbe cualquier entrada incorrecta.

#### **Ejemplo 2: $ L = a^+ $**  
**Lenguaje:**  
$$
L = a^+ = \{a,\, a^2,\, a^3,\, \dots\}
$$
Es decir, todas las cadenas compuestas de al menos una $a$.

**Diseño del AFD:**

1. **Análisis del Lenguaje:**  
   A diferencia de $a^*$, el lenguaje $a^+$ no acepta la cadena vacía, por lo que el autómata debe rechazar $\lambda$.

2. **Construcción del Autómata:**  
   - **Estados:**  
     Se requiere al menos dos estados:  
     - $q_0$: estado inicial, **no es de aceptación**.  
     - $q_1$: estado al que se transita tras leer al menos un $a$; este es un **estado de aceptación**.
   - **Transiciones:**  
     - Desde $q_0$, al leer $a$, se transita a $q_1$.  
     - Desde $q_1$, al leer $a$, se permanece en $q_1$ (lazo sobre sí mismo).  
     
   **Diagrama Completo:**

3. **Observación:**  
   La cadena vacía $\lambda$ es rechazada porque $q_0$ no es de aceptación.

#### **Ejemplo 3: $ L = $ Lenguaje de las cadenas que contienen exactamente dos $a$'s**  
**Alfabeto:** $\Sigma = \{a, b\}$.

**Lenguaje:**  
$$
L = b^* a \, b^* a \, b^*
$$
Todas las cadenas con exactamente dos ocurrencias del símbolo $a$; pueden aparecer cualquier cantidad de $b$'s antes, entre y después de las dos $a$'s.

**Diseño del AFD:**

1. **Análisis del Lenguaje:**  
   La información relevante es el número de $a$'s leídas hasta el momento: 0, 1 o ≥ 2.  
   - 0 $a$'s leídas → estado inicial.  
   - 1 $a$ leída → estado intermedio.  
   - 2 $a$'s leídas → estado de aceptación.  
   - Más de 2 $a$'s → estado trampa (rechazo).

2. **Construcción del Autómata:**  

   | Estado | Significado | Aceptador |
   |--------|-------------|-----------|
   | $q_0$ | 0 $a$'s leídas | No |
   | $q_1$ | 1 $a$ leída | No |
   | $q_2$ | 2 $a$'s leídas | **Sí** |
   | $q_3$ | >2 $a$'s (trampa) | No |

   **Tabla de transición:**

   | Estado | $a$ | $b$ |
   |--------|-----|-----|
   | $q_0$ | $q_1$ | $q_0$ |
   | $q_1$ | $q_2$ | $q_1$ |
   | $q_2$ | $q_3$ | $q_2$ |
   | $q_3$ | $q_3$ | $q_3$ |

   **Diagrama:**  

<p float="center">
  <img src="https://github.com/carlosalvarezh/TeoriaComputacion/blob/main/images/TC03_Graph10.PNG?raw=true" width="350" />
</p>

3. **Caracterización formal:**  
$$
L(M) = \{w \in \{a,b\}^* \mid w \text{ contiene exactamente dos ocurrencias de } a\}
$$

4. **Observación:**  
   Solo las cadenas que lleguen a $q_2$ al terminar la lectura son aceptadas. Si llega una tercera $a$, el autómata pasa a $q_3$ y no hay retorno.

#### **Ejemplo 4: $ L = $ Lenguaje de las cadenas de $\{0,1\}$ que tienen longitud par**  
**Alfabeto:** $\Sigma = \{0,1\}$.

**Lenguaje:**  
Todas las cadenas cuya longitud es un número par (0, 2, 4, …). La cadena vacía $\lambda$ tiene longitud 0, que es par, y por tanto pertenece al lenguaje.

**Diseño del AFD:**

1. **Análisis del Lenguaje:**  
   La información relevante es la paridad de la longitud leída hasta el momento.  
   - Longitud par → estado de aceptación.  
   - Longitud impar → estado de no-aceptación.  
   Cada símbolo leído cambia la paridad: par → impar → par → …

2. **Construcción del Autómata:**  

   | Estado | Significado | Aceptador |
   |--------|-------------|-----------|
   | $q_0$ | Longitud par | **Sí** |
   | $q_1$ | Longitud impar | No |

   **Tabla de transición:**

   | Estado | $0$ | $1$ |
   |--------|-----|-----|
   | $q_0$ | $q_1$ | $q_1$ |
   | $q_1$ | $q_0$ | $q_0$ |

   **Diagrama:**  

<p float="center">
  <img src="https://github.com/carlosalvarezh/TeoriaComputacion/blob/main/images/TC03_Graph11.PNG?raw=true" width="250" />
</p>

3. **Caracterización formal:**  
$$
L(M) = \{w \in \{0,1\}^* \mid |w| \text{ es par}\}
$$

4. **Observación:**  
   $q_0$ es a la vez estado inicial y de aceptación: la cadena vacía (longitud 0, par) se acepta. Cualquier símbolo —sea `0` o `1`— cambia la paridad, por lo que las transiciones de $q_0$ y $q_1$ son simétricas en ambos símbolos.

#### **Ejemplo 5: Lenguaje de cadenas con un número par de ceros**

**Alfabeto:** $\Sigma = \{0, 1\}$.  
**Lenguaje:** Todas las cadenas de $\Sigma^*$ que contienen un **número par de ceros** (0, 2, 4, …). Ejemplos:

| Cadena | # de ceros | ¿En $L$? |
|--------|-----------|---------|
| $\lambda$ | 0 (par) | Sí |
| $11$ | 0 (par) | Sí |
| $01$ | 1 (impar) | No |
| $001$ | 2 (par) | Sí |
| $10010$ | 3 (impar) | No |

**Diseño del AFD:**

1. **Análisis del Lenguaje:**  
   La información relevante es la **paridad del número de ceros** leídos hasta el momento.

2. **Construcción del Autómata:**  

   | Estado | Significado | Aceptador |
   |--------|-------------|-----------|
   | $q_0$ | # de ceros leídos es par | **Sí** |
   | $q_1$ | # de ceros leídos es impar | No |

   **Tabla de transición:**

   | Estado | $0$ | $1$ |
   |--------|-----|-----|
   | $q_0$ | $q_1$ | $q_0$ |
   | $q_1$ | $q_0$ | $q_1$ |

   **Diagrama:**

<p float="center">
  <img src="https://github.com/carlosalvarezh/TeoriaComputacion/blob/main/images/TC03_Graph12.PNG?raw=true" width="250" />
</p>

3. **Caracterización formal:**  
$$
L(M) = \{w \in \{0,1\}^* \mid \text{el número de ceros en } w \text{ es par}\}
$$

4. **Observación:**  
   Leer un `1` no cambia la paridad del conteo de ceros, por lo que en ambos estados hay un lazo para el símbolo `1`. Leer un `0` conmuta la paridad, por lo que el `0` siempre cruza entre $q_0$ y $q_1$. Este autómata es análogo al del Ejemplo 4, pero la "paridad" se cuenta solo sobre los ceros, ignorando los unos.

#### **Ejemplo 6: Lenguaje de cadenas que terminan en $b$**

**Alfabeto:** $\Sigma = \{a, b\}$.  
**Lenguaje:** Todas las cadenas que **terminan con el símbolo** $b$.  
- Pertenecen al lenguaje: $b$, $ab$, $bbb$, $aab$, …  
- No pertenecen: $a$, $abba$, $\lambda$, …

**Diseño del AFD:**

1. **Análisis del Lenguaje:**  
   La información relevante es el **último símbolo leído**. Solo hay dos posibilidades: el último símbolo fue $b$, o no (o no se ha leído nada).

2. **Construcción del Autómata:**  

   | Estado | Significado | Aceptador |
   |--------|-------------|-----------|
   | $q_0$ | Último símbolo no es $b$ (o cadena vacía) | No |
   | $q_1$ | Último símbolo es $b$ | **Sí** |

   **Tabla de transición:**

   | Estado | $a$ | $b$ |
   |--------|-----|-----|
   | $q_0$ | $q_0$ | $q_1$ |
   | $q_1$ | $q_0$ | $q_1$ |

   **Diagrama:**  
<p float="center">
  <img src="https://github.com/carlosalvarezh/TeoriaComputacion/blob/main/images/TC03_Graph13.PNG?raw=true" width="250" />
</p>

3. **Caracterización formal:**  
$$
L(M) = \{w \in \{a,b\}^* \mid w \text{ termina en } b\} = \{a,b\}^* b
$$

4. **Ejemplo de ejecución** — cadena `abb`:

   | Paso | Estado actual | Símbolo leído | Estado siguiente |
   |------|--------------|--------------|-----------------|
   | 0 | $q_0$ (inicio) | $a$ | $q_0$ |
   | 1 | $q_0$ | $b$ | $q_1$ |
   | 2 | $q_1$ | $b$ | $q_1$ |
   | — | $q_1$ (fin) | — | **Acepta** |

5. **Observación:**  
   Leer $a$ siempre lleva a $q_0$ (el último símbolo deja de ser $b$); leer $b$ siempre lleva a $q_1$. Por esta razón las transiciones de ambos estados en $a$ y $b$ son idénticas: solo importa cuál fue el último símbolo, no el historial anterior.

### **Observaciones Generales a estos ejemplos**

- **Estado Limbo:**  
  Aunque no aparezca en los diagramas, las transiciones que no están dibujadas van a un estado de rechazo (o trampa). Esto ayuda a que el diagrama muestre solamente aquellas trayectorias que potencialmente pueden conducir a la aceptación.

- **Estado Inicial vs. Estado de Aceptación:**  
  Tener cuidado en distinguir cuándo un estado inicial es de aceptación y cuándo no.  
  - En el ejemplo del número par de ceros, $q_0$ es a la vez inicial y de aceptación, ya que la cadena vacía ($\lambda$) contiene 0 ceros, lo cual es par.  
  - En el ejemplo de “terminar en b”, el estado inicial no es de aceptación, pues una cadena sin leer (vacía) no termina en `b`.

- **Extensión a Otros Modelos:**  
  Los mismos principios se aplican al diseñar autómatas para cualquier lenguaje regular que se describa por condiciones como “tener un número determinado de apariciones”, “terminar con cierto símbolo”, “ser de longitud par o impar”, etc.

  La clave está en definir estados que representen la información relevante para el lenguaje (por ejemplo, si el conteo de ceros es par o impar, o si el último símbolo leído es `b`), y luego dibujar las transiciones adecuadas con base en el símbolo que se lee. Con la convención de omitir arcos al estado de trampa, los diagramas resultan más manejables y se centran en las trayectorias que pueden llevar a la aceptación.
