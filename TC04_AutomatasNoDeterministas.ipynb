# **Capítulo 3: Autómatas Finitos No Deterministas (AFN)**

## **3.1  Autómatas finitos no deterministas (AFN)**

Un autómata finito no determinista extiende la idea del AFD permitiendo que, ante un mismo símbolo de entrada, puedan activarse **varias** transiciones desde el mismo estado, o incluso ninguna.  

La definición formal introduce un ligero cambio: la función de transición ya no devuelve **un** estado, sino un **conjunto** de estados posibles.

## **3.2 Definición formal**

$$
M=\bigl(\Sigma,\,Q,\,q_{0},\,F,\,\Delta\bigr)
$$

| Componente                        | Descripción                                                                     |
| --------------------------------- | ------------------------------------------------------------------------------- |
| $\Sigma$                          | alfabeto de entrada                                                             |
| $Q$                               | conjunto finito de estados                                                      |
| $q_{0}$                           | estado inicial                                                                  |
| $F\subseteq Q,\,F\neq\varnothing$ | estados de aceptación                                                           |
| $\Delta:Q\times\Sigma\to\wp(Q)$   | función de transición; $\wp(Q)$ es el conjunto de todos los subconjuntos de $Q$ |

Para $q\in Q$ y $s\in\Sigma$:

$$
\Delta(q,s)=\{\,q_{i_{1}},q_{i_{2}},\dots,q_{i_{k}}\,\}
$$

*Interpretación:* estando en $q$ y leyendo $s$, el autómata **puede elegir** cualquiera de los estados listados y, tras dar el paso, avanza el cabezal una celda a la derecha.

Si $\Delta(q,s)=\varnothing$ y el autómata llega a la combinación $(q,s)$, el cómputo se **interrumpe** (no hay transición posible desde allí).

## **3.3 Cómputo no determinista**

***Diagramas.*** En el grafo de transiciones esto se traduce en **arcos paralelos** que salen del mismo nodo con la misma etiqueta.

<p float="center">
  <img src="https://github.com/carlosalvarezh/TeoriaComputacion/blob/main/images/TC04_AFN01.PNG?raw=true" width="200" />
</p>

Cada arco representa una alternativa que el autómata puede tomar al procesar ese símbolo.

**Ejecución.** Cuando una cadena $u\in\Sigma^{*}$ se coloca en la cinta, el autómata genera simultáneamente todas las trayectorias compatibles con las reglas de $\Delta$.

* Puede darse el caso de que algunas trayectorias se detengan prematuramente por falta de transición (cómputo abortado, primera figura).
* Otras trayectorias pueden avanzar hasta consumir la cadena completa.

## **3.4 Criterio de aceptación**

El lenguaje reconocido por $M$ es

$$
L(M)=\Bigl\{\,u\in\Sigma^{*}\mid
\exists\text{ al menos un recorrido completo de }u\text{ que termine en }q\in F\Bigr\}.
$$

- *Una cadena se acepta si **alguna** de las posibles ejecuciones lee la cadena entera y finaliza en un estado de aceptación.*  

- No es necesario que **todas** las trayectorias terminen bien; basta con una.

**Observaciones clave**

* **No determinismo lógico.** El modelo no especifica cómo se elige la rama; conceptualmente se exploran todas en paralelo.  

* **Equivalencia con AFD.** Todo lenguaje aceptado por un AFN es regular; existe siempre un AFD equivalente. La diferencia está solo en la comodidad para describir el autómata.  

* **Abortos seguros.** Las trayectorias que se abortan no influyen en el resultado si al menos una rama exitosa llega a $F$.  

* **Compactación de diagramas.** Al permitir arcos duplicados, un AFN suele requerir menos estados y flechas que su versión determinista, lo que simplifica algunos diseños.

En los próximos apartados se presentarán ejemplos concretos donde el modelo no determinista reduce drásticamente el esfuerzo de construcción frente al determinista, y se mostrará el procedimiento sistemático (construcción “del subconjunto”) que convierte cualquier AFN en un AFD equivalente.


## **3.5 Ejemplos**

### **3.5.1 Ejemplo 1**

El AFN $M$ de la figura trabaja sobre el alfabeto $\Sigma=\{a,b\}$.  

<p float="center">
  <img src="https://github.com/carlosalvarezh/TeoriaComputacion/blob/main/images/TC04_AFN02.PNG?raw=true" width="200" />
</p>

Las transiciones permiten que, al leer un símbolo, la máquina **se divida** en varias copias que evolucionan de forma independiente.  
Para la cadena $u=abb$ veremos que:

* algunas copias terminan sin transición posible (**cómputo abortado**);  

* otras consumen todo $u$ y paran en un estado **no** perteneciente a $F$ (**rechazo**);  

* al menos una copia consume $u$ y queda en un estado de $F$ (**aceptación**).

Como basta **una** trayectoria exitosa para aceptar, $u\in L(M)$.

**Función de transición**

$$
\begin{array}{|c|c|c|}
\hline
\Delta & a & b\\\hline
q_0 & \{q_0,q_1,q_3\} & \varnothing\\
q_1 & \{q_1\} & \{q_2\}\\
q_2 & \varnothing & \{q_1,q_2\}\\
q_3 & \varnothing & \{q_3\}\\\hline
\end{array}
$$

Estados de aceptación (doble circunferencia en la figura):
$F=\{q_0,q_1,q_3\}$.

#### **Cómputos para $u=abb$**

> Cada diagrama-cinta (figs. siguientes) muestra, bajo la posición que se está leyendo, las copias activas y el estado de cada una.

##### **1. Cómputo de rechazo**  

<p float="center">
  <img src="https://github.com/carlosalvarezh/TeoriaComputacion/blob/main/images/TC04_AFN03.PNG?raw=true" width="250" />
</p>

1. Empieza en $q_0$ y lee **a** → copia única en $q_1$.
2. Lee el primer **b** → copia se mueve a $q_2$.
3. Lee el segundo **b** → copia permanece en $q_2$.
4. La cadena termina y la copia está en $q_2\notin F$ ⇒ **rechazo**.

##### **2. Primer cómputo de aceptación**  

<p float="center">
  <img src="https://github.com/carlosalvarezh/TeoriaComputacion/blob/main/images/TC04_AFN04.PNG?raw=true" width="250" />
</p>

1. $q_0 \xrightarrow{a} q_3$.
2. Primer **b**: $q_3\xrightarrow{b}q_3$.
3. Segundo **b**: $q_3\xrightarrow{b}q_3$.
4. Fin de la cadena en $q_3\in F$ ⇒ **aceptación**.

##### **3. Segundo cómputo de aceptación**  

<p float="center">
  <img src="https://github.com/carlosalvarezh/TeoriaComputacion/blob/main/images/TC04_AFN05.PNG?raw=true" width="250" />
</p>

1. $q_0 \xrightarrow{a} q_1$.
2. Primer **b**: $q_1\xrightarrow{b}q_2$.
3. Segundo **b**: $q_2\xrightarrow{b}q_1$.
4. Termina en $q_1\in F$ ⇒ **aceptación**.

##### **4. Cómputo abortado**  

<p float="center">
  <img src="https://github.com/carlosalvarezh/TeoriaComputacion/blob/main/images/TC04_AFN06.PNG?raw=true" width="250" />
</p>

1. $q_0\xrightarrow{a} q_0$ (usando la rama que se queda en $q_0$).
2. Primer **b**: no hay transición definida desde $q_0$ con $b$ ($\Delta(q_0,b)=\varnothing$).
3. Esa copia se detiene prematuramente ⇒ **abortada** (no alcanza a leer todo $u$).


**Conclusión**

En la ejecución paralela del AFN sobre $u=abb$:

* existe al menos un camino que consume la cadena completa y acaba en $q_1$ o $q_3$ (estados de aceptación),  

* por lo tanto $u$ pertenece al lenguaje reconocido por $M$, aun cuando coexistan trayectorias abortadas o que terminen en rechazo.


### **3.5.2 Ejemplo 2**

#### **Lenguaje objetivo**  

En un ejemplo de la sección 3.3 se diseñó el siguiente AFD que acepta el lenguaje de las cadenas sobre $\Sigma=\{a, b\}$ que terminan en $b$:

$$
L=\{\,w\in\{a,b\}^{*}\mid\text{el símbolo final de }w\text{ es }b\,\}
$$

Equivalente a la expresión regular $(a\cup b)^{*}b$.

- #### **AFD**

<p float="center">
  <img src="https://github.com/carlosalvarezh/TeoriaComputacion/blob/main/images/TC03_Graph13.PNG?raw=true" width="250" />
</p>

| Estado | Significado                                | ¿Acepta? |
| ------ | ------------------------------------------ | -------- |
| $q_0$  | “la última letra vista es $a$ (o ninguna)” | No       |
| $q_1$  | “la última letra vista es $b$”             | Sí       |

* Transiciones deterministas:

  * Desde $q_0$   —$a$→ $q_0$ (otro $a$ mantiene el estado “no-aceptar”).
  * Desde $q_0$   —$b$→ $q_1$ (el último símbolo pasa a ser $b$).
  * Desde $q_1$   —$b$→ $q_1$ (un nuevo $b$ sigue siendo aceptable).
  * Desde $q_1$   —$a$→ $q_0$ (un $a$ al final anula la condición de aceptación).

El autómata “recuerda” de forma explícita el carácter leído más recientemente, por eso necesita **dos** estados.

- ### **AFN (misma tarea, diseño más directo)**

<p float="center">
  <img src="https://github.com/carlosalvarezh/TeoriaComputacion/blob/main/images/TC04_AFN07.PNG?raw=true" width="250" />
</p>

* Un único control principal $q_0$ con bucles en **a** y **b**: la lectura puede continuar indefinidamente sin perderse.
* Una transición extra **b**→ $q_1$ (estado de aceptación).
* Al leer un $b$ en $q_0$ existen **dos** posibilidades simultáneas:

  1. quedarse en $q_0$ y seguir procesando,
  2. saltar a $q_1$ y “decidir” que ese $b$ será el último símbolo.

Ese desdoblamiento convierte al modelo en no determinista (un mismo par estado-símbolo tiene dos salidas).

#### **¿Por qué el AFN es “más fácil de concebir”?**

1. **No necesita memorizar el último carácter.**
   En el AFD hubo que crear un estado especial para “acabé en $b$”.
   El AFN se limita a **“apostar”**: cada vez que ve un $b$, abre una rama que declara “éste será el final”.
   Si luego llega otro símbolo, esa rama fallará, pero puede haber otra apuesta más adelante.

2. **Refleja la expresión regular casi de forma literal.**

   * Los bucles sobre $q_0$ cubren $(a\cup b)^{*}$;
   * la flecha $b$ hacia $q_1$ materializa el $b$ final.
     El diseño se obtiene en un solo trazo mental, sin analizar casos de memoria.

3. **Menos estados y transiciones conceptuales.**
   Solo se definen dos estados; todas las ramas que aborten después del salto a $q_1$ son irrelevantes mientras al menos una termine justo allí.
   El diseñador se concentra en “¿cuándo puedo aceptar?” en lugar de “¿qué condición debo recordar?”.

#### **Evaluación rápida de una cadena**

Cadena de ejemplo `a b a b`:

| Símbolo leído | Ramas vivas   | Observaciones                                                                     |
| ------------- | ------------- | --------------------------------------------------------------------------------- |
| `a`           | $q_0$         | solo bucle en $q_0$.                                                              |
| `b`           | $q_0$ y $q_1$ | rama que salta a $q_1$ supone que este puede ser el final.                        |
| `a`           | $q_0$         | la rama en $q_1$ falla (no hay salida con $a$); sobrevive la que seguía en $q_0$. |
| `b`           | $q_0, q_1$    | se renueva la apuesta; al terminar, hay una copia en $q_1$ ⇒ la cadena se acepta. |

El AFD llega a $q_1$ tras el último símbolo, pero tuvo que alternar entre $q_0$ y $q_1$ para “contar” la letra final; el AFN lo resolvió delegando la suposición al momento oportuno.

En síntesis, el AFN de la figura 2 es “más fácil de concebir” porque traduce de forma inmediata la idea **“en algún punto leo el último $b$”** a una bifurcación no determinista, sin obligar a crear una memoria explícita del carácter final.


#### **Ejemplo cotidiano — detector de “mensaje de cierre” en un chatbot**

*Supuestos*

* Cada aportación del chat se clasifica automáticamente como

  * **`a`** : mensaje normal (saludo, pregunta, comentario, etc.),
  * **`b`** : mensaje de cierre (“Adiós”, “Chao”, “Bye”).
* El bot de soporte debe marcar la conversación como **finalizada** si el **último** mensaje que llega es de cierre.

##### **AFD**

Para saber si la última entrada fue `b` el bot mantiene una memoria explícita:

* **Estado `q0`** – “el mensaje más reciente no es de cierre”
* **Estado `q1`** – “el mensaje más reciente sí es de cierre” (estado de aceptación)

Cada vez que entra un mensaje:

* `a` deja (o devuelve) al bot en `q0`.
* `b` mueve al bot a `q1`.
* Si estando en `q1` llega un nuevo `a`, la máquina vuelve a `q0` y la sesión deja de considerarse finalizada.

Este control funciona, pero obliga a alternar continuamente entre dos estados para reflejar cada nuevo mensaje.

##### **AFN**

Se conserva `q0` (inicio) y se añade `q1` (aceptación), pero la lógica cambia:

* En `q0` hay **sendas transiciones en bucle** con `a` y con `b`: el bot siempre puede seguir leyendo.
* Además, al leer un `b` existe **una segunda salida** que va a `q1`.

El comportamiento resulta natural:

1. Cuando aparece un mensaje `b`, el bot **abre dos hipótesis simultáneas**:

   * una copia se queda en `q0` (por si el diálogo continúa),
   * la otra copia salta a `q1` y apuesta a que ese `b` será el último.
2. Si llegan más mensajes, las copias que estaban en `q1` se descartan —no hay salidas—, pero basta con que **el último** vuelva a ser `b` para que una nueva copia alcance `q1` y la conversación quede marcada como finalizada.

##### **Por qué este AFN describe el problema con menos esfuerzo**

* No lleva un registro explícito del “último tipo de mensaje”; simplemente **intenta** declarar cada `b` como cierre y deja que los hechos confirmen o invaliden esa apuesta.
* El diagrama refleja directamente la expresión regular $(a\cup b)^{*}b$:

  * bucles en `q0` para “cualquier secuencia de mensajes”,
  * una flecha `b` hacia `q1` para el cierre definitivo.
* Si el diseñador añade otro posible token de cierre (por ejemplo `c = “Hasta luego”`), basta con duplicar la flecha `c` de `q0` a `q1`; no hay que replantear la lógica interna.

En la práctica, muchos filtros de texto y motores de expresiones regulares trabajan exactamente con esta estrategia: cada vez que aparece un símbolo candidato, crean una rama que da por concluido el patrón y siguen leyendo por si ese intento resulta inválido.


### **3.5.3 Ejemplo 3**

Sea $\Sigma=\{a,b\}$.  Queremos un autómata que acepte **todas las cadenas que comienzan con al menos una $a$** y, después de esa primera $a$, permiten **exactamente una de dos alternativas**:

1. seguir leyendo únicamente $a$’s (uno o más)
2. leer únicamente $b$’s (cero o más)

Formalmente

$$
L = ab^{*}\;\cup\;a^{+}.
$$

<p float="center">
  <img src="https://github.com/carlosalvarezh/TeoriaComputacion/blob/main/images/TC04_AFN08.PNG?raw=true" width="250" />
</p>

El AFN de la figura cumple $L(M)=L$ porque, al leer la primera $a$ en $q_{0}$, se bifurca:

* una copia salta a $q_{1}$ y se queda allí mientras aparezcan $b$’s;
* otra copia salta a $q_{2}$ y se mantiene mientras sigan llegando $a$’s.

Ambos estados son de aceptación; cualquier otra combinación de símbolos aborta o termina en rechazo.

#### **Aplicación práctica de este AFN - Control de acceso a un edificio con dos procedimientos distintos**

* **$a$**  = registrar la tarjeta de empleado en el lector.
* **$b$**  = abrir la puerta batiente (sensor detecta el paso).

Reglas del sistema:

1. Todo ingreso empieza obligatoriamente registrando la tarjeta **al menos una vez**.
2. Después de la primera lectura hay dos protocolos válidos:

   * **Protocolo A** – se requieren *varias verificaciones de tarjeta* antes de abrir (caso de visitantes internos que vuelven a mostrar la tarjeta a vigilantes sucesivos).
     Esto genera la secuencia $a^{+}$ y corresponde al camino $q_{0}\to q_{2}$ con bucle en $a$.

   * **Protocolo B** – basta con la primera lectura; a continuación el usuario cruza hasta que el sensor deje de detectar pasos.
     Produce la secuencia $ab^{*}$ y coincide con el camino $q_{0}\to q_{1}$ con bucle en $b$.

* Cualquier intento de alternar lecturas y pasos (por ejemplo `aba`, `abbab`) no está previsto, de modo que la copia que lo intente se aborta y no influye en la decisión final.

Con este paralelismo el diagrama del AFN resulta fácil de concebir:
una flecha para la primera tarjeta y, desde allí, dos ciclos independientes que modelan los dos protocolos admitidos.


### **3.5.4 Ejemplo 4**

### **Lenguaje objetivo**

$$
L=(01\;\cup\;010)^{*}
$$

Cada palabra en $L$ es una concatenación, en cualquier cantidad, de los **bloques**
*`01`*  o *`010`*.
Equivalentemente, la entrada se puede leer como una secuencia de “uno o más ceros seguidos de un uno”, con la restricción de que no aparezcan dos unos consecutivos.

#### **Primera solución – cuatro estados**

<p float="center">
  <img src="https://github.com/carlosalvarezh/TeoriaComputacion/blob/main/images/TC04_AFN09.PNG?raw=true" width="250" />
</p>

* **$q_0$** inicial y de aceptación.

  * Lee `0` y se bifurca:

    * a $q_1$ para intentar el bloque `01`;
    * a $q_2$ para intentar el bloque `010`.
* **$q_1$** espera `1` para completar `01`; al recibirlo regresa a $q_0$.
* **$q_2$** espera otro `0` y avanza a $q_3$.
* **$q_3$** espera `1` para cerrar `010` y volver a $q_0$.

El diseño sigue literalmente la descomposición en dos bloques distintos, pero mantiene **rutas paralelas** (`01` y `010`) con sus propios pares de estados intermedios.

#### **Segunda solución – tres estados**

<p float="center">
  <img src="https://github.com/carlosalvarezh/TeoriaComputacion/blob/main/images/TC04_AFN10.PNG?raw=true" width="250" />
</p>


| Estado | Papel dentro del ciclo de bloques        | Transiciones                                                          |
| ------ | ---------------------------------------- | --------------------------------------------------------------------- |
| $q_0$  | comienzo de un nuevo bloque (aceptación) | `0` → $q_1$                                                           |
| $q_1$  | se ha leído un `0`                       | `1` → $q_0$ (completa `01`) <br> `1` no permitido (dos unos seguidos) |
| $q_1$  |                                          | `0` → $q_2$ (posible `010`)                                           |
| $q_2$  | se ha leído `00`; espera `1`             | `1` → $q_0$ (completa `010`)                                          |

**Cómo cubre ambos bloques**

1. Si tras el primer `0` aparece `1`, la trayectoria $q_0 \xrightarrow{0} q_1 \xrightarrow{1} q_0$ reconoce `01`.
2. Si aparece otro `0`, se toma $q_0 \xrightarrow{0} q_1 \xrightarrow{0} q_2$; el `1` siguiente lleva de $q_2$ a $q_0$, completando `010`.

En ambos casos el recorrido vuelve a $q_0$, listo para procesar otro bloque.
Cualquier símbolo que rompa el patrón (p.e. dos unos seguidos) no tiene transición definida y aborta.

#### **Por qué la segunda versión es más eficiente**

1. **Menos estados**
   Reducir de cuatro a tres estados simplifica la tabla de transición y el coste de la conversión posterior a AFD (si se requiere).

2. **Compartición de prefijos**
   El prefijo `0` es común a `01` y `010`. El diseño con tres estados reutiliza $q_1$ para ambos casos, mientras que la versión de cuatro estados crea dos ramas paralelas. Compartir camino evita duplicar lógica.

3. **Menos bifurcaciones no deterministas**

   * En la figura 1, desde $q_0$ el símbolo `0` genera dos copias simultáneas.
   * En la figura 2, `0` crea **una sola** copia que decide después—con la segunda lectura—si la ruta es `01` o `010`.
     Menos copias implican menos ramas activas y una simulación más ligera.

4. **Claridad en el ciclo**
   Todo bloque termina siempre en $q_0$; el diagrama revela de inmediato la estructura “leer bloque, volver a inicio”, haciendo más fácil razonar sobre cadenas repetidas.

En síntesis, la figura 2 acepta exactamente el mismo lenguaje porque modela los mismos dos bloques, pero los compacta en un único camino con decisión diferida, reduciendo estados y transiciones redundantes.

#### **Aplicación práctica que sigue el patrón $(01 \cup 010)^{*}$ - Máquina expendedora de café con dos modos de servicio**

| Símbolo | Acción detectada por la máquina    |
| ------- | ---------------------------------- |
| **0**   | Se introduce una moneda de \$1 000 |
| **1**   | Se pulsa el botón **SERVIR**       |

La expendedora permite dos tamaños de bebida:

1. **Tamaño regular** – se paga una sola moneda y luego se pulsa **SERVIR**
   → secuencia **01**

2. **Tamaño grande** – se pagan **dos** monedas consecutivas y luego se pulsa **SERVIR**
   → secuencia **010**

El cliente puede comprar cualquier cantidad de bebidas en una misma sesión; cada vez que termina un servicio, la máquina vuelve a esperar la siguiente moneda.
Por diseño **no se acepta**:

* Pulsar **SERVIR** sin haber insertado la(s) moneda(s) requeridas.
* Insertar más de dos monedas seguidas antes de servir (implicaría dos ceros adicionales).
* Pulsar dos veces **SERVIR** sin pago intermedio (implicaría dos unos consecutivos).

#### **Correspondencia con el AFN de tres estados (figura 2)**

* **$q_0$** – “lista para recibir la primera moneda” (estado de aceptación porque puede cerrar inmediatamente la sesión sin comprar).
* **$q_1$** – “se ha recibido la primera moneda”.
* **$q_2$** – “se han recibido dos monedas, falta pulsar **SERVIR**”.

| Trayectoria en el AFN                                             | Operación del usuario  | Tamaño servido |
| ----------------------------------------------------------------- | ---------------------- | -------------- |
| $q_0 \xrightarrow{0} q_1 \xrightarrow{1} q_0$                     | 1 moneda + **SERVIR**  | Regular        |
| $q_0 \xrightarrow{0} q_1 \xrightarrow{0} q_2 \xrightarrow{1} q_0$ | 2 monedas + **SERVIR** | Grande         |

Después de servir, el autómata (y la máquina) regresan a $q_0$, listos para repetir el proceso, de modo que una sesión completa es una concatenación arbitraria de **01** y **010**, exactamente el lenguaje $(01 \cup 010)^{*}$.

## **3.7 Equivalencia computacional entre AFD y AFN**

### **3.7.1 Introducción**

La teoría de lenguajes formales sostiene que **todo autómata finito no determinista se puede transformar en un autómata finito determinista que reconozca exactamente el mismo conjunto de cadenas**. Esto es relevante en la práctica por dos razones:

* El modelo no determinista facilita el **diseño**: basta con añadir transiciones alternativas donde convenga, sin preocuparse por la unicidad de la salida.  

* El modelo determinista facilita la **ejecución eficiente**: en cualquier instante existe un único estado activo, lo que permite implementaciones directas en hardware o en software con complejidad lineal en la longitud de la entrada.

#### **De AFD a AFN ― conversión inmediata**

Si disponemos de un AFD $M=(\Sigma,Q,q_{0},F,\delta)$ basta definir

$$
\Delta(q,a)=\{\delta(q,a)\}\qquad\forall q\in Q,\;a\in\Sigma
$$

para obtener un AFN $M'=(\Sigma,Q,q_{0},F,\Delta)$.
Cada entrada de la tabla de transición pasa a ser un **conjunto unitario**; el comportamiento no se altera.

#### **De AFN a AFD ― construcción por subconjuntos**

> ***Teorema 3.7*** Para todo AFN $M=(\Sigma,Q,q_{0},F,\Delta)$ existe un AFD $M'=(\Sigma,Q',\{q_{0}\},F',\delta')$ con $L(M)=L(M')$.

**Idea central**
Un AFN puede mantener varias copias activas; basta con que **una** copie llegue a aceptar la cadena.  

El truco consiste en dejar de rastrear cada copia por separado y, en su lugar, **tratar el conjunto de copias activas como un único estado** del nuevo autómata determinista.

1. **Estados del AFD**

   $$
   Q' \subseteq \wp(Q)
   $$

   Cada elemento de $Q'$ es un subconjunto de $Q$; representa *todas las posiciones posibles* en que el AFN podría hallarse tras leer cierto prefijo.

2. **Estado inicial**
   $\{q_{0}\}$ —al inicio solo la copia original está activa.

3. **Función de transición**
   Para un conjunto $S\subseteq Q$ y un símbolo $a\in\Sigma$:

   $$
   \delta'(S,a)=\bigcup_{q\in S}\Delta(q,a)
   $$

   Es decir, se combina en un solo paso lo que el AFN alcanzaría en paralelo.

4. **Estados de aceptación**

   $$
   F'=\{\,S\in Q' \mid S\cap F\neq\varnothing\,\}
   $$

   Si alguna copia del AFN llega a un estado aceptable, el subconjunto lo refleja.

5. **Cierre de la tabla**
   Se genera la tabla de $\delta'$ de forma iterativa:

   * se parte de $\{q_{0}\}$;
   * se emplea cada símbolo de $\Sigma$ para producir nuevos subconjuntos;
   * el proceso continúa hasta que no aparezcan combinaciones inéditas.

El número máximo de estados en $Q'$ es $2^{|Q|}$. Aunque puede crecer exponencialmente en teoría, en la práctica muchos subconjuntos no se alcanzan y los compiladores modernos aplican técnicas de poda y minimización adicionales.

#### **Importancia del resultado**

* **Propiedad de cierre**: los lenguajes regulares no cambian de categoría al pasar entre modelos; esto justifica el uso indistinto de AFN, AFD o expresiones regulares en especificaciones formales.
* **Implementación de analizadores léxicos**: las herramientas (por ejemplo, `lex` o `flex`) construyen primero un AFN a partir de expresiones regulares del usuario y luego aplican este procedimiento para obtener un AFD optimizado que se ejecuta a gran velocidad.
* **Diseño guiado**: el ingeniero puede comenzar con el modelo más cómodo (normalmente AFN) sabiendo que siempre existe una versión determinista lista para producción.

En los apartados siguientes se presenta la construcción paso a paso sobre un ejemplo concreto y se muestra cómo se minimiza el AFD resultante para obtener la máquina más compacta posible.

* **$ab^{*}$** : la cadena empieza con un solo `a` y puede ir seguida de **cualquier cantidad de `b` (incluido cero)**.  

* **$a^{+}$** : la cadena contiene **uno o más `a` consecutivos** y nada más.

En ambos casos el primer símbolo es siempre `a`; el resto determina qué variante se ha elegido.

#### **1. AFN original $M$ (figura 1)**

| Estado $\Delta$               | entrada `a`   | entrada `b` | Comentario breve                                                    |
| -------------------- | :-------------: | :-----------: | ------------------------------------------------------------------- |
| $q_0$ *(inicial)*    | $\{q_1,q_2\}$ | ∅           | tras leer el primer `a` se crean dos ramas: “solo `b`” o “solo `a`” |
| $q_1$ *(aceptación)* | ∅             | $\{q_1\}$   | camino $ab^{*}$: cada `b` mantiene la aceptación                    |
| $q_2$ *(aceptación)* | $\{q_2\}$     | ∅           | camino $a^{+}$: cada `a` extiende la racha de `a`                   |

En palabras, de la tabla se indica:

1. En $q_0$ solo se admite `a`. Ese símbolo **clona** el cómputo:

   * Rama 1 → $q_1$ (luego solo `b`).
   * Rama 2 → $q_2$ (luego solo `a`).  

2. Una rama que reciba un símbolo incompatible se queda sin transición y se aborta.  

3. Una cadena se acepta si **alguna rama** alcanza $q_1$ o $q_2$ habiendo consumido toda la entrada.

#### **2. Determinización: AFD $M'$**

Para cada prefijo de la entrada reunimos **todas** las ramas activas del AFN en un subconjunto $S\subseteq\{q_0,q_1,q_2\}$. Ese subconjunto se trata como un solo estado del AFD.

| Estado (subconjunto) | `a` →         | `b` →     | Significado                                               |
| -------------------- | ------------- | --------- | --------------------------------------------------------- |
| $\{q_0\}$            | $\{q_1,q_2\}$ | ∅         | inicio; aún no se decidió la variante                     |
| $\{q_1\}$            | ∅             | $\{q_1\}$ | solo queda la variante de `b` repetidos                   |
| $\{q_2\}$            | $\{q_2\}$     | ∅         | solo queda la variante de más `a`                         |
| $\{q_1,q_2\}$        | $\{q_2\}$     | $\{q_1\}$ | ambas ramas vivas; el símbolo actual elige cuál sobrevive |

#### **3. Importancia del proceso**

* **Verifica la equivalencia formal** entre modelos: cualquier lenguaje descrito de forma intuitiva mediante un AFN puede obtener una versión determinista, lista para implementarse sin exploración simultánea de ramas.

#### **4· Ejemplo práctico — Filtro de correo sospechoso**

| Símbolo | Acción del usuario                                      |
| ------- | ------------------------------------------------------- |
| `a`     | Escribe la palabra **“URGENTE”** al inicio del asunto   |
| `b`     | Añade un signo de exclamación **!** al final del asunto |

**Reglas del filtro**

1. Solo se estudian mensajes cuyo asunto comienza con **URGENTE** (`a`).
2. Después de esa palabra, el remitente **debe** escoger **una** de dos rutas:

   * ruta **“solo exclamaciones”**: puede colocar cualquier cantidad de signos de exclamación (`b`, `bb`, `bbb`, …);
   * ruta **“solo más URGENTE”**: puede repetir la palabra **URGENTE** una o más veces (`a`, `aa`, …);
     Mezclar ambas cosas (por ejemplo `ab`, `aba`) invalida el formato.

**Correspondencia con los autómatas**

* $q_0$ — filtro esperando la primera ocurrencia de **URGENTE**.
* `a` divide la evaluación:

  * copia “solo signos” → $q_1$ (ciclo en `b`), marca el asunto como sospechoso si se agota la cadena.
  * copia “solo URGENTE extra” → $q_2$ (ciclo en `a`), también considera sospechoso al terminar.
* El AFD resultante implementa el mismo filtrado sin ramificar procesos: tras la primera palabra mantiene internamente el subconjunto $\{q_1,q_2\}$ y deja que el símbolo siguiente decida cuál ruta sigue viva.

Así, la teoría de determinización se traduce en un motor de reglas único, rápido y fácil de mantener, que detecta automáticamente asuntos que caen en el patrón “URGENTE + grito” o “URGENTE repetido”.


## **3.7.3 Definiciones**

### **3.7.3.1 Extensión de la función de transición en un AFD**

Sea $M=\left(\Sigma, Q, q_0, F, \delta\right)$ un AFD. La función de transición $\delta: Q \times \Sigma \longrightarrow Q$ se extiende a una función $\widehat{\delta}: Q \times \Sigma^* \longrightarrow Q$ por medio de la siguiente definición recursiva:

$$
\left\{\begin{array}{l}
\widehat{\delta}(q, \lambda)=q, \quad q \in Q \\
\widehat{\delta}(q, a)=\delta(q, a), \quad q \in Q, \quad a \in \Sigma, \\
\widehat{\delta}(q, w a)=\delta(\widehat{\delta}(q, w), a), \quad q \in Q, a \in \Sigma, w \in \Sigma^*
\end{array}\right.
$$

La tabla original $\delta:Q\times\Sigma\to Q$ indica ―símbolo por símbolo― a qué estado se pasa; sin embargo, para caracterizar el lenguaje aceptado necesitamos saber en qué estado termina el autómata tras leer **toda** una cadena.  

Por eso se define la **función extendida**

$$
\widehat{\delta}:Q\times\Sigma^{*}\longrightarrow Q,
$$

mediante la recursión:

$$
\boxed{
\begin{aligned}
\widehat{\delta}(q,\lambda) &= q &&\text{(la cadena vacía no provoca ningún cambio)}\\[2pt]
\widehat{\delta}(q,a) &= \delta(q,a) &&\text{(caso base de una sola letra)}\\[2pt]
\widehat{\delta}(q,wa) &= \delta\bigl(\,\widehat{\delta}(q,w),\,a\bigr) &&\text{(procesar \(w\) y luego el último \(a\))}
\end{aligned}}
$$

* **Primera línea**: leer $\lambda$ implica “quedarse donde se está”.  

* **Segunda línea**: para una única letra se usa la tabla original.  

* **Tercera línea**: para una palabra $wa$ se procesa primero el prefijo $w$; al estado resultante se aplica la transición con la letra final $a$.

Con esta extensión:

$$
L(M)=\bigl\{\,w\in\Sigma^{*}\;\big|\;\widehat{\delta}(q_{0},w)\in F\,\bigr\}.
$$

> **Lectura**: una cadena se acepta si el estado que se alcanza tras procesarla completa pertenece a $F$.

Por comodidad, a partir de aquí se omite el acento y se escribe $\delta(q,w)$ en lugar de $\widehat{\delta}(q,w)$.


### **3.7.3.2 Extensión de la función de transición en un AFN**

Sea $M=\left(\Sigma, Q, q_0, F, \Delta\right)$ un AFN. La función de transición $\Delta: Q \times \Sigma \longrightarrow \wp(Q)$ se extiende inicialmente a conjuntos de estados. Para $a \in \Sigma$ y $S \subseteq F$ se define

$$
\Delta(S, a):=\bigcup_{q \in S} \Delta(q, a)
$$

Luego se extiende $\Delta$ a una función $\widehat{\Delta}: Q \times \Sigma^* \longrightarrow \wp(Q)$, de manera similar a como se hace para los AFD. Recursivamente,

$$
\left\{\begin{array}{l}
\widehat{\Delta}(q, \lambda)=\{q\}, \quad q \in Q \\
\widehat{\Delta}(q, a)=\Delta(q, a), \quad q \in Q, a \in \Sigma, \\
\widehat{\Delta}(q, w a)=\Delta(\widehat{\Delta}(q, w), a)=\bigcup_{p \in \widehat{\Delta}(q, w)} \Delta(p, a), \quad q \in Q, a \in \Sigma, w \in \Sigma^*
\end{array}\right.
$$


#### (a) Paso intermedio: de estados individuales a **conjuntos**

Para $a\in\Sigma$ y cualquier subconjunto $S\subseteq Q$:

$$
\boxed{\;\Delta(S,a)\;=\;\bigcup_{q\in S}\Delta(q,a)\;}
$$

Es decir, al leer $a$ desde un conjunto $S$, el autómata avanza cada copia que esté en $q$ según la tabla y reúne los resultados.

#### (b) Extensión completa a cadenas

Se define

$$
\widehat{\Delta}:Q\times\Sigma^{*}\longrightarrow\wp(Q)
$$

(la potencia $\wp(Q)$ es el conjunto de todos los subconjuntos de $Q$) con la recursión:

$$
\boxed{
\begin{aligned}
\widehat{\Delta}(q,\lambda) &= \{q\} &&\text{(un solo estado activo y ningún movimiento)}\\[2pt]
\widehat{\Delta}(q,a) &= \Delta(q,a) &&\text{(un símbolo, tabla original)}\\[2pt]
\widehat{\Delta}(q,wa) &= \Delta\bigl(\widehat{\Delta}(q,w),a\bigr)\\
                       &= \displaystyle\bigcup_{p\in\widehat{\Delta}(q,w)} \Delta(p,a) 
                       &&\text{(avanzar **todas** las copias generadas por \(w\))}
\end{aligned}}
$$

* La primera línea produce un conjunto de un solo elemento: la copia que no se ha movido.
* La segunda usa directamente la tabla $\Delta$.
* La tercera procesa el prefijo $w$, obtiene el conjunto de copias vivas y, para cada una, aplica la transición con $a$.

Con esta extensión:

$$
L(M)=\bigl\{\,w\in\Sigma^{*}\;\big|\;\widehat{\Delta}(q_{0},w)\cap F\neq\varnothing\,\bigr\}.
$$

> **Lectura**: la cadena se acepta si, después de leerla completa, **al menos una** de las copias termina en un estado de aceptación.

Como antes, se permite escribir $\Delta(q,w)$ en lugar de $\widehat{\Delta}(q,w)$ cuando el contexto lo deja claro.

Estas extensiones son la herramienta formal que:

* conecta la tabla de un autómata con la definición de su lenguaje;  

* posibilita la **construcción por subconjuntos** (AFN → AFD) demostrando que el no determinismo no añade poder de reconocimiento;  

* fundamenta implementaciones prácticas (analizadores léxicos, buscadores de patrones) que primero generan un AFN y luego lo convierten en un AFD eficiente.
