# **CapГ­tulo 3: AutГіmatas Finitos No Deterministas (AFN)**

Los **autГіmatas finitos deterministas** (AFD) exigen que, para cada par (estado, sГ­mbolo de entrada), exista exactamente una transiciГіn definida. Este requisito facilita la implementaciГіn, pero puede complicar el diseГ±o: en muchas situaciones es mГЎs natural describir el comportamiento de una mГЎquina admitiendo que, ante un mismo sГ­mbolo, se activen varias transiciones simultГЎneas o ninguna.

En este capГ­tulo presentamos los **autГіmatas finitos no deterministas** (AFN), que relajan esa restricciГіn. El resultado teГіrico central del capГ­tulo demuestra que, pese a su mayor flexibilidad aparente, los AFN **no reconocen lenguajes adicionales**: todo AFN puede convertirse en un AFD equivalente mediante la **construcciГіn por subconjuntos**. La clase de lenguajes reconocibles sigue siendo exactamente la de los **lenguajes regulares**.

El capГ­tulo estГЎ organizado de la siguiente manera:
1. DefiniciГіn formal del AFN y su criterio de aceptaciГіn.
2. Ejemplos comparativos AFD vs. AFN para ilustrar las ventajas de diseГ±o del modelo no determinista.
3. La **construcciГіn por subconjuntos** y las extensiones formales de la funciГіn de transiciГіn.
4. Los **AFN-О»** (con transiciones espontГЎneas) y su equivalencia con los AFN ordinarios.

## **3.1 AutГіmatas finitos no deterministas (AFN)**

Un autГіmata finito no determinista extiende la idea del AFD permitiendo que, ante un mismo sГ­mbolo de entrada, puedan activarse **varias** transiciones desde el mismo estado, o incluso ninguna.  

La definiciГіn formal introduce un ligero cambio: la funciГіn de transiciГіn ya no devuelve **un** estado, sino un **conjunto** de estados posibles.

## **3.2 DefiniciГіn formal**

$$
M=\bigl(\Sigma,\,Q,\,q_{0},\,F,\,\Delta\bigr)
$$

| Componente                        | DescripciГіn                                                                     |
| --------------------------------- | ------------------------------------------------------------------------------- |
| $\Sigma$                          | alfabeto de entrada                                                             |
| $Q$                               | conjunto finito de estados                                                      |
| $q_{0}$                           | estado inicial                                                                  |
| $F\subseteq Q,\,F\neq\varnothing$ | estados de aceptaciГіn                                                           |
| $\Delta:Q\times\Sigma\to\wp(Q)$   | funciГіn de transiciГіn; $\wp(Q)$ es el conjunto de todos los subconjuntos de $Q$ |

Para $q\in Q$ y $s\in\Sigma$:

$$
\Delta(q,s)=\{\,q_{i_{1}},q_{i_{2}},\dots,q_{i_{k}}\,\}
$$

**InterpretaciГіn:** estando en $q$ y leyendo $s$, el autГіmata **puede tomar** cualquiera de las transiciones listadas y, tras dar el paso, avanza el cabezal una celda a la derecha.

El hecho de que el codominio sea $\wp(Q)$ вЂ” el conjunto potencia вЂ” significa que $\Delta(q,s)$ puede ser un subconjunto de cualquier tamaГ±o. Esto da lugar a **tres casos cualitativamente distintos**:

| Caso | Ejemplo | Significado |
| ---- | ------- | ----------- |
| Conjunto con varios elementos | $\Delta(q,s)=\{q_1, q_2, q_3\}$ | No determinismo puro: el autГіmata se ramifica en tantas copias como elementos tenga el conjunto, una por cada destino posible. |
| Conjunto unitario | $\Delta(q,s)=\{q_1\}$ | Hay una Гєnica opciГіn: el comportamiento en ese punto es idГ©ntico al de un AFD. Un AFD es simplemente el caso especial donde **todos** los conjuntos son unitarios. |
| Conjunto vacГ­o | $\Delta(q,s)=\varnothing$ | No existe ninguna transiciГіn definida para esa combinaciГіn. La copia del autГіmata que llega a este punto se detiene sin haber consumido toda la cadena вЂ” se dice que ese cГіmputo se **aborta**. |

> **Nota clave:** que un cГіmputo se aborte no implica que la cadena sea rechazada. Si otras copias del autГіmata logran consumir la cadena completa y terminar en un estado de aceptaciГіn, la cadena se acepta de todas formas. El criterio de aceptaciГіn se estudiarГЎ en la secciГіn 3.4.

## **3.3 CГіmputo no determinista**

***Diagramas.*** En el grafo de transiciones el no determinismo se traduce en **arcos paralelos** que salen del mismo nodo con la misma etiqueta.

<p float="center">
  <img src="https://github.com/carlosalvarezh/TeoriaComputacion/blob/main/images/TC04_AFN01.PNG?raw=true" width="200" />
</p>

Cada arco representa una alternativa que el autГіmata puede tomar al procesar ese sГ­mbolo.

**EjecuciГіn.** Cuando una cadena $u\in\Sigma^{*}$ se coloca en la cinta, el autГіmata genera simultГЎneamente todas las trayectorias compatibles con las reglas de $\Delta$. Se puede pensar en esto como si el autГіmata se **clonara** cada vez que tiene varias opciones: cada copia sigue su propio camino de forma independiente.

* Puede darse el caso de que algunas trayectorias se detengan prematuramente por falta de transiciГіn (cГіmputo abortado).
* Otras trayectorias pueden avanzar hasta consumir la cadena completa.

**Ejemplo mГ­nimo de ejecuciГіn paralela**

Sea un AFN con $\Sigma=\{a,b\}$, $Q=\{q_0,q_1,q_2\}$ y la tabla parcial:

$$\Delta(q_0,a)=\{q_1,q_2\},\quad\Delta(q_1,b)=\{q_1\},\quad\Delta(q_2,b)=\varnothing$$

Procesemos la cadena $u=ab$:

| Paso | SГ­mbolo leГ­do | Copias activas **antes** | Copias activas **despuГ©s** | ObservaciГіn |
| ---- | :-----------: | :----------------------: | :------------------------: | ----------- |
| 0 | вЂ” | $\{q_0\}$ | $\{q_0\}$ | Estado inicial; una sola copia |
| 1 | $a$ | $\{q_0\}$ | $\{q_1,\,q_2\}$ | La copia se **bifurca** en dos |
| 2 | $b$ | $\{q_1,\,q_2\}$ | $\{q_1\}$ | $\Delta(q_1,b)=\{q_1\}$; la copia en $q_2$ se **aborta** porque $\Delta(q_2,b)=\varnothing$ |

Al terminar de leer $u$, la Гєnica copia viva estГЎ en $q_1$. Si $q_1\in F$, la cadena se acepta.

> NГіtese que el conjunto de copias activas evoluciona asГ­: $\{q_0\}\to\{q_1,q_2\}\to\{q_1\}$. Precisamente estos **conjuntos** serГЎn los estados del AFD equivalente cuando se aplique la construcciГіn por subconjuntos (secciГіn 3.7).

## **3.4 Criterio de aceptaciГіn**

El lenguaje reconocido por $M$ es

$$
L(M)=\Bigl\{\,u\in\Sigma^{*}\mid
\exists\text{ al menos un recorrido completo de }u\text{ que termine en }q\in F\Bigr\}.
$$

- *Una cadena se acepta si **alguna** de las posibles ejecuciones lee la cadena entera y finaliza en un estado de aceptaciГіn.*

- No es necesario que **todas** las trayectorias terminen bien; basta con una.

**Tipos de resultado de un cГіmputo individual**

Cada trayectoria (copia del autГіmata) puede terminar de tres maneras distintas. La siguiente tabla resume quГ© significa cada una para la decisiГіn final:

| Tipo de resultado | ВїConsumiГі toda la cadena? | ВїTerminГі en $q\in F$? | ВїContribuye a la aceptaciГіn? |
| ----------------- | :-----------------------: | :-------------------: | :--------------------------: |
| **AceptaciГіn** | SГ­ | SГ­ | **SГ­** вЂ” basta con que exista una sola de estas |
| **Rechazo** | SГ­ | No | No |
| **Aborto** | No ($\Delta(q,s)=\varnothing$ antes de terminar) | N/A | No |

> La cadena $u$ pertenece a $L(M)$ **si y solo si** existe al menos una trayectoria de tipo "AceptaciГіn". Las trayectorias de rechazo y aborto no influyen en el resultado final, siempre que haya al menos una de aceptaciГіn.

**Observaciones clave**

* **No determinismo lГіgico.** El modelo no especifica cГіmo se elige la rama; conceptualmente se exploran todas en paralelo. En la prГЎctica, la construcciГіn por subconjuntos (secciГіn 3.7) permite simular este comportamiento con un solo estado activo.

* **Equivalencia con AFD.** Todo lenguaje aceptado por un AFN es regular; existe siempre un AFD equivalente. La diferencia estГЎ solo en la comodidad para describir el autГіmata.

* **Abortos seguros.** Las trayectorias que se abortan no influyen en el resultado si al menos una rama exitosa llega a $F$.

* **CompactaciГіn de diagramas.** Al permitir arcos duplicados, un AFN suele requerir menos estados y flechas que su versiГіn determinista, lo que simplifica el diseГ±o.

En los prГіximos apartados se presentarГЎn ejemplos concretos donde el modelo no determinista reduce drГЎsticamente el esfuerzo de construcciГіn frente al determinista, y se mostrarГЎ el procedimiento sistemГЎtico (construcciГіn "del subconjunto") que convierte cualquier AFN en un AFD equivalente.

### **3.5.1 Ejemplo 1**

El AFN $M$ de la figura trabaja sobre el alfabeto $\Sigma=\{a,b\}$.  

<p float="center">
  <img src="https://github.com/carlosalvarezh/TeoriaComputacion/blob/main/images/TC04_AFN02.PNG?raw=true" width="300" />
</p>

Las transiciones permiten que, al leer un sГ­mbolo, la mГЎquina **se divida** en varias copias que evolucionan de forma independiente.  
Para la cadena $u=abb$ veremos que:

* algunas copias terminan sin transiciГіn posible (**cГіmputo abortado**);  

* otras consumen todo $u$ y paran en un estado **no** perteneciente a $F$ (**rechazo**);  

* al menos una copia consume $u$ y queda en un estado de $F$ (**aceptaciГіn**).

Como basta **una** trayectoria exitosa para aceptar, $u\in L(M)$.

**FunciГіn de transiciГіn**

$$
\begin{array}{|c|c|c|}
\hline
\Delta & a & b\\\hline
q_0 & \{q_0,q_1,q_3\} & \varnothing\\
q_1 & \{q_1\} & \{q_2\}\\
q_2 & \varnothing & \{q_1,q_2\}\\
q_3 & \varnothing & \{q_3\}\\\hline
\end{array}
$$

Estados de aceptaciГіn (doble circunferencia en la figura):
$F=\{q_0,q_1,q_3\}$.

#### **CГіmputos para $u=abb$**

2. Lee el primer **b** в†’ copia se mueve a $q_2$.
3. Lee el segundo **b** в†’ copia permanece en $q_2$.
4. La cadena termina y la copia estГЎ en $q_2\notin F$ в‡’ **rechazo**.

##### **2. Primer cГіmputo de aceptaciГіn**

* por lo tanto $u$ pertenece al lenguaje reconocido por $M$, aun cuando coexistan trayectorias abortadas o que terminen en rechazo.


### **3.5.2 Ejemplo 2**

#### **Lenguaje objetivo**

$$
L=\{\,w\in\{a,b\}^{*}\mid\text{el sГ­mbolo final de }w\text{ es }b\,\}
$$

Equivalente a la expresiГіn regular $(a\cup b)^{*}b$. Es uno de los lenguajes mГЎs sencillos que sirve para comparar el esfuerzo de diseГ±o entre un AFD y un AFN.

#### **AFD**

<p float="center">
  <img src="https://github.com/carlosalvarezh/TeoriaComputacion/blob/main/images/TC03_Graph13.PNG?raw=true" width="250" />
</p>

| Estado | Significado                                | ВїAcepta? |
| ------ | ------------------------------------------ | -------- |
| $q_0$  | "la Гєltima letra vista es $a$ (o ninguna)" | No       |
| $q_1$  | "la Гєltima letra vista es $b$"             | SГ­       |

Transiciones:
* $q_0 \xrightarrow{a} q_0$, $\quad q_0 \xrightarrow{b} q_1$
* $q_1 \xrightarrow{b} q_1$, $\quad q_1 \xrightarrow{a} q_0$

El autГіmata **recuerda explГ­citamente** el carГЎcter leГ­do mГЎs recientemente; por eso necesita dos estados y debe alternar entre ellos con cada entrada.

#### **AFN (misma tarea, diseГ±o mГЎs directo)**

<p float="center">
  <img src="https://github.com/carlosalvarezh/TeoriaComputacion/blob/main/images/TC04_AFN07.PNG?raw=true" width="250" />
</p>

* Un control principal $q_0$ con bucles en $a$ y en $b$: la lectura puede continuar sin restricciГіn.
* Una transiciГіn adicional $b \to q_1$ (estado de aceptaciГіn).
* Al leer un $b$ en $q_0$ existen **dos** posibilidades simultГЎneas:
  1. quedarse en $q_0$ y seguir procesando,
  2. saltar a $q_1$ y "apostar" a que ese $b$ serГЎ el Гєltimo sГ­mbolo.

Ese desdoblamiento convierte el modelo en no determinista.

#### **ВїPor quГ© el AFN es mГЎs fГЎcil de concebir?**

1. **No necesita memorizar el Гєltimo carГЎcter.** El AFN se limita a *apostar*: cada vez que aparece un $b$, abre una rama que declara "Г©ste serГЎ el final". Si despuГ©s llega otro sГ­mbolo, esa rama falla; pero puede haber otra apuesta mГЎs adelante.

2. **Refleja la expresiГіn regular de forma casi literal.** Los bucles en $q_0$ cubren $(a\cup b)^{*}$; la flecha $b \to q_1$ materializa el $b$ final. El diseГ±o se obtiene en un Гєnico paso mental.

3. **El diseГ±ador se concentra en "ВїcuГЎndo acepto?"** en lugar de "ВїquГ© condiciГіn debo recordar?".

#### **EvaluaciГіn rГЎpida de la cadena $abab$**

| SГ­mbolo leГ­do | Ramas vivas   | Observaciones                                                                     |
| ------------- | ------------- | --------------------------------------------------------------------------------- |
| $a$           | $q_0$         | solo bucle en $q_0$.                                                              |
| $b$           | $q_0,\; q_1$ | la copia que salta a $q_1$ supone que este $b$ puede ser el final.                |
| $a$           | $q_0$         | la rama en $q_1$ falla (no hay salida con $a$); sobrevive la que seguГ­a en $q_0$. |
| $b$           | $q_0,\; q_1$ | se renueva la apuesta; al terminar, hay una copia en $q_1 \in F$ в‡’ **acepta**.    |

El AFD llega igualmente a $q_1$ tras el Гєltimo sГ­mbolo, pero tuvo que alternar entre $q_0$ y $q_1$ para "rastrear" la letra final; el AFN lo resolviГі delegando la decisiГіn al momento oportuno.

#### **Ejemplo cotidiano вЂ” Detector de "mensaje de cierre" en un chatbot**

Un bot de soporte clasifica cada mensaje como $a$ (normal) o $b$ (cierre: "AdiГіs", "Bye", etc.) y debe marcar la conversaciГіn como **finalizada** cuando el **Гєltimo** mensaje recibido sea de tipo $b$.

* El **AFD** mantiene un estado explГ­cito que refleja si el mensaje mГЎs reciente fue $a$ o $b$, y alterna entre $q_0$ y $q_1$ con cada entrada.
* El **AFN** simplemente "apuesta" cada vez que aparece un $b$: abre una rama que declara ese $b$ como cierre definitivo mientras otra copia sigue en $q_0$. Si llega un mensaje posterior, la rama apostada falla; si el diГЎlogo termina ahГ­, esa rama acepta.

La ventaja prГЎctica es evidente: aГ±adir un nuevo token de cierre (p.ej. $c =$ "Hasta luego") requiere solo una flecha adicional $c \to q_1$, sin replantear ninguna lГіgica interna.

### **3.5.3 Ejemplo 3**

Sea $\Sigma=\{a,b\}$.  Queremos un autГіmata que acepte **todas las cadenas que comienzan con al menos una $a$** y, despuГ©s de esa primera $a$, permiten **exactamente una de dos alternativas**:

1. seguir leyendo Гєnicamente $a$вЂ™s (uno o mГЎs)
2. leer Гєnicamente $b$вЂ™s (cero o mГЎs)

Formalmente

$$
L = ab^{*}\;\cup\;a^{+}.
$$

<p float="center">
  <img src="https://github.com/carlosalvarezh/TeoriaComputacion/blob/main/images/TC04_AFN08.PNG?raw=true" width="250" />
</p>

El AFN de la figura cumple $L(M)=L$ porque, al leer la primera $a$ en $q_{0}$, se bifurca:

* una copia salta a $q_{1}$ y se queda allГ­ mientras aparezcan $b$вЂ™s;
* otra copia salta a $q_{2}$ y se mantiene mientras sigan llegando $a$вЂ™s.

Ambos estados son de aceptaciГіn; cualquier otra combinaciГіn de sГ­mbolos aborta o termina en rechazo.

#### **AplicaciГіn prГЎctica de este AFN - Control de acceso a un edificio con dos procedimientos distintos**

* **$a$**  = registrar la tarjeta de empleado en el lector.
* **$b$**  = abrir la puerta batiente (sensor detecta el paso).

Reglas del sistema:

1. Todo ingreso empieza obligatoriamente registrando la tarjeta **al menos una vez**.
2. DespuГ©s de la primera lectura hay dos protocolos vГЎlidos:

   * **Protocolo A** вЂ“ se requieren *varias verificaciones de tarjeta* antes de abrir (caso de visitantes internos que vuelven a mostrar la tarjeta a vigilantes sucesivos).
     Esto genera la secuencia $a^{+}$ y corresponde al camino $q_{0}\to q_{2}$ con bucle en $a$.

   * **Protocolo B** вЂ“ basta con la primera lectura; a continuaciГіn el usuario cruza hasta que el sensor deje de detectar pasos.
     Produce la secuencia $ab^{*}$ y coincide con el camino $q_{0}\to q_{1}$ con bucle en $b$.

* Cualquier intento de alternar lecturas y pasos (por ejemplo `aba`, `abbab`) no estГЎ previsto, de modo que la copia que lo intente se aborta y no influye en la decisiГіn final.

Con este paralelismo el diagrama del AFN resulta fГЎcil de concebir:
una flecha para la primera tarjeta y, desde allГ­, dos ciclos independientes que modelan los dos protocolos admitidos.


### **3.5.4 Ejemplo 4**

#### **Lenguaje objetivo**

$$
L=(01\;\cup\;010)^{*}
$$

Cada palabra en $L$ es una concatenaciГіn, en cualquier cantidad, de los **bloques** `01` o `010`. Equivalentemente, la entrada puede leerse como una secuencia de "un cero seguido de un uno" o "dos ceros seguidos de un uno", sin que aparezcan nunca dos unos consecutivos.

#### **Primera soluciГіn вЂ“ cuatro estados**

<p float="center">
  <img src="https://github.com/carlosalvarezh/TeoriaComputacion/blob/main/images/TC04_AFN09.PNG?raw=true" width="250" />
</p>

* **$q_0$** inicial y de aceptaciГіn.
  * Lee `0` y se bifurca:
    * a $q_1$ para intentar el bloque `01`;
    * a $q_2$ para intentar el bloque `010`.
* **$q_1$** espera `1` para completar `01`; al recibirlo regresa a $q_0$.
* **$q_2$** espera otro `0` y avanza a $q_3$.
* **$q_3$** espera `1` para cerrar `010` y volver a $q_0$.

El diseГ±o sigue literalmente la descomposiciГіn en dos bloques distintos, pero mantiene **rutas paralelas** (`01` y `010`) con sus propios pares de estados intermedios.

#### **Segunda soluciГіn вЂ“ tres estados**

</p>

| Estado | Papel dentro del ciclo de bloques        | Transiciones                                                       |
| ------ | ---------------------------------------- | ------------------------------------------------------------------ |
| $q_0$  | comienzo de un nuevo bloque (aceptaciГіn) | `0` в†’ $q_1$                                                        |
| $q_1$  | se ha leГ­do un `0`                       | `1` в†’ $q_0$ (completa `01`); `0` в†’ $q_2$ (inicia bloque `010`)    |
| $q_2$  | se han leГ­do dos `0`; espera `1`         | `1` в†’ $q_0$ (completa `010`)                                       |

**CГіmo cubre ambos bloques**

1. Si tras el primer `0` aparece `1`: la trayectoria $q_0 \xrightarrow{0} q_1 \xrightarrow{1} q_0$ reconoce `01`.
2. Si aparece otro `0`: se toma $q_0 \xrightarrow{0} q_1 \xrightarrow{0} q_2$; el `1` siguiente lleva de $q_2$ a $q_0$, completando `010`.

En ambos casos el recorrido vuelve a $q_0$, listo para procesar otro bloque. Cualquier sГ­mbolo que rompa el patrГіn (p.ej. dos unos seguidos) no tiene transiciГіn definida y aborta.

#### **Por quГ© la segunda versiГіn es mГЎs eficiente**

1. **Menos estados.** Reducir de cuatro a tres estados simplifica la tabla de transiciГіn y el coste de la conversiГіn posterior a AFD.

2. **ComparticiГіn de prefijos.** El prefijo `0` es comГєn a `01` y `010`. El diseГ±o con tres estados reutiliza $q_1$ para ambos casos, mientras que la versiГіn de cuatro estados crea dos ramas paralelas desde el inicio.

3. **Menos bifurcaciones no deterministas.** En la primera soluciГіn, desde $q_0$ el sГ­mbolo `0` genera dos copias simultГЎneas. En la segunda, `0` crea **una sola** copia que decide despuГ©s вЂ”con la segunda lecturaвЂ” si la ruta es `01` o `010`. Menos copias implican una simulaciГіn mГЎs ligera.

4. **Claridad en el ciclo.** Todo bloque termina siempre en $q_0$; el diagrama revela de inmediato la estructura "leer bloque в†’ volver al inicio", facilitando el razonamiento sobre cadenas repetidas.

En sГ­ntesis, la segunda soluciГіn acepta exactamente el mismo lenguaje pero compacta los dos bloques en un Гєnico camino con decisiГіn diferida, reduciendo estados y transiciones redundantes.

#### **AplicaciГіn prГЎctica вЂ” MГЎquina expendedora de cafГ© con dos modos de servicio**

| SГ­mbolo | AcciГіn detectada por la mГЎquina    |
| ------- | ---------------------------------- |
| **0**   | Se introduce una moneda de \$1 000 |
| **1**   | Se pulsa el botГіn **SERVIR**       |

La expendedora permite dos tamaГ±os de bebida:

1. **TamaГ±o regular** вЂ“ una moneda, luego **SERVIR** в†’ secuencia **01**
2. **TamaГ±o grande** вЂ“ dos monedas consecutivas, luego **SERVIR** в†’ secuencia **010**

El cliente puede comprar cualquier cantidad de bebidas en una misma sesiГіn; cada servicio completo devuelve la mГЎquina al estado de espera. Por diseГ±o **no se acepta**:

* Pulsar **SERVIR** sin haber insertado la(s) moneda(s) requeridas.
* Insertar mГЎs de dos monedas seguidas antes de servir.
* Pulsar dos veces **SERVIR** sin pago intermedio (dos unos consecutivos).

**Correspondencia con el AFN de tres estados**

* **$q_0$** вЂ“ "lista para recibir la primera moneda" (aceptaciГіn: sesiГіn terminada sin compra).
* **$q_1$** вЂ“ "se ha recibido la primera moneda".
* **$q_2$** вЂ“ "se han recibido dos monedas; falta pulsar **SERVIR**".

| Trayectoria en el AFN                                              | OperaciГіn del usuario  | TamaГ±o servido |
| ------------------------------------------------------------------ | ---------------------- | -------------- |
| $q_0 \xrightarrow{0} q_1 \xrightarrow{1} q_0$                      | 1 moneda + **SERVIR**  | Regular        |
| $q_0 \xrightarrow{0} q_1 \xrightarrow{0} q_2 \xrightarrow{1} q_0$ | 2 monedas + **SERVIR** | Grande         |

DespuГ©s de cada servicio, el autГіmata regresa a $q_0$, de modo que una sesiГіn completa es una concatenaciГіn arbitraria de `01` y `010`, exactamente el lenguaje $(01 \cup 010)^{*}$.

## **3.7 Equivalencia computacional entre AFD y AFN**

### **3.7.1 IntroducciГіn**

La teorГ­a de lenguajes formales establece que **todo autГіmata finito no determinista puede transformarse en un autГіmata finito determinista que reconozca exactamente el mismo conjunto de cadenas**. Esto es relevante en la prГЎctica por dos razones:

* El modelo no determinista facilita el **diseГ±o**: basta con aГ±adir transiciones alternativas donde convenga, sin preocuparse por la unicidad de la salida.

* El modelo determinista facilita la **ejecuciГіn eficiente**: en cualquier instante existe un Гєnico estado activo, lo que permite implementaciones directas en hardware o en software con complejidad lineal en la longitud de la entrada.

> **Idea fundamental de la conversiГіn**
>
> El AFN puede tener varias copias activas simultГЎneamente. La observaciГіn clave es que, en lugar de rastrear cada copia por separado, podemos rastrear el **conjunto de todas las copias activas** como si fuera un Гєnico objeto. Ese conjunto se convierte en el estado del AFD equivalente.
>
> Dicho de otra forma: si el AFN tiene $n$ estados, en cualquier momento el conjunto de copias activas es un subconjunto de $Q$ вЂ” y hay a lo sumo $2^n$ subconjuntos posibles. El AFD simplemente enumera esos subconjuntos como sus estados.

#### **De AFD a AFN вЂ• conversiГіn inmediata**

Si disponemos de un AFD $M=(\Sigma,Q,q_{0},F,\delta)$ basta definir

$$
\Delta(q,a)=\{\delta(q,a)\}\qquad\forall q\in Q,\;a\in\Sigma
$$

para obtener un AFN $M'=(\Sigma,Q,q_{0},F,\Delta)$.
Cada entrada de la tabla de transiciГіn pasa a ser un **conjunto unitario**; el comportamiento no se altera.

#### **De AFN a AFD вЂ• construcciГіn por subconjuntos**

> ***Teorema 3.7*** Para todo AFN $M=(\Sigma,Q,q_{0},F,\Delta)$ existe un AFD $M'=(\Sigma,Q',\{q_{0}\},F',\delta')$ con $L(M)=L(M')$.

El procedimiento se construye en cinco pasos:

1. **Estados del AFD**

   $$Q' \subseteq \wp(Q)$$

   Cada elemento de $Q'$ es un subconjunto de $Q$; representa *todas las posiciones posibles* en que el AFN podrГ­a hallarse tras leer cierto prefijo.

2. **Estado inicial**
   $\{q_{0}\}$ вЂ” al inicio solo la copia original estГЎ activa.

3. **FunciГіn de transiciГіn**
   Para un conjunto $S\subseteq Q$ y un sГ­mbolo $a\in\Sigma$:

   $$\delta'(S,a)=\bigcup_{q\in S}\Delta(q,a)$$

   Es decir, se combina en un solo paso lo que el AFN alcanzarГ­a en paralelo.

4. **Estados de aceptaciГіn**

   $$F'=\{\,S\in Q' \mid S\cap F\neq\varnothing\,\}$$

   Si alguna copia del AFN llega a un estado aceptable, el subconjunto lo refleja.

5. **Cierre de la tabla**
   Se genera la tabla de $\delta'$ de forma iterativa:
   * se parte de $\{q_{0}\}$;
   * se emplea cada sГ­mbolo de $\Sigma$ para producir nuevos subconjuntos;
   * el proceso continГєa hasta que no aparezcan combinaciones inГ©ditas.

El nГєmero mГЎximo de estados en $Q'$ es $2^{|Q|}$. Aunque puede crecer exponencialmente en teorГ­a, en la prГЎctica muchos subconjuntos no se alcanzan desde el estado inicial, y los compiladores modernos aplican tГ©cnicas adicionales de minimizaciГіn.

#### **Importancia del resultado**

* **Propiedad de cierre**: los lenguajes regulares no cambian de categorГ­a al pasar entre modelos; esto justifica el uso indistinto de AFN, AFD o expresiones regulares en especificaciones formales.

* **ImplementaciГіn de analizadores lГ©xicos**: las herramientas (por ejemplo, `lex` o `flex`) construyen primero un AFN a partir de expresiones regulares del usuario y luego aplican este procedimiento para obtener un AFD optimizado que se ejecuta a gran velocidad.

* **DiseГ±o guiado**: el ingeniero puede comenzar con el modelo mГЎs cГіmodo (normalmente AFN) sabiendo que siempre existe una versiГіn determinista lista para producciГіn.

### **3.7.3 ExtensiГіn formal de la funciГіn de transiciГіn**

#### **3.7.3.1 ExtensiГіn de la funciГіn de transiciГіn en un AFD**

Sea $M=\left(\Sigma, Q, q_0, F, \delta\right)$ un AFD. La funciГіn de transiciГіn $\delta: Q \times \Sigma \longrightarrow Q$ se extiende a una funciГіn $\widehat{\delta}: Q \times \Sigma^* \longrightarrow Q$ por medio de la siguiente definiciГіn recursiva:

$$
\left\{\begin{array}{l}
\widehat{\delta}(q, \lambda)=q, \quad q \in Q \\
\widehat{\delta}(q, a)=\delta(q, a), \quad q \in Q, \quad a \in \Sigma, \\
\widehat{\delta}(q, w a)=\delta(\widehat{\delta}(q, w), a), \quad q \in Q, a \in \Sigma, w \in \Sigma^*
\end{array}\right.
$$

La tabla original $\delta:Q\times\Sigma\to Q$ indica вЂ”sГ­mbolo por sГ­mboloвЂ” a quГ© estado se pasa; sin embargo, para caracterizar el lenguaje aceptado necesitamos saber en quГ© estado termina el autГіmata tras leer **toda** una cadena.

Por eso se define la **funciГіn extendida**

$$
\widehat{\delta}:Q\times\Sigma^{*}\longrightarrow Q,
$$

mediante la recursiГіn:

$$
\boxed{
\begin{aligned}
\widehat{\delta}(q,\lambda) &= q &&\text{(la cadena vacГ­a no provoca ningГєn cambio)}\\[2pt]
\widehat{\delta}(q,a) &= \delta(q,a) &&\text{(caso base: una sola letra)}\\[2pt]
\widehat{\delta}(q,wa) &= \delta\bigl(\,\widehat{\delta}(q,w),\,a\bigr) &&\text{(procesar el prefijo \(w\) y luego aplicar \(a\))}
\end{aligned}}
$$

* **Primera lГ­nea**: leer $\lambda$ implica "quedarse donde se estГЎ".
* **Segunda lГ­nea**: para una Гєnica letra se usa la tabla original.
* **Tercera lГ­nea**: para una palabra $wa$ se procesa primero el prefijo $w$; al estado resultante se aplica la transiciГіn con la letra final $a$.

**Ejemplo de aplicaciГіn**

Sea el AFD con $Q=\{q_0,q_1\}$, $\Sigma=\{a,b\}$ y tabla de transiciГіn:

$$\delta(q_0,a)=q_1,\quad\delta(q_1,a)=q_1,\quad\delta(q_0,b)=q_0,\quad\delta(q_1,b)=q_0$$

Calculemos $\widehat{\delta}(q_0,\, aba)$ paso a paso, aplicando la tercera regla de dentro hacia afuera (de izquierda a derecha):

$$
\begin{aligned}
\widehat{\delta}(q_0,\,aba)
&= \delta\!\bigl(\widehat{\delta}(q_0,\,ab),\;a\bigr)
&&\text{(separo la Гєltima letra: }w=ab,\;a=a\text{)}\\[4pt]
&= \delta\!\bigl(\delta\!\bigl(\widehat{\delta}(q_0,\,a),\;b\bigr),\;a\bigr)
&&\text{(separo de nuevo: }w=a,\;a=b\text{)}\\[4pt]
&= \delta\!\bigl(\delta\!\bigl(\delta(q_0,\,a),\;b\bigr),\;a\bigr)
&&\text{(caso base: una sola letra)}\\[4pt]
&= \delta\!\bigl(\delta(q_1,\;b),\;a\bigr)
&&\text{(}\delta(q_0,a)=q_1\text{)}\\[4pt]
&= \delta(q_0,\;a)
&&\text{(}\delta(q_1,b)=q_0\text{)}\\[4pt]
&= q_1
&&\text{(}\delta(q_0,a)=q_1\text{)}
\end{aligned}
$$

La cadena $aba$ deja al autГіmata en $q_1$. Si $q_1\in F$, la cadena se acepta.

La recursiГіn simplemente "desenvuelve" la cadena sГ­mbolo a sГ­mbolo de izquierda a derecha, aplicando la tabla $\delta$ en cada paso вЂ” exactamente lo que hace el autГіmata al recorrer la cinta.

Con esta extensiГіn:

$$
L(M)=\bigl\{\,w\in\Sigma^{*}\;\big|\;\widehat{\delta}(q_{0},w)\in F\,\bigr\}.
$$

> **Lectura**: una cadena se acepta si el estado que se alcanza tras procesarla completa pertenece a $F$.

Por comodidad, a partir de aquГ­ se omite el acento y se escribe $\delta(q,w)$ en lugar de $\widehat{\delta}(q,w)$.

#### **3.7.3.2 ExtensiГіn de la funciГіn de transiciГіn en un AFN**

Sea $M=\left(\Sigma, Q, q_0, F, \Delta\right)$ un AFN. La funciГіn de transiciГіn $\Delta: Q \times \Sigma \longrightarrow \wp(Q)$ se extiende inicialmente a conjuntos de estados. Para $a \in \Sigma$ y $S \subseteq Q$ se define

$$
\Delta(S, a):=\bigcup_{q \in S} \Delta(q, a)
$$

Luego se extiende $\Delta$ a una funciГіn $\widehat{\Delta}: Q \times \Sigma^* \longrightarrow \wp(Q)$, de manera similar a como se hace para los AFD. Recursivamente,

$$
\left\{\begin{array}{l}
\widehat{\Delta}(q, \lambda)=\{q\}, \quad q \in Q \\
\widehat{\Delta}(q, a)=\Delta(q, a), \quad q \in Q, a \in \Sigma, \\
\widehat{\Delta}(q, w a)=\Delta(\widehat{\Delta}(q, w), a)=\bigcup_{p \in \widehat{\Delta}(q, w)} \Delta(p, a), \quad q \in Q, a \in \Sigma, w \in \Sigma^*
\end{array}\right.
$$


#### (a) Paso intermedio: de estados individuales a **conjuntos**

Para $a\in\Sigma$ y cualquier subconjunto $S\subseteq Q$:

$$
\boxed{\;\Delta(S,a)\;=\;\bigcup_{q\in S}\Delta(q,a)\;}
$$

Es decir, al leer $a$ desde un conjunto $S$, el autГіmata avanza cada copia que estГ© en algГєn $q \in S$ segГєn la tabla, y reГєne todos los resultados en un Гєnico conjunto. Paso a paso:

1. Se toman todos los estados $q$ que forman el conjunto $S$.
2. Para cada uno se aplica $\Delta(q,a)$, obteniendo un conjunto de estados destino.
3. Todos esos conjuntos se unifican: el resultado es $\Delta(S,a)$.


#### (b) ExtensiГіn completa a cadenas

Se define

$$
\widehat{\Delta}:Q\times\Sigma^{*}\longrightarrow\wp(Q)
$$

con la recursiГіn:

$$
\boxed{
\begin{aligned}
\widehat{\Delta}(q,\lambda) &= \{q\} &&\text{(un solo estado activo; ningГєn movimiento)}\\[2pt]
\widehat{\Delta}(q,a) &= \Delta(q,a) &&\text{(un sГ­mbolo; se usa la tabla original)}\\[2pt]
\widehat{\Delta}(q,wa) &= \Delta\bigl(\widehat{\Delta}(q,w),a\bigr)
                       = \displaystyle\bigcup_{p\in\widehat{\Delta}(q,w)} \Delta(p,a)
                       &&\text{(avanzar todas las copias generadas por \(w\))}
\end{aligned}}
$$

* La primera lГ­nea produce un conjunto unitario: la copia que no se ha movido.
* La segunda usa directamente la tabla $\Delta$.
* La tercera procesa el prefijo $w$, obtiene el conjunto de copias vivas y, para cada una, aplica la transiciГіn con $a$.

**Ejemplo de aplicaciГіn**

Usamos el AFN del Ejemplo 1 (secciГіn 3.5.1), con la tabla:

$$
\Delta(q_0,a)=\{q_0,q_1,q_3\},\quad\Delta(q_1,b)=\{q_2\},\quad\Delta(q_2,b)=\{q_1,q_2\},\quad\Delta(q_3,b)=\{q_3\}
$$

(el resto de entradas es $\varnothing$). Calculemos $\widehat{\Delta}(q_0,\,ab)$:

**Paso 1** вЂ” Procesar el prefijo $a$ (caso base, una sola letra):
$$\widehat{\Delta}(q_0,\,a) = \Delta(q_0,\,a) = \{q_0,\,q_1,\,q_3\}$$

**Paso 2** вЂ” Procesar $b$ desde el conjunto $\{q_0,q_1,q_3\}$ (tercera regla):
$$
\widehat{\Delta}(q_0,\,ab)
= \Delta\!\bigl(\{q_0,q_1,q_3\},\;b\bigr)
= \Delta(q_0,b)\;\cup\;\Delta(q_1,b)\;\cup\;\Delta(q_3,b)
= \varnothing\;\cup\;\{q_2\}\;\cup\;\{q_3\}
= \{q_2,\,q_3\}
$$

Como $q_3\in F=\{q_0,q_1,q_3\}$, tenemos $\widehat{\Delta}(q_0,ab)\cap F=\{q_3\}\neq\varnothing$ в†’ la cadena $ab$ **se acepta**.

Este resultado coincide con lo que verГ­amos trazando los cГіmputos a mano: existe al menos una trayectoria que termina en $q_3\in F$.

Con esta extensiГіn, el lenguaje aceptado por $M$ queda caracterizado formalmente como:

$$
L(M)=\bigl\{\,w\in\Sigma^{*}\;\big|\;\widehat{\Delta}(q_{0},w)\cap F\neq\varnothing\,\bigr\}.
$$

> **Lectura**: la cadena se acepta si, despuГ©s de leerla completa, **al menos una** de las copias termina en un estado de aceptaciГіn.

Como antes, se permite escribir $\Delta(q,w)$ en lugar de $\widehat{\Delta}(q,w)$ cuando el contexto lo deja claro.

Estas extensiones son la herramienta formal que:

* conecta la tabla de un autГіmata con la definiciГіn de su lenguaje;  

* posibilita la **construcciГіn por subconjuntos** (AFN в†’ AFD) demostrando que el no determinismo no aГ±ade poder de reconocimiento;  

* fundamenta implementaciones prГЎcticas (analizadores lГ©xicos, buscadores de patrones) que primero generan un AFN y luego lo convierten en un AFD eficiente.

## **3.9 AutГіmatas con transiciones $\boldsymbol{\lambda}$ (AFN- $\boldsymbol{\lambda}$ )**

### **3.9.1 DefiniciГіn formal**

Un AFN-$\lambda$ es un autГіmata no determinista

$$
M=\bigl(\Sigma,\,Q,\,q_{0},\,F,\,\Delta\bigr)
$$

cuyas transiciones permiten, ademГЎs de los sГ­mbolos del alfabeto, el uso de la **cadena vacГ­a $\lambda$**:

$$
\Delta: Q \times (\Sigma \;\cup\;\{\lambda\}) \;\longrightarrow\; \wp(Q).
$$

#### **ВїQuГ© significa una transiciГіn $\lambda$?**

$$
\Delta(q,\lambda)=\{p_{1},p_{2},\dots,p_{k}\}
$$

* El autГіmata se encuentra en $q$.
* **Sin leer** el sГ­mbolo que estГЎ bajo la cabeza, y **sin avanzar** a la siguiente celda, puede "saltar" a **cualquiera** de los estados $p_{1},\dots,p_{k}$.

Estas transiciones se denominan indistintamente **$\lambda$-transiciones**, **transiciones nulas** o **transiciones espontГЎneas**. En el diagrama se dibujan como arcos etiquetados $\lambda$.

#### **IntuiciГіn operacional**

1. **Separar fases internas.** El autГіmata puede "prepararse" o cambiar de modo moviГ©ndose entre estados invisiblemente, antes de consumir el siguiente sГ­mbolo.

2. **Simplificar el diseГ±o.** Muchos patrones se expresan con menos estados cuando se permite insertar pasos vacГ­os.

3. **Puentes entre sub-mГЎquinas.** Se usan para conectar varios autГіmatas pequeГ±os que reconocen sub-expresiones de una expresiГіn regular.

#### **Criterio de aceptaciГіn**

Una palabra $w\in\Sigma^{*}$ se **acepta** si existe al menos una trayectoria que:

* **comienza** en $q_{0}$,
* **consume exactamente los sГ­mbolos de $w$** en el mismo orden,
* puede intercalar **cero o mГЎs pasos $\lambda$** en cualquier punto,
* **termina** en un estado de $F$.

Al igual que en un AFN convencional:

* Puede haber **varias ejecuciones paralelas** sobre la misma entrada.
* Algunas de ellas pueden **abortar** (quedar sin transiciГіn).
* La entrada se acepta si **al menos una** ejecuciГіn completa finaliza en aceptaciГіn.

#### **Por quГ© resultan Гєtiles en teorГ­a y prГЎctica**

| Ventaja                                           | ExplicaciГіn breve                                                                               |
| ------------------------------------------------- | ----------------------------------------------------------------------------------------------- |
| DiseГ±o modular                                    | Se encadenan fГЎcilmente autГіmatas mГЎs pequeГ±os mediante arcos $\lambda$.                        |
| Correspondencia directa con expresiones regulares | Los operadores de concatenaciГіn y opciГіn se traducen muy naturalmente a transiciones vacГ­as.    |
| Paso intermedio en compiladores                   | Herramientas como *flex* generan primero un $\lambda$-AFN y luego lo determinizan.              |
| Pruebas teГіricas                                  | Facilitan demostraciones de cierre de los lenguajes regulares (uniГіn, concatenaciГіn, estrella). |

En la siguiente secciГіn se mostrarГЎ cГіmo eliminar las transiciones vacГ­as para obtener un AFN sin $\lambda$ equivalente, garantizando que la potencia expresiva no aumenta: **aГ±adir $\lambda$-pasos simplifica el modelado, pero no permite reconocer lenguajes mГЎs allГЎ de los regulares.**

## **3.10 Equivalencia computacional entre los AFN-$\lambda$ y los AFN**

En este apartado mostramos que **aГ±adir transiciones О» no amplГ­a el poder de reconocimiento**: todo AFN-О» puede transformarse en un AFN sin О» que acepte exactamente el mismo lenguaje. De forma mГЎs grГЎfica, las transiciones espontГЎneas simplemente se **simulan** con rutas adicionales, sin modificar el conjunto de palabras reconocidas.

**RelaciГіn de inclusiГіn entre los tres modelos**

$$
\text{AFD} \;\subseteq\; \text{AFN} \;\subseteq\; \text{AFN-}\lambda
$$

Cada modelo es una extensiГіn del anterior: todo AFD es un AFN (con conjuntos unitarios), y todo AFN es un AFN-О» (sin transiciones vacГ­as). La equivalencia en **poder de reconocimiento** вЂ” los lenguajes que cada modelo puede aceptar вЂ” se establece en sentido contrario:

| ConversiГіn | DirecciГіn | ObservaciГіn |
| ---------- | --------- | ----------- |
| AFD в†’ AFN  | Trivial   | Envolver cada transiciГіn en un conjunto unitario |
| AFN в†’ AFD  | ConstrucciГіn por subconjuntos (secciГіn 3.7) | Puede aumentar el nГєmero de estados |
| AFN в†’ AFN-О» | Trivial | El AFN ya es un AFN-О» sin transiciones vacГ­as |
| **AFN-О» в†’ AFN** | **EliminaciГіn de О» (esta secciГіn)** | Se usa la О»-clausura |

1. **AFN в†’ AFN-О»**
   Cualquier AFN $M=(\Sigma,Q,q_{0},F,\Delta)$ es ya un AFN-О» que **no** utiliza transiciones О» (es decir, $\Delta(q,\lambda)=\varnothing$ para todo $q$).

2. **AFN-О» в†’ AFN**
   El resultado recГ­proco se enuncia en el siguiente teorema.

### **3.10.1 Teorema**

Dado un AFN-О»

$$
M = \bigl(\Sigma,\;Q,\;q_{0},\;F,\;\Delta\bigr),
$$

existe un AFN

$$
M' = \bigl(\Sigma,\;Q,\;q_{0},\;F',\;\Delta'\bigr)
$$

tal que $L(M') = L(M)$.

**Bosquejo de la demostraciГіn**

Para eliminar las О»-transiciones de $M$ construimos $M'$ de manera puramente algorГ­tmica, apoyГЎndonos en la nociГіn de **О»-clausura**.

1. **О»-clausura de un estado**

   $$
   \lambda[q]
   = \{\,p\in Q \mid
       \text{existe un camino de 0 Гі mГЎs saltos }q\xrightarrow{\lambda}\cdots\xrightarrow{\lambda}p
     \}.
   $$

   * Incluye siempre $q$ (cero saltos).
   * ReГєne todos los estados alcanzables **solo** con movimientos О».

2. **О»-clausura de un conjunto**
   Para $S=\{q_{1},\dots,q_{k}\}\subseteq Q$,

   $$
     \lambda[S]
     = \bigcup_{i=1}^{k} \lambda[q_{i}],
     \qquad
     \lambda[\varnothing] = \varnothing.
   $$

3. **Nueva funciГіn de transiciГіn**

   $$
   \boxed{
     \Delta'(q,a)
     \;=\;
     \lambda\bigl[\;\Delta\bigl(\lambda[q],\,a\bigr)\bigr]
   }
   $$

   **InterpretaciГіn paso a paso:**
   * *Antes* de leer $a$: se calculan todos los estados alcanzables desde $q$ solo con О» ($\lambda[q]$).
   * *Con* $a$: se aplica $\Delta(\cdot, a)$ a cada estado de $\lambda[q]$.
   * *DespuГ©s* de $a$: se vuelven a activar las О»-transiciones salientes desde los estados resultantes.

4. **Estados de aceptaciГіn**

   $$
     F'
     = \bigl\{\,q\in Q\mid \lambda[q]\;\cap\;F\neq\varnothing\,\bigr\}.
   $$

   Un estado $q$ es aceptante en $M'$ si **algГєn** estado alcanzable por О» desde $q$ era aceptante en $M$.

Con estos cuatro pasos, $M'$ replica fielmente cada cГіmputo de $M$ **sin** usar transiciones О». La construcciГіn es completamente mecГЎnica y se basa en recorrer grafos finitos (cГЎlculo de clausuras y uniones), garantizando $L(M')=L(M)$.

---

**Ejemplo de aplicaciГіn.**

A continuaciГіn aplicaremos el procedimiento al AFN-О» presentado en la secciГіn 3.9.2, cuyo lenguaje es

$$

$$

Nuestro objetivo es **eliminar** las transiciones $\lambda$ construyendo un AFN equivalente sin pasos espontГЎneos. El primer paso consiste en calcular la **О»-clausura** de cada estado, es decir, el conjunto de todos los estados a los que se puede llegar **solo** con saltos $\lambda$, incluyendo el propio estado de partida.

**CГЎlculo de las О»-clausuras**

En el grafo hay dos transiciones $\lambda$: de $q_{0}$ a $q_{1}$, y de $q_{1}$ a $q_{2}$. No hay $\lambda$ desde $q_{2}$. Por tanto:

$$
\begin{aligned}
\lambda[q_{0}]
&= \{\,q_{0}\}\;\cup\;\{q_{1}\}\;\cup\;\{q_{2}\}
 = \{q_{0},\,q_{1},\,q_{2}\},\\[4pt]
\lambda[q_{1}]
&= \{\,q_{1}\}\;\cup\;\{q_{2}\}
 = \{q_{1},\,q_{2}\},\\[4pt]
\lambda[q_{2}]
&= \{q_{2}\}.
\end{aligned}
$$

* **$\lambda[q_{0}]$** incluye $q_{0}$ (cero saltos), luego $q_{1}$ y finalmente $q_{2}$ al encadenar las dos О»-transiciones.
* **$\lambda[q_{1}]$** arranca en $q_{1}$ y salta una vez a $q_{2}$.
* **$\lambda[q_{2}]$** queda solo en $q_{2}$ porque no hay О»-salidas.

**Nueva funciГіn de transiciГіn $\Delta'$**

La funciГіn $\Delta^{\prime}:Q\times\{a,b,c\}\to\wp(\{q_0,q_1,q_2\})$ queda:

$$
\begin{aligned}
\Delta^{\prime}(q_0,a)&=\lambda\bigl[\Delta(\{q_0,q_1,q_2\},a)\bigr]
                      =\lambda[\{q_0\}]
                      =\{q_0,q_1,q_2\},\\
\Delta^{\prime}(q_0,b)&=\lambda\bigl[\Delta(\{q_0,q_1,q_2\},b)\bigr]
                      =\lambda[\{q_1\}]
                      =\{q_1,q_2\},\\
\Delta^{\prime}(q_0,c)&=\lambda\bigl[\Delta(\{q_0,q_1,q_2\},c)\bigr]
                      =\lambda[\{q_2\}]
                      =\{q_2\},\\
\Delta^{\prime}(q_1,a)&=\lambda\bigl[\Delta(\{q_1,q_2\},a)\bigr]
                      =\lambda[\varnothing]
                      =\varnothing,\\
\Delta^{\prime}(q_1,b)&=\lambda\bigl[\Delta(\{q_1,q_2\},b)\bigr]
                      =\lambda[\{q_1\}]
                      =\{q_1,q_2\},\\
\Delta^{\prime}(q_1,c)&=\lambda\bigl[\Delta(\{q_1,q_2\},c)\bigr]
                      =\lambda[\{q_2\}]
                      =\{q_2\},\\
\Delta^{\prime}(q_2,a)&=\lambda\bigl[\Delta(\{q_2\},a)\bigr]
                      =\lambda[\varnothing]
                      =\varnothing,\\
\Delta^{\prime}(q_2,b)&=\lambda\bigl[\Delta(\{q_2\},b)\bigr]
                      =\lambda[\varnothing]
                      =\varnothing,\\
\Delta^{\prime}(q_2,c)&=\lambda\bigl[\Delta(\{q_2\},c)\bigr]
                      =\lambda[\{q_2\}]
                      =\{q_2\}.
\end{aligned}
$$

InterpretaciГіn de las transiciones clave:

1. **$\Delta'(q_0,a)=\{q_0,q_1,q_2\}$**: leer $a$ desde $q_0$ devuelve a $q_0$; las О»-clausuras activan tambiГ©n $q_1$ y $q_2$.
2. **$\Delta'(q_0,b)=\{q_1,q_2\}$**: la $b$ solo la consume $q_1$ (en la clausura de $q_0$); luego О» activa $q_2$.
3. **$\Delta'(q_0,c)=\{q_2\}$**: la $c$ solo la consume $q_2$.
4. **$\Delta'(q_1,a)=\varnothing$**: ni $q_1$ ni $q_2$ tienen transiciГіn con $a$.
5. **$\Delta'(q_1,b)=\{q_1,q_2\}$**: leer $b$ mantiene en $q_1$; la О»-clausura aГ±ade $q_2$.
6. **$\Delta'(q_1,c)=\{q_2\}$**: la $c$ solo llega a $q_2$.
7вЂ“9. **Desde $q_2$**: solo la $c$ tiene transiciГіn ($q_2$ cicla con $c$); $a$ y $b$ no tienen salida.
