# **5. Otras Propiedades de los lenguajes regulares**

En los capítulos anteriores hemos visto que los lenguajes regulares se caracterizan por poder describirse con autómatas finitos o expresiones regulares, y que estos modelos son equivalentes (Teorema de Kleene). Además, los lenguajes regulares **se comportan bien** ante operaciones habituales:

* **Operaciones booleanas**
  *Unión, intersección y complemento* conservan la regularidad. Si $L_1$ y $L_2$ son regulares, también lo son $L_1\cup L_2$, $L_1\cap L_2$ y $\Sigma^*\setminus L_1$.

* **Homomorfismos**
  Dado un homomorfismo $\phi:\Sigma\to\Gamma^*$, tanto la imagen $\phi(L)$ como la **imagen inversa** $\phi^{-1}(L)$ de un lenguaje regular $L\subseteq\Gamma^*$ son regulares.

* **Problemas de decisión**
  Para lenguajes regulares existen algoritmos eficaces que deciden:

  * *Pertinencia* ($w\in L$?)
  * *Vacío* ($L=\emptyset$?)
  * *Finitud* (¿$L$ tiene un número finito de cadenas?)
  * *Equivalencia* ($L_1=L_2$?)
  * *Inclusión* ($L_1\subseteq L_2$?)

Sin embargo, **no todo lenguaje** sobre un alfabeto dado es regular. Para demostrar que un lenguaje *no* puede reconocerse con un autómata finito usamos la herramienta siguiente.

## **5.1. Lema de bombeo**

El **lema de bombeo** ([pumping lemma](https://en.wikipedia.org/wiki/Pumping_lemma_for_regular_languages)) es una propiedad que todo [lenguaje regular](https://es.wikipedia.org/wiki/Lenguaje_regular) debe cumplir. Si un lenguaje **no** obedece esta propiedad, **no** puede ser regular.

> **Lema de bombeo.**
> Sea $L$ un lenguaje regular sobre $\Sigma$. Entonces existe un entero $n\ge1$ (la *longitud de bombeo* para $L$) tal que **toda** cadena $w\in L$ con $|w|\ge n$ puede descomponerse
>
> $$
> w = u\,v\,x
> $$
>
> cumpliendo:
>
> 1. $\lvert u\,v\rvert \le n$,
> 2. $v\neq\lambda$ (el *bloque* $v$ no está vacío),
> 3. para **todo** $i\ge0$, la cadena bombeada
>
>    $$
>      u\,v^i\,x
>    $$
>
>    también pertenece a $L$.

donde

1. **Constante $n$**
   Depende **solo** de $L$. Es la cabeza de “cuántos **estados tiene el autómata mínimo**” (o equivalente en expresiones regulares).

2. **Cadena larga**
   Si eliges cualquier $w\in L$ con longitud $\lvert w\rvert\ge n$, entonces al procesarla habrá visitado **más de $n$ estados** en el autómata, forzando algún **ciclo**.

3. **Descomposición**

   * $u$ recoge los primeros símbolos hasta llegar al ciclo.
   * $v$ es la parte que se repite cada vez que el autómata recorre ese ciclo.
   * $x$ es el resto final de la cadena.

4. **Bombeo**
   Al repetir ($i>1$) o eliminar ($i=0$) el bloque $v$, seguimos circulando en el mismo ciclo, de modo que **la cadena resultante** también es aceptada.

- **Uso típico para probar no–regularidad**

Para demostrar que un lenguaje $L$ **no** es regular:

1. **Supón** que $L$ sí es regular y deja $n$ su constante de bombeo.
2. **Elige** cuidadosamente una cadena $w\in L$ con $|w|\ge n$.
3. **Para toda** descomposición $w=uvx$ con $|uv|\le n$ y $v\neq\lambda$, exhibe un $i$ para el cual

   $$
   u\,v^i\,x\;\notin\;L.
   $$
4. Al llegar a una contradicción con la propiedad del lema de bombeo, **concluye** que $L$ no puede ser regular.

Con esta herramienta se cierra el ciclo: hemos visto **cómo** construir autómatas de expresiones regulares (Teorema de Kleene, Parte I), **cómo** extraer expresiones de autómatas (Parte II), y **cómo** certificar que algunos lenguajes **no** entran en la clase regular (Lema de bombeo).


### **5.1.1 Ejemplos**

#### **Ejemplo 1:** Usar el lema de bombeo para demostrar que

$$
L \;=\;\{\,a^{i}b^{i}\mid i\ge0\}
$$

no es regular.

**Solución**

1. **Suposición inicial y constante de bombeo**
   Supongamos, buscando contradicción, que $L$ **sí** es regular.
   Entonces existe un entero $n\ge1$, la *constante de bombeo* para $L$.

2. **Elección de la cadena larga**
   Elijamos

   $$
   w = a^{n}\,b^{n}\;\in\;L.
   $$

   Obsérvese que $\lvert w\rvert = 2n\ge n$, cumpliendo la hipótesis del lema.

3. **Descomposición según el lema**
   El lema de bombeo garantiza la existencia de tres subcadenas $u,v,x$ tales que

   * $w = u\,v\,x$,
   * $\lvert u\,v\rvert \le n$,
   * $\lvert v\rvert \ge 1$,
     y para **todo** $i\ge0$, la cadena bombeada
     $\,u\,v^{i}\,x$ pertenece a $L$.

4. **Identificación de $u,v,x$**
   Como $\lvert u\,v\rvert\le n$, **tanto** $u$ **como** $v$ consisten **solo** en `a`’s (los primeros $n$ símbolos de $w$ son todos `a`).
   Por tanto podemos escribir

   $$
   u = a^{r},\quad
   v = a^{s},\quad
   \text{con }r\ge0,\;s\ge1,
   $$

   y luego

   $$
   x = a^{\,n-(r+s)}\,b^{n}.
   $$

5. **Caso $i=0$ (“bombear hacia abajo”)**
   Formamos

   $$
   u\,v^{0}\,x = u\,x
                 = a^{r}\;a^{\,n-(r+s)}\;b^{n}
                 = a^{\,n-s}\;b^{n}.
   $$

   Para que $a^{n-s}b^{n}$ pertenezca a $L$ sería necesario $n-s = n$, pero como $s\ge1$ esto **no** se cumple.
   Entonces $u\,v^{0}\,x \notin L$, contradiciendo la propiedad de bombeo.

6. **Caso $i=2$ (“bombear hacia arriba”)**
   Alternativamente, con $i=2$:

   $$
   u\,v^{2}\,x
   = a^{r}\;a^{2s}\;a^{\,n-(r+s)}\;b^{n}
   = a^{\,n+s}\;b^{n}.
   $$

   Dado que $s\ge1$, también $n+s\neq n$, por lo que $a^{n+s}b^{n}\notin L$.
   Nuevamente, contradicción.

7. **Conclusión**
   En **ambos** casos hemos hallado $i$ tal que $u\,v^{i}\,x\notin L$, violando el lema de bombeo.
   Por lo tanto, nuestra suposición original era falsa: **$L$ no es regular**.

> **Nota.** El mismo razonamiento vale para
> $\{\,a^{i}b^{i}:i\ge1\}$, pues basta elegir
> $w=a^{n}b^{n}$ con $n\ge1$ y repetir los pasos anteriores.


#### **Ejemplo 2:** Demostrar que el lenguaje de los palíndromos sobre $\{a,b\}$ no es regular

Recordemos que un **palíndromo** es una cadena $w$ que coincide con su reversa:

$$
w = w^{R}.
$$

Denotemos

$$
L = \{\,w\in\{a,b\}^*\mid w = w^{R}\}.
$$

Vamos a ver por qué $L$ **no** puede reconocerse con un autómata finito, usando el lema de bombeo.

##### **Solución**

1. **Supongamos** que $L$ es regular. Entonces existe una **constante de bombeo** $n\ge1$ que cumple las condiciones del lema.

2. **Elección de la cadena test**
   Tomemos

   $$
   w = a^{n}\;b\;a^{n}.
   $$

   * Claramente $\lvert w\rvert = 2n+1 \ge n$.
   * Y $w$ **sí** es un palíndromo (misma secuencia de $a$ antes y después de la $b$).

3. **Descomposición del lema**
   El lema de bombeo garantiza que podemos partir

   $$
   w = u\,v\,x
   \quad\text{con}\quad
   \lvert u\,v\rvert \le n,\;\lvert v\rvert \ge1,
   $$

   de modo que **para todo** $i\ge0$, la “cadena bombeada”
   $\;u\,v^{i}\,x$;$\in L$.

4. **Localización de $u$ y $v$**
   Dado que $\lvert u\,v\rvert\le n$, las posiciones de $u$ y $v$ quedan **dentro** del bloque inicial de $a$s. Por tanto

   $$
   u = a^{r},\quad
   v = a^{s},
   \quad
   \text{con }0\le r,\;1\le s,\;r+s\le n.
   $$

5. **Forma de $x$**
   Como $w = a^{n}b\,a^{n} = a^{r}\,a^{s}\,x$, despejamos

   $$
   x = a^{\,n-(r+s)}\,b\,a^{n}.
   $$

6. **Bombeo hacia abajo** ($i=0$)
   Al poner $i=0$ obtenemos

   $$
   u\,v^{0}\,x \;=\; u\,x
   \;=\;
   a^{r}\;a^{\,n-(r+s)}\,b\,a^{n}
   \;=\;
   a^{\,n-s}\;b\;a^{n}.
   $$

   Esta nueva cadena **no** es un palíndromo, porque tras el `b` quedan $n$ `a`’s, mientras que antes hay solo $n-s$ `a`’s (y $n-s<n$). Por tanto
   $\;u\,v^{0}\,x\notin L$.

7. **Contradicción**
   Hemos hallado un $i$ (concretamente $i=0$) para el cual la cadena bombeada no pertenece a $L$, violando la condición “para todo $i\ge0$, $u\,v^{i}\,x\in L$”.

8. **Conclusión**
   La suposición de que $L$ es regular conduce a una contradicción.

   $$
   \boxed{L\text{ no es regular.}}
   $$


> **Idea clave.** Al “bombear” las $a$'s del prefijo se desajusta la simetría requerida para palíndromos, demostrando que ningún autómata finito (con memoria limitada) puede verificar igualdad perfecta de ambas mitades.


### **5.1.2 Ejemplos cotidianos verificados con el Lema de Bombeo**

#### **- Ejemplo 1: Patente de vehículo (dos letras + cuatro dígitos)**

En muchos países las matrículas de automóviles siguen el formato

$$
\text{L}\,\text{L}\,\text{D}\,\text{D}\,\text{D}\,\text{D},
$$

donde cada **L** es una letra (A–Z) y cada **D** un dígito (0–9).

**Lenguaje formal**

$$
L_{pat} = \{\,\ell_1\ell_2d_1d_2d_3d_4 \mid \ell_i\in A\!-\!Z,\;d_j\in0\!-\!9\}.
$$

Es un lenguaje **regular** porque cabe describirlo con la expresión regular

$$
[A\text{-}Z][A\text{-}Z][0\text{-}9][0\text{-}9][0\text{-}9][0\text{-}9].
$$

**Verificación con el lema de bombeo**

* La longitud de **todas** las cadenas de $L_{pat}$ es fija (=6).
* Si elegimos cualquier constante de bombeo $n>6$, **no existe** $w\in L_{pat}$ con $\lvert w\rvert\ge n$.
* El lema exige que “todas las $w$ largas se puedan bombear”, pero al no haber $w$ que cumpla $\lvert w\rvert\ge n$ la condición se satisface **vacuamente**.
* Con ello, $L_{pat}$ cumple el lema y es regular.


#### **- Ejemplo 2: Secuencias de semáforo que terminan en verde**

Un peatón ve una serie de indicadores: **R**ojo (R) o **V**erde (V). Queremos modelar todas las secuencias que **terminan en verde**.

**Lenguaje formal**

$$
L_{sem} = \{\,w\in\{R,V\}^* \mid w\text{ acaba en }V\}
      = (R\cup V)^*\,V.
$$

Es regular, ya que la expresión $(R|V)^*V$ lo describe perfectamente.

**Verificación con el lema de bombeo**

* Tomemos la constante de bombeo $n$ que el lema garantice.
* Sea $w\in L_{sem}$ con $\lvert w\rvert\ge n$. Entonces $w=u\,v\,x$ cumple $\lvert u\,v\rvert\le n$ y $v\neq\lambda$.
* Como $\lvert u\,v\rvert\le n$, el bloque $v$ está formado solo por símbolos del prefijo de longitud $\le n$, ninguno de los cuales afecta al **último** símbolo de $w$.
* Bombear $v$ (poner $i=0$ o $i>1$) **no altera** el símbolo final de $w$, que sigue siendo $V$.
* Por tanto, para todo $i\ge0$, la cadena $u\,v^i\,x$ todavía termina en $V$ y pertenece a $L_{sem}$.
* Esto verifica la propiedad de bombeo, confirmando que $L_{sem}$ es regular.


#### **- Ejemplo3: Contrajemplo - Igual número de escaneos y embolsados**

En un supermercado un cajero **escanea** cada artículo (`s`) y luego lo **embolsa** (`b`). Supongamos que, por política, **el número de escaneos debe coincidir con el de embolsados** en cada transacción.

**Lenguaje formal**

$$
L_{sb} = \{\,s^n b^n \mid n\ge0\}.
$$

Es el conjunto de cadenas con $n$ veces `s` seguidas de $n$ veces `b`.

**Por qué no es regular (uso del lema de bombeo)**

1. **Supongamos** que $L_{sb}$ fuera regular y deje $n$ su constante de bombeo.
2. **Escojamos** la cadena

   $$
   w = s^n\,b^n\in L_{sb},
   \quad |w|=2n\ge n.
   $$
3. Según el lema, $w$ se descompone en $u\,v\,x$ con
   $\lvert u\,v\rvert\le n$ y $v\neq\lambda$.
   Esto implica que $v$ contiene **solo** símbolos `s` (porque los primeros $n$ símbolos son todos `s`).
4. Tomemos $i=0$. La cadena bombeada

   $$
   u\,v^{0}\,x = u\,x
   $$

   tiene **menos** `s` que `b` (ya que hemos eliminado uno o más `s` de la primera parte), en concreto
   $\#s< n$ pero $\#b=n$, luego

   $$
   u\,x\notin L_{sb}.
   $$
5. Esto **contradice** la condición “para todo $i\ge0$, $u\,v^i\,x\in L$”.
6. Concluimos que **no** existe autómata finito que reconozca $L_{sb}$, es decir, $L_{sb}$ **no** es regular.

**Resumen**

* Los dos primeros lenguajes (**placas de coche**, **señales que terminan en verde**) cumplen la propiedad de bombeo y son regulares.
* El tercero (**escaneo–embolsado igual**) viola el lema y, por lo tanto, no puede ser regular.

## **5.2. Propiedades de clausura**

Las propiedades de clausura nos indican que, si partimos de lenguajes que ya sabemos reconocer con autómatas finitos (es decir, lenguajes regulares), al aplicarles ciertas operaciones seguiremos obteniendo lenguajes que también son regulares. En otras palabras, “la regularidad se conserva” al combinar, repetir o modificar lenguajes por medio de estas operaciones.

#### **5.2.1 Teorema.**

Si $L, L_1$ y $L_2$ son lenguajes regulares sobre un alfabeto $\Sigma$, entonces también lo son los siguientes:

|     | Notación                      | Operación            |
| --- | ----------------------------- | -------------------- |
| (1) | $L_1 \cup L_2$                | unión                |
| (2) | $L_1 L_2$                     | concatenación        |
| (3) | $L^*$                         | estrella de Kleene   |
| (4) | $L^{+}$                       | clausura positiva    |
| (5) | $\overline{L} = \Sigma^* - L$ | complemento          |
| (6) | $L_1 \cap L_2$                | intersección         |
| (7) | $L_1 - L_2$                   | diferencia           |
| (8) | $L_1 \triangleleft L_2$       | diferencia simétrica |

##### **Explicación de cada operación**

1. **Unión** $\;L_1 \cup L_2$
   Reúne todas las cadenas que pertenecen a $L_1$ *o* a $L_2$ (o a ambos).

   > *Intuitivo:* “¿Está en $L_1$ o está en $L_2$?”

2. **Concatenación** $\;L_1 L_2$
   Forma cadenas tomando primero una de $L_1$ y pegando después una de $L_2$.

   > *Intuitivo:* “Escribe algo de $L_1$ y a continuación algo de $L_2$.”

3. **Estrella de Kleene** $\;L^*$
   Permite repetir cualquier cantidad de veces (incluso cero) las cadenas de $L$.

   > *Intuitivo:* “¿Cuántas veces quieres repetir una cadena de $L$?”

4. **Clausura positiva** $\;L^{+}$
   Similar a la estrella, pero exige al menos una repetición.

   > *Intuitivo:* “Repite cadenas de $L$, pero al menos una vez.”

5. **Complemento** $\;\overline{L} = \Sigma^* - L$
   Todas las cadenas posibles sobre $\Sigma$ *que no* están en $L$.

   > *Intuitivo:* “Todo lo que no era de $L$.”

6. **Intersección** $\;L_1 \cap L_2$
   Solo las cadenas que están *a la vez* en $L_1$ y en $L_2$.

   > *Intuitivo:* “¿Está tanto en $L_1$ como en $L_2$?”

7. **Diferencia** $\;L_1 - L_2$
   Las cadenas de $L_1$ *que no* aparecen en $L_2$.

   > *Intuitivo:* “Está en $L_1$ pero no en $L_2$.”

8. **Diferencia simétrica** $\;L_1 \triangleleft L_2$
   Las cadenas que están en exactamente uno de los dos lenguajes, pero no en ambos.

   > *Intuitivo:* “¿Está en uno u otro, pero no en los dos al mismo tiempo?”

#### **¿Por qué siguen siendo regulares?**
Para cada una de estas operaciones podemos construir de forma mecánica un nuevo autómata (o expresiones regulares equivalentes) partiendo de los autómatas originales. Por ejemplo:

* **Unión e intersección**: existe la construcción de producto de autómatas con estados pareados.
* **Concatenación y estrella**: basta con “conectar” autómatas, añadiendo transiciones $\varepsilon$.
* **Complemento**: si tenemos un autómata determinista completo, invertimos sus estados finales y no finales.
* **Diferencias** y **diferencia simétrica** se deducen combinando complemento e intersección o unión de forma aditiva.

Cada paso mantiene la forma finita y determinista (o puede llevarse a DFA), por lo que el resultado siempre es un autómata finito: ¡la regularidad se conserva!


### **5.2.2 Otros aspectos importantes a considerar**

**Recordatorio.** Un *álgebra booleana de conjuntos* sobre un conjunto base $X$ es una colección $\mathcal{A}\subseteq\wp(X)$ que cumple:

1. Está cerrada bajo **unión** ($\cup$), **intersección** ($\cap$) y **complemento** ($\bar{\,\cdot\,}$).
2. Contiene al conjunto vacío $\varnothing$ y al conjunto completo $X$.


### **5.2.3 Corolario**

La familia $\mathcal{R}\subseteq\wp(\Sigma^*)$ de *todos* los lenguajes regulares sobre un alfabeto $\Sigma$ forma un álgebra booleana de conjuntos.

**Demostración sencilla**

1. Por el Teorema 3.2.1 sabemos que $\mathcal{R}$ está cerrada bajo unión, intersección y complemento.
2. Además, los lenguajes $\varnothing$ y $\Sigma^*$ (todas las cadenas posibles sobre $\Sigma$) son regulares.
3. Por tanto, $\mathcal{R}$ cumple los requisitos para ser un álgebra booleana.


### **5.2.4 ¿Qué debemos tener en cuenta?**

1. **Uniones finitas**

   * Si juntamos un *número finito* de lenguajes regulares, el resultado sigue siendo regular.
   * Ejemplo: $L = L_1\cup L_2\cup L_3$ es regular siempre que $L_1,L_2,L_3$ lo sean.

2. **Uniones infinitas**

   * Una unión *infinita* de lenguajes regulares **no garantiza** regularidad.
   * **Ejemplo concreto:**

     $$
       L \;=\;\{\,a^n b^n : n\ge1\}
         \;=\;\bigcup_{i=1}^{\infty}\{\,a^i b^i\}.
     $$

     Cada $\{a^i b^i\}$ es regular (solo contiene una cadena), pero la unión $L$ **no** es regular.

3. **Sublenguajes**

   * Tomar un *subconjunto* de un lenguaje regular NO asegura que siga siendo regular.
   * Dicho de otra forma, un lenguaje regular puede “albergar” sublenguajes que no lo son.
   * **Ejemplo:**

     $$
       a^* b^*\quad\text{es regular, pero}\quad
       \{\,a^n b^n : n\ge1\}
       \subset a^* b^*
     $$

     Aquí el sublenguaje $\{a^n b^n\}$ **no** es regular.


### **5.2.5 Uso de las propiedades de clausura para probar no-regularidad**

Cuando queremos demostrar que un lenguaje $L$ *no* es regular, a menudo suponemos lo contrario y aplicamos operaciones de clausura que sí preservan la regularidad. Si al final obtenemos un lenguaje que *sabemos* no es regular, la contradicción nos obliga a concluir que nuestra suposición inicial (que $L$ *sí* es regular) era falsa.

**Ejemplo esquemático:**

1. Supongamos $L$ regular.
2. Apliquemos alguna combinación de unión, intersección o complemento para llegar a $\{a^n b^n\}$.
3. Como $\{a^n b^n\}$ no es regular, la hipótesis de partida falla.
4. Concluimos: **$L$ no es regular**.

De este modo, las propiedades de clausura no solo nos ayudan a construir nuevos lenguajes regulares, sino también a identificar con rigor los que no lo son.

### **5.2.6 Ejemplos**

#### - **Ejemplo 1:**

**Lenguaje**

$$
L \;=\;\{\,a^i b^j : i,j \ge 0,\; i \neq j\}.
$$

**Idea clave**
Si $L$ fuera regular, como los lenguajes regulares son cerrados bajo complemento, su complemento

$$
\overline{L} \;=\;\Sigma^* - L
$$

también lo sería. Pero fijémonos en

$$
\Sigma^* = a^* b^* \quad(\text{todas las cadenas de ’a’ seguidas de ’b’}),
$$

de modo que

$$
\overline{L} = a^* b^* - L
           = \{\,a^i b^j : i=j\}
           = \{\,a^n b^n : n\ge0\},
$$

y ya sabemos (por el lema de bombeo o por argumentos anteriores) que $\{a^n b^n: n\ge0\}$ **no** es regular.

**Conclusión**
La suposición “$L$ es regular” lleva a que $\{a^n b^n\}$ también lo sea, contradicción. Por tanto, **$L$ no es regular**.


#### - **Ejemplo 2:**

**Lenguaje**

$$
L \;=\;\{\,w\,b^n : w\in\Sigma^*,\;|w|=n,\;n\ge1\}
\quad\text{sobre }\Sigma=\{a,b\}.
$$

Es decir, cadenas formadas por alguna palabra $w$ seguida de $n$ letras $b$, donde $n$ es la longitud de $w$.

**Paso 1: intersectar con un lenguaje regular**
Intersectamos $L$ con el lenguaje regular $a^*b^*$ (todas las cadenas consistentes en un bloque de $a$s seguido de un bloque de $b$s).

$$
L\;\cap\;a^*b^*
\;=\;\{\,w b^n : |w|=n\}\;\cap\;\{a^*b^*\}
\;=\;\{\,a^n b^n : n\ge1\}.
$$

**Paso 2: recordar que $\{a^n b^n\}$ no es regular**
Por lo tanto, si $L$ fuera regular, la intersección con un regular produciría un lenguaje no regular, lo cual es imposible.

**Conclusión**
**$L$ no es regular**.

#### - **Ejemplo 3: Rutas de montaña**

**Contexto**
Imaginemos un alfabeto $\Sigma=\{U,D\}$ donde “$U$” significa “subida” y “$D$” “bajada”. Definimos

$$
L = \{\,\text{secuencia de ’U’ y ’D’ en la que el número de subidas *no* coincide con el de bajadas}\}.
$$

Este lenguaje modela rutas que no vuelven al punto de partida.

1. **Suponemos** $L$ regular.
2. Como los regulares son cerrados bajo complemento, $\overline{L}$ sería regular.
3. Pero

   $$
   \overline{L}
     = \{\,\text{secuencias con mismo número de ’U’ y ’D’}\}
     = \{\,U^nD^n : n\ge0\},
   $$

   que **no** es regular.
4. **Conclusión:** $L$ no es regular.



#### - **Ejemplo 4: Listado de tareas con encabezado**

**Contexto**
Supongamos que un pequeño programa lee una cadena del formato

$$
\text{«header}\#\text{items»},
$$

donde «header» es una palabra de longitud $n$ (por ejemplo un título) y «items» es exactamente la lista de $n$ símbolos “$\ast$” (cada “$\ast$” representa un ítem). Definimos

$$
L = \{\,w\#\ast^n : w\in\{\texttt{a–z}\}^*,\;|w|=n\}.
$$

Para verificar en tiempo constante que el número de ítems coincide con la longitud del encabezado:

1. Intersectamos con el regular $\{\texttt{a–z}\}^*\#\ast^*$ (todas las cadenas con un bloque de letras, “#”, y luego estrellas).
2. Obtenemos

   $$
   L\cap(\{\texttt{a–z}\}^*\#\ast^*)
   = \{\,a^n\#\ast^n : n\ge1\},
   $$

   equivalente a $\{a^n\#b^n\}$ cambiado de símbolos.
3. Como $\{a^n\#\ast^n\}$ no es regular, la suposición de regularidad de $L$ falla.

**Conclusión:** este sencillo formato de encabezado+lista **no** puede reconocerse con un autómata finito.


En todos estos casos vemos cómo, tomando el supuesto de regularidad y usando **cerradura** (complemento, intersección), llegamos a un lenguaje $\{a^n b^n\}$ o su análogo, que sabemos no es regular. De este modo, **demostramos con rigor** que el lenguaje original tampoco lo es.

## **5.3. Propiedades de clausura para autómatas**

Las operaciones sobre lenguajes regulares que vimos en el Teorema 3.2.1 no son sólo “propiedades abstractas”: para cada una de ellas existe un **método concreto** (un algoritmo de construcción) que, dado(s) autómata(s) de partida, produce otro autómata que acepta exactamente el lenguaje resultante.

### **5.3.1 Teorema**

> **Enunciado.**
> Sean $M, M_1$ y $M_2$ autómatas finitos (deterministas, no deterministas, o con transiciones $\varepsilon$). Denotemos
>
> $$
> L(M)=L,\quad L(M_1)=L_1,\quad L(M_2)=L_2.
> $$
>
> Entonces existen **construcciones efectivas** para obtener autómatas que reconozcan los siguientes lenguajes:
>
> 1. $L_1 \cup L_2$ (unión)
> 2. $L_1L_2$ (concatenación)
> 3. $L^*$ (estrella de Kleene)
> 4. $L^+$ (clausura positiva)
> 5. $\overline{L} = \Sigma^* - L$ (complemento)
> 6. $L_1 \cap L_2$ (intersección)
> 7. $L_1 - L_2$ (diferencia)
> 8. $L_1 \triangleleft L_2$ (diferencia simétrica)

A continuación, describimos en términos sencillos **cómo** se “programa” cada operación:

1. **Unión ($L_1\cup L_2$)**

   * Tomar los dos autómatas $M_1, M_2$.
   * Crear un nuevo estado inicial $q_0'$ con transiciones $\varepsilon$ hacia los inicios originales de $M_1$ y $M_2$.
   * Los estados finales son la unión de los finales de $M_1$ y $M_2$.

2. **Concatenación ($L_1L_2$)**

   * Partir de $M_1$ y $M_2$.
   * Añadir transiciones $\varepsilon$ desde cada estado final de $M_1$ al estado inicial de $M_2$.
   * El estado inicial es el de $M_1$; los estados finales son los de $M_2$.

3. **Estrella de Kleene ($L^*$)**

   * Partir de $M$.
   * Añadir un nuevo estado inicial/final $q'$.
   * Conectar $q'$ → inicio$(M)$ con $\varepsilon$, y de cada final de $M$ de vuelta a inicio$(M)$ con $\varepsilon$.

4. **Clausura positiva ($L^+$)**

   * Igual que la estrella, pero sin permitir la transición $\varepsilon$ inicial → final directa; obliga a usar **al menos** una copia de $M$.

5. **Complemento ($\overline{L}$)**

   * Primero convertir $M$ en un **AFD completo** (si no lo es ya):

     * Asegurarse de que cada estado tenga transición definida para cada símbolo (si falta, apuntar a un “sumidero”).
   * Luego, intercambiar los estados finales y no finales.

6. **Intersección ($L_1\cap L_2$)**

   * Construir el **autómata producto** (ver Teorema 3.3.2 más abajo).
   * Estados: parejas $(q_1,q_2)$.
   * Transición conjunta: sobre símbolo $a$, cada componente avanza según su función delta.
   * Finales: $(q_1,q_2)$ donde $q_1$ y $q_2$ eran finales.

7. **Diferencia ($L_1 - L_2$)**

   * Usar el complemento y la intersección:

     $$
     $$

> $$L_1 - L_2 \;=\; L_1 \,\cap\, \overline{L_2}.··


8. **Diferencia simétrica ($L_1\triangleleft L_2$)**

   * Se define como

     $$
     $$

> $$(L_1 - L_2) \;\cup\; (L_2 - L_1).$$

* Basta combinar construcciones de diferencia y luego de unión.


### **5.3.2 Teorema (Intersección de AFD)**

> **Enunciado.**
> Dados dos AFD
>
> $$
> M_1=(\Sigma, Q_1, q_1, F_1, \delta_1),
> \quad
> M_2=(\Sigma, Q_2, q_2, F_2, \delta_2),
> $$
>
> podemos construir un AFD
>
> $$
> M = (\Sigma,\;Q_1\times Q_2,\;(q_1,q_2),\;F_1\times F_2,\;\delta)
> $$
>
> donde la función de transición
>
> $$
> \delta\bigl((q_i,q_j),\,a\bigr)
> \;=\;
> \bigl(\delta_1(q_i,a),\;\delta_2(q_j,a)\bigr).
> $$
>
> Este autómata reconoce
>
> $$
> L(M)\;=\;L(M_1)\;\cap\;L(M_2).
> $$

**Explicación:**

* **Estados:** cada uno representa “estar simultáneamente” en un estado de $M_1$ **y** en un estado de $M_2$.
* **Inicio:** la pareja de los inicios de $M_1$ y $M_2$.
* **Finales:** aquellas parejas donde **ambos** componentes sean estados finales en sus autómatas originales.
* **Movimiento:** al leer un símbolo $a$, cada componente avanza según su propio autómata.

Con esta construcción “paralela”, $M$ sólo acepta si **ambos** $M_1$ y $M_2$ aceptarían la misma cadena, lo que implementa exactamente la intersección.

## **5.4. Homomorfismos**

### **5.4.1 Definición**

Un **homomorfismo** es una forma de **traducir** cadenas de un alfabeto $\Sigma$ a cadenas de otro alfabeto $\Gamma$ (que puede ser igual a $\Sigma$), letra por letra.

* Se define una función

  $$
    h:\Sigma^*\;\longrightarrow\;\Gamma^*
  $$

  tal que:

  1. $h(\lambda)=\lambda$ (la cadena vacía se mantiene vacía).
  2. Si $u=a_1a_2\cdots a_n$ con cada $a_i\in\Sigma$, entonces

     $$
       h(u)\;=\;h(a_1)\;h(a_2)\;\cdots\;h(a_n).
     $$
* Basta especificar **qué hace** $h$ con cada símbolo de $\Sigma$; el resto se extiende automáticamente letra a letra.


**- Ejemplo**

* $\Sigma=\{a,b,c\}$, $\Gamma=\{0,1\}$.
* Definimos

  $$
    h(a)=0,\quad h(b)=1,\quad h(c)=010.
  $$
* **¿Cómo funciona?**

  1. Para cada $a$ aparece un “0”.
  2. Para cada $b$ aparece un “1”.
  3. Para cada $c$ aparece la secuencia “010”.
* **Algunos cálculos:**

  $$
    h(a^2 b^2)
    =h(a)\,h(a)\,h(b)\,h(b)
    =0\;0\;1\;1
    =0011.
  $$

  $$
    h(a\,c\,b^2\,c)
    =h(a)\;h(c)\;h(b)\;h(b)\;h(c)
    =0\;(010)\;1\;1\;(010)
    =001011010.
  $$
* En forma de lenguaje:

  $$
    h\bigl(a^* b^* c^*\bigr)
    =0^*\,1^*\,\bigl(010\bigr)^*.
  $$

**- Ejemplo: Código Morse**

Imaginemos que queremos traducir **textos simples** al alfabeto de puntos y rayas del **Código Morse**.

1. **Alfabeto de partida** $\Sigma =\{\texttt{A},\dots,\texttt{Z},\texttt{0},\dots,\texttt{9}\}$.
2. **Alfabeto de llegada** $\Gamma =\{\,\texttt{.},\texttt{-}\}$.
3. Definimos el homomorfismo $m:\Sigma^*\to\Gamma^*$ que a cada letra asigna su secuencia Morse. Por ejemplo:

   $$
     m(\texttt{A}) = \texttt{.-},\quad
     m(\texttt{B}) = \texttt{-...},\quad
     m(\texttt{S}) = \texttt{...},\quad
     m(\texttt{1}) = \texttt{.----},\;\dots
   $$
4. Para traducir una palabra como **“SOS”**, aplicamos:

   $$
     m(\texttt{SOS})
     = m(\texttt{S})\,m(\texttt{O})\,m(\texttt{S})
     = \texttt{...}\;\texttt{---}\;\texttt{...}
     = \texttt{...---...}.
   $$
5. Si queremos traducir una frase entera, basta concatenar la traducción de cada símbolo, y (si lo deseamos) insertar separadores entre letras o palabras.

Este procedimiento es **exactamente** un homomorfismo: cada carácter se sustituye por una pequeña cadena de puntos y rayas, y la traducción de una cadena entera es la **concatenación** de esas pequeñas cadenas.

**Resumen:**

* Un homomorfismo **reemplaza** símbolo a símbolo, extendiéndose por concatenación.
* Con el ejemplo dado, transformamos $a,b,c$ en “0”, “1” y “010”.
* Con el ejemplo cotidiano, transformamos letras y dígitos en señales Morse.
* La gran ventaja es que, si partimos de un lenguaje **regular**, su imagen homomorfa seguirá siendo **regular**.



### **5.4.2 Teorema**


Sea $h: \Sigma^* \to \Gamma^*$ un homomorfismo.

**(1) Propiedades básicas de $h$**

Para **cadenas** $u,v\in\Sigma^*$ y **lenguajes** $A,B\subseteq\Sigma^*$, valen las siguientes igualdades:

1. **Concatenación de cadenas**

   $$
     h(u\,v)\;=\;h(u)\;h(v).
   $$

   > *Significado:* Traducir primero $u$ y luego $v$ es lo mismo que traducir la cadena completa $u\,v$.

2. **Unión de lenguajes**

   $$
     h(A\;\cup\;B)
     \;=\;
     h(A)\;\cup\;h(B).
   $$

   > *Significado:* La imagen homomorfa de todo lo que está en $A$ o en $B$ son las cadenas que vienen de $A$ más las que vienen de $B$.

3. **Concatenación de lenguajes**

   $$
     h(A\,B)
     \;=\;
     h(A)\;h(B).
   $$

   > *Significado:* Si `AB = {xy | x en A, y en B}`, traducir cada par $(x,y)$ y luego juntar es lo mismo que juntar primero todas las traducciones de $A$ con todas las de $B$.

4. **Estrella de Kleene**

   $$
     h\bigl(A^*\bigr)
     \;=\;
     \bigl[h(A)\bigr]^*.
   $$

   > *Significado:* Repetir (cero o más veces) cadenas de $A$ y luego traducir es igual a repetir (cero o más veces) las traducciones de $A$.

**(2) Conservación de la regularidad**

> Si $L\subseteq\Sigma^*$ es un **lenguaje regular** (reconocible por un autómata finito), entonces su imagen homomorfa
>
> $$
>   h(L) \;=\;\{\,h(w):w\in L\}
> $$
>
> **también** es regular sobre $\Gamma$.

**Ampliando la explicación**

1. **Extender letra a letra**
   Por definición, un homomorfismo traduce cada símbolo de $\Sigma$ a una pequeña “frase” en $\Gamma^*$. Para traducir una cadena, simplemente encadena esas frases en el mismo orden.

2. **Por qué respetan las operaciones**

   * **Concatenar antes o después** de traducir no cambia el resultado: la traducción “se distribuye” sobre la concatenación.
   * **Unir** dos conjuntos de cadenas y luego traducir da el mismo conjunto que traducir cada uno por separado y unir los resultados.
   * **Repetir** cadenas también conmute con la traducción: traducir una secuencia repetida equivale a repetir las traducciones.

3. **Preservación de la regularidad**

   * Dado un **automáta finito** que reconoce $L$, podemos construir otro automáta finito que reconozca $h(L)$ simplemente “expandiendo” cada transición original: cada vez que el primer autómata lee un símbolo $a$, el segundo debe leer la secuencia $h(a)$.
   * Así, cualquier lenguaje regular bajo $\Sigma$ se transforma en otro lenguaje que sigue siendo reconocible por un autómata de tamaño finito.

**Importancia práctica:**

* Nos asegura que, si empezamos con un lenguaje fácil de reconocer (regular), al traducir su alfabeto —por ejemplo, codificar caracteres, comprimir información, cambiar notaciones— seguimos en el ámbito de los lenguajes regulares, sin complicarnos con autómatas infinitos ni construcciones más potentes.



La parte (2) del Teorema 5.4.1 nos permite, **por contradicción**, demostrar que un lenguaje **no** es regular de forma muy elegante:

1. **Supongamos** que $L$ es regular.
2. Entonces, por el teorema, su **imagen homomorfa** $h(L)$ también debe ser regular.
3. Pero si elegimos un homomorfismo $h$ tal que $h(L)$ sea un lenguaje que **ya sabemos** no es regular, llegamos a una contradicción.
4. Concluimos que $L$ **no** puede ser regular.

**Ejemplo 1**

* **Lenguaje a estudiar**

  $$
    L \;=\;\{\,0^i1^i : i\ge1\}.
  $$
* **Homomorfismo**

  $$
    h(0)=a,\quad h(1)=b.
  $$
* **Imagen**

  $$
    h(L)
    = \{\,h(0)^i\,h(1)^i : i\ge1\}
    = \{\,a^i b^i : i\ge1\}.
  $$
* Sabemos por la **sección 3.1** que $\{a^i b^i\}$ **no** es regular.
* **Contradicción:** si $L$ fuese regular, $h(L)$ también lo sería.
* **Conclusión:** $L=\{0^i1^i\}$ no es regular.

**Ejemplo 2**

* **Lenguaje a estudiar**

  $$
    L \;=\;\{\,0^n2\,1^n : n\ge1\}
  $$

  (cadenas de ceros, un “2” en medio, y luego el mismo número de unos).
* **Homomorfismo**

  $$
    h(0)=0,\quad h(1)=1,\quad h(2)=\lambda\quad(\text{la cadena vacía}).
  $$
* **Imagen**

  $$
    h(L)
    = \{\,0^n\,\lambda\,1^n : n\ge1\}
    = \{\,0^n1^n : n\ge1\},
  $$

  que acabamos de ver **no** es regular.
* **Conclusión:** $L=\{0^n2\,1^n\}$ no es regular.

**Ejemplo 3: Chequeo de corchetes marcados**

Sea el lenguaje

$$
  L \;=\;\{\,(^{n}\,[\,)^{n}\,] : n\ge1\},
$$

esto es, cadenas que consisten en:

1. $n$ paréntesis abiertos “$($”,
2. un corchete “$[$”,
3. $n$ paréntesis cerrados “$)$”,
4. y finalmente un corchete de cierre “$]$”.

**¿Es $L$ regular?**

1. Definimos el homomorfismo

   $$
     h(()= (,\quad h())= ),\quad h([\!]=\lambda,\quad h(])=\lambda.
   $$

   Es decir, “borramos” los corchetes y dejamos sólo los paréntesis.  

2. La **imagen** es

$$
  h(L)
  \;=\;
  \bigl\{\,
    \underbrace{(\,(}_{n}\dots\underbrace{)\,)}_{n}
    \;\colon\;n\ge1
  \bigr\}
  \;=\;
  \bigl\{\,
    \underbrace{(\dots(}_{n}
    \underbrace{)\dots)}_{n}
    : n\ge1
  \bigr\}.
$$

   que es precisamente el lenguaje de **paréntesis equilibrados** $\{(^n)^n\}$, bien conocido **no** regular.  

3. Por contradicción, $L$ **no** puede ser regular, pues de otro modo $h(L)$ sí lo sería.

**Resumen de la técnica:**

* Elegimos $h$ que transforme nuestro lenguaje $L$ en otro $h(L)$ de forma sencilla (p. ej. reemplazando un símbolo por vacío).
* Si $h(L)$ es un caso ya probado como no regular, concluimos que $L$ tampoco lo es.
* Así **ahorramos** tener que aplicar directamente el Lema de Bombeo al lenguaje original.

## **5.5 Imagen inversa de un homomorfismo**

Dado un homomorfismo

$$
  h:\Sigma^* \;\longrightarrow\;\Gamma^*
$$

y un lenguaje $B\subseteq\Gamma^*$, la **imagen inversa** de $B$ bajo $h$ se define como

$$
  h^{-1}(B)
  \;:=\;
  \{\,u\in\Sigma^* : h(u)\in B\}.
$$

Es decir, es el conjunto de todas las cadenas en $\Sigma^*$ que, al aplicarles $h$, caen dentro de $B$.


### **5.5.1 Teorema**

> **Enunciado.**
> Sea $h:\Sigma^*\to\Gamma^*$ un homomorfismo y $B\subseteq\Gamma^*$ un **lenguaje regular**. Entonces
>
> $$
>   h^{-1}(B)
> $$
>
> **también** es un lenguaje regular (pero ahora sobre $\Sigma$).

1. **Interpretación de $h^{-1}(B)$.**

   * Piensa en $h$ como un traductor de mensajes: cada palabra (cadena) en $\Sigma^*$ termina traducida en $\Gamma^*$.
   * El conjunto $B$ es un “club” de mensajes en $\Gamma^*$.
   * La imagen inversa $h^{-1}(B)$ es **el grupo de mensajes originales** (antes de traducir) que, al pasar por $h$, terminan **sí o sí** dentro del club $B$.

2. **¿Por qué sigue siendo regular?**

   * Dado un autómata (DFA) que reconoce $B$, podemos construir otro autómata que, **en lugar de leer directamente símbolos de $\Gamma$**, simule la lectura de cada $h(a)$ al recibir el símbolo $a\in\Sigma$.
   * Concretamente, cada transición que en el DFA de $B$ va sobre un símbolo $x\in\Gamma$, la reemplazamos por un pequeño sub-autómata (o varias transiciones $\varepsilon$) que consumen exactamente la cadena $x = h(a)$.
   * El resultado es un autómata finito sobre $\Sigma$ que acepta precisamente las cadenas $u$ cuya traducción $h(u)$ era aceptada en $B$.
   * Como construimos un **autómata finito**, $h^{-1}(B)$ es regular.

**Ejemplo: Números escritos en inglés**

1. **Alfabeto de entrada**
   $\Sigma = \{\texttt{0},\texttt{1},\dots,\texttt{9}\}$ (dígitos).

2. **Alfabeto de salida**
   $\Gamma =$ letras de la A a la Z.

3. **Homomorfismo $h$:**
   Convierte cada dígito en su nombre en inglés:

   $$
     h(\texttt{0})=\texttt{zero},\;h(\texttt{1})=\texttt{one},\;\dots,\;h(\texttt{9})=\texttt{nine}.
   $$

   Y para una secuencia de dígitos, p. ej. “$\texttt{314}$”,

   $$
     h(\texttt{314})
     =\texttt{three}\;\texttt{one}\;\texttt{four}
     =\texttt{threeonefour}.
   $$

4. **Lenguaje regular $B\subseteq\Gamma^*$:**
   Las palabras en inglés de **exactamente 3 letras**,

   $$
     B = \{\,w\in\{\texttt{a–z}\}^* : |w|=3\}.
   $$

   (Es sencillo construir un DFA que acepte justo las cadenas de longitud 3.)

5. **Imagen inversa**

   $$
     h^{-1}(B)
     = \{\,u\in\Sigma^* : h(u)\in B\}.
   $$

   ¿Qué dígitos o combinaciones de dígitos producen nombres de longitud 3?

   $$
     h^{-1}(B) = \{\texttt{1},\texttt{2},\texttt{6},\texttt{10}\},
   $$

   porque

   * “one” (3 letras) ← “1”
   * “two” (3 letras) ← “2”
   * “six” (3 letras) ← “6”
   * “ten” (3 letras) ← “10”
     Ningún otro número de un dígito ni dos dígitos da un nombre de longitud 3.

6. **Conclusión:**

   * Dado que $B$ (las cadenas de longitud 3) es regular, su **imagen inversa** bajo $h$ —el conjunto $\{1,2,6,10\}$— **también** es regular.
   * Efectivamente, $\{1,2,6,10\}$ es finito, y por tanto trivialmente reconocible por un autómata finito.
**Resumen:**

* La imagen inversa $h^{-1}(B)$ recoge todas las entradas que, tras traducir por $h$, quedan en $B$.
* Construyendo un autómata que “deshace” la traducción, vemos que si $B$ era regular, entonces $h^{-1}(B)$ lo sigue siendo.
* Esta propiedad es extremadamente valiosa tanto para construir nuevos lenguajes reconocibles como para razonar, por ejemplo, sobre pruebas de no regularidad.


## **5.6 Algoritmos de decisión**

### **5.6.1 Introducción**

Un **algoritmo** —en el sentido que usaremos aquí— es un procedimiento claro y preciso que, para **cualquier** entrada, **termina** después de un número finito de pasos, entregando un resultado.

### **Problema de decisión**

<p float="center">
  <img src="https://github.com/carlosalvarezh/TeoriaComputacion/blob/main/images/TC05_AFN20.PNG?raw=true" width="500" />
</p> 

* Tenemos una **propiedad** $\mathcal{P}$ que se puede preguntar sobre **autómatas** (por ejemplo: “¿rechaza todas las cadenas?”, “¿tiene al menos un ciclo?”, etc.).
* El **problema de decisión** para $\mathcal{P}$ consiste en diseñar un **algoritmo** que tome **cualquier** autómata $M$ y responda **SI** o **NO** a la pregunta:

> **“¿Satisface $M$ la propiedad $\mathcal{P}$?”**

* Si **existe** tal algoritmo, decimos que el problema $\mathfrak{P}$ es **decidible** (o **resoluble**).
* Si **no** existe ningún algoritmo que funcione para **todos** los autómatas, $\mathfrak{P}$ es **indecidible** (o **irresoluble**).

> **Importante:** No basta saber responder sí/no para unos pocos ejemplos de autómatas. El algoritmo debe funcionar para **cualquier** autómata en un tiempo finito.

### Dos ejemplos contrastantes

1. **Problema de pertenencia**

   $$
   \mathcal{P}_{\text{memb}}:
   \begin{cases}
     \text{Entrada:} & M\ (\text{autómata})\;\text{y}\;w\in\Sigma^*,\\
     \text{Pregunta:} & \text{¿acepta }M\;la\;cadena\;w\;?
   \end{cases}
   $$

   * **Solución sencilla:**

     1. Convertir $M$ en un AFD $M'$ (si no lo es).
     2. Simular $M'$ sobre $w$.
     3. Si llega a un estado final, responde **SI**; si no, responde **NO**.
   * Este problema es **decidible**, pues el algoritmo siempre termina después de leer las $|w|$ posiciones de la cadena.

2. **Problema de vacuidad**

   $$
   \mathcal{P}_{\varnothing}:
   \begin{cases}
     \text{Entrada:} & M\ (\text{autómata}),\\
     \text{Pregunta:} & \text{¿}L(M)=\varnothing\;?\;
       \bigl(\text{¿acepta \(M\) \emph{alguna} cadena?}\bigr)
   \end{cases}
   $$

   * **No sirve** la idea de probar todas las cadenas $w\in\Sigma^*$ —¡son infinitas!—, ni simular una por una.
   * Se necesita un **nuevo método** (por ejemplo, analizar la estructura del grafo de estados y ver si existe un camino desde el inicial a algún estado final).

### **5.6.2 Teorema (Decidibilidad de la vacuidad)**

Sea $\Sigma$ un alfabeto dado. Existe un algoritmo para el siguiente problema de decisión referente a autómatas sobre $\Sigma$ :  

> Dado un autómata (AFD o AFN o AFN- $\lambda$ ) $M, ¿Es L(M) \neq \varnothing$ ? (es decir, ¿acepta $M$ alguna cadena?).

### **5.6.3 Idea del algoritmo (esquema)**

1. **Construir** el grafo dirigido de $M$:

   * **Nodos:** los estados del autómata.
   * **Aristas:** cada transición con etiqueta real o $\varepsilon$.
2. **Buscar** (por ejemplo, con DFS o BFS) si existe **algún** camino desde el **estado inicial** hasta **algún** estado final.
3. **Si** lo hay, significa que existe alguna cadena que lleva al autómata a un final ⇒ **$L(M)\neq\varnothing$**.
4. **Si no**, el autómata nunca llega a un estado final ⇒ **$L(M)=\varnothing$**.
5. El recorrido de un grafo finito con DFS/BFS siempre termina en pasos finitos ⇒ es un **algoritmo** de decisión.

**Resumen:**

* Un **problema de decisión** pregunta “¿sí/no?” para **cualquier** autómata.
* **Decidible** significa que hay un **algoritmo** que siempre lo responde en tiempo finito.
* **Indecidible** significa que **no existe** tal algoritmo general.
* El **problema de pertenencia** y el **problema de vacuidad** son dos ejemplos clásicos: el primero se resuelve simulando la cadena, el segundo usando búsqueda en el grafo de estados.


### **5.6.4 Corolario.**

Sea $\Sigma$ un alfabeto cualquiera. A partir de autómatas sobre $\Sigma$ podemos resolver de forma automática dos problemas muy importantes:

1. **¿Es $L(M_1)\subseteq L(M_2)$?**
2. **¿Es $L(M_1)=L(M_2)$?**

Para ambos casos primero necesitamos un **subprocedimiento** que calcule, en cualquier autómata $M$, el conjunto de **estados accesibles** desde su estado inicial:

#### Subrutina: Estados accesibles (ALC)

```algorithm
INICIALIZAR:
  ALC := { q₀ }    ← q₀ = estado inicial de M

REPETIR:
  para cada q ∈ ALC:
    mira todas las transiciones q —[a]→ p  (a puede ser un símbolo o ε)
    añade p a ALC

HASTA:
  no se añada ningún estado nuevo en una pasada
```

> **¿Qué hace?**
>
> 1. Comienza con el único estado que seguro “se usa”: el inicial.
> 2. Itera: por cada estado que ya está en ALC, agrega todos sus destinos.
> 3. Repite hasta que no queden más estados nuevos.
> 4. Al terminar, **ALC** contiene exactamente los estados que **se pueden alcanzar** desde el inicio con alguna cadena.


#### **(1) Decidir si $L(M_1)\subseteq L(M_2)$**

Para comprobar que **toda** cadena que acepta $M_1$ también la acepta $M_2$, seguimos este esquema:

1. **Construir** un autómata para $\overline{L(M_2)}$ (el **complemento** de $M_2$).
2. **Formar** el autómata de **intersección**

   $$
     M_I \;=\; M_1 \;\cap\; \bigl(\text{complemento de }M_2\bigr).
   $$
3. **Calcular** los **estados accesibles** de $M_I$ con la subrutina ALC.
4. **Determinar** si alguno de estos accesibles es **estado final** de $M_I$:

   * Si sí, existe una cadena que está en $L(M_1)$ pero **no** en $L(M_2)$: **NO** es subconjunto.
   * Si no, ninguna cadena rompe la inclusión: **SÍ**, $L(M_1)\subseteq L(M_2)$.

> **Idea clave:**
> $L(M_1)\subseteq L(M_2)$ equivale a
>
> $$
>   L(M_1)\;\cap\;\overline{L(M_2)}
>   \;=\;\varnothing.
> $$
>
> Verificamos vacuidad del lenguaje resultante usando los estados accesibles.

#### **(2) Decidir si $L(M_1)=L(M_2)$**

Para igualdad de lenguajes basta comprobar **inclusión mutua**:

1. Verificar $L(M_1)\subseteq L(M_2)$ con el método anterior.
2. Verificar $L(M_2)\subseteq L(M_1)$ intercambiando roles.
3. Si **ambas** inclusiones son verdaderas, entonces $L(M_1)=L(M_2)$. De lo contrario, no.

> **Resumen:**
>
> * El subprocedimiento de **estados accesibles** (ALC) sirve para detectar, en cualquier autómata, si hay caminos que llevan a un estado final.
> * Con él podemos **decidir** vacuidad, inclusión y, por extensión, igualdad de los lenguajes que los autómatas reconocen.
> * **Todo** se reduce a construir automóviles derivados (complemento, intersección) y aplicar ALC: es un **algoritmo** que siempre termina y contesta **SI** o **NO**.


### **5.6.5 Teorema**

Sea $M$ un **AFD** con $n$ estados y sea $L = L(M)$.

> **Afirmación.**
> $L$ es **infinito** (contiene infinitas cadenas) **si y solo si** existe alguna cadena $w$ aceptada por $M$ cuya longitud satisface
>
> $$
>   n \;\le\; |w| \;<\;2n.
> $$

#### **Explicación**

1. **“Sólo si”** ($\Rightarrow$):

   * Si $L$ es infinito, $M$ acepta cadenas de **longitudes arbitrariamente grandes**.
   * Entre todas estas cadenas existe alguna mínima con longitud $\ge n$.
   * Pero por el **Pumping Lemma** aplicado a autómatas (o por el hecho de que cualquier recorrido de más de $n$ pasos debe repetir un estado), podemos “bombear” esa cadena para generar infinitas más, todas con longitud **entre** $n$ y $2n-1$.
   * En particular, hay al menos **una** aceptada en ese rango.

2. **“Si”** ($\Leftarrow$):

   * Si encontramos un $w$ con $n\le|w|<2n$, entonces al recorrer $M$ sobre $w$ hemos visitado $n$ transiciones en $n$ o más pasos, lo que **forzosamente** implica un **ciclo** (dos veces el mismo estado).
   * Al “repetir” ese ciclo (bombearlo), obtenemos **infinitas** cadenas distintas, todas aceptadas.
   * Por tanto, $L$ no puede ser finito; debe ser infinito.

> **Conclusión:**
> Basta buscar **una sola** cadena en el rango de longitudes $[n,\,2n)$ para decidir si $L$ es infinito.


### **5.6.6 Corolario.**

Sea $\Sigma$ un alfabeto dado.

> **Problema de decisión:**
> “Dado un autómata $M$ (AFD, AFN o AFN–$\lambda$), ¿es $L(M)$ **infinito**?”

**Afirmación:** Existe un **algoritmo** que responde **SI** o **NO** a esta pregunta en tiempo finito.

#### Esquema del algoritmo

1. Contar cuántos estados $n$ tiene $M$.
2. **Recorrer** $M$ en **anchura** (BFS) o **profundidad** (DFS) para buscar alguna cadena aceptada de longitud entre $n$ y $2n-1$:

   * Mientras generamos cadenas crecientes, detenemos la búsqueda si encontramos una de longitud $k$ tal que $n\le k<2n$ y que lleve a un estado final.
3. * Si la encontramos → **SI**, $L(M)$ es infinito.
   * Si agotamos **todas** las posibles rutas de longitud menor que $2n$ sin éxito → **NO**, $L(M)$ es finito.

> **Nota:** El espacio de búsqueda hasta longitud $2n-1$ es **finito**, y BFS/DFS sobre el grafo de estados termina en tiempo polinomial en $n$.


### **5.6.7 Extensión a expresiones regulares**

* Un problema de decisión sobre **lenguajes regulares** (representados como expresiones regulares $R$) es:

  > “Dada una expresión $R$, ¿es $L(R)$ infinito?”
* Como **sabemos convertir** una expresión regular en un autómata finito (y viceversa), **todo** algoritmo decidible para autómatas se traslada a expresiones regulares.

> **Corolario práctico:**
> Gracias al Corolario 3.6.4, el problema
>
> $$
>   \text{“Dada una expresión regular \(R\), ¿es \(L(R)\) infinito?”}
> $$
>
> es **decidible**.


**Resumen:**

* El Teorema 5.6.5 reduce la pregunta “¿$L(M)$ es infinito?” a buscar una **cadena de longitud entre $n$ y $2n-1$**.
* El Corolario 5.6.6 convierte este criterio en un **algoritmo** de decisión efectivo tanto para **autómatas** como para **expresiones regulares**.
